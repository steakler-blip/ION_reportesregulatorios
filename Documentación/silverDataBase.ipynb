# [dbo].[001_ENT_CONSUMO]

```sql
/****** Object:  StoredProcedure [dbo].[001_ENT_CONSUMO]    Script Date: 01/06/2026 20:11:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[001_ENT_CONSUMO] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[001_ENT_CONSUMO]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[001_ENT_CONSUMO]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[001_ENT_CONSUMO]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END

		INSERT INTO [SILVER].[RR].[001_ENT_CONSUMO] (
			[ID_PERSONA],
			[ID_CREDITO],
			[PRODUCTO],
			[FECHAALTA],
			[FECHADISPREST],
			[FECHAFINREST],
			[FECHAVENCIMIENTO],
			[EDO_CTA_PAPEL],
			[TIPORESTRUCTURA],
			[NUMDISPOSICIONES],
			[PERIODICIDADFACTURACION],
			[PLAZOTOTAL],
			[PLAZOREMANENTE],
			[IMPORTEORIGINAL],
			[TASAINTERES],
			[MEDIOADQUISICION],
			[QUITASCONDONACIONES],
			[BONIFICACIONESDESCUENTOS],
			[SALDOCAPITALVIGENTE],
			[SALDOINTERESVIGENTE],
			[SALDOCAPITALVENCIDO],
			[SALDOINTERESVENCIDO],
			[INTERESCARTERAVENCIDA],
			[FECHACORTE],
			[MONTOEXIGIBLE],
			[MONTOEXIGIBLECAP],
			[MONTOEXIGIBLEINT],
			[MONTOEXIGIBLECOM],
			[MONTOEXIGIBLEFALTAPAG],
			[MONTOEXIGIBLEIVAOTROS],
			[PAGOREALIZADO],
			[PAGOREALIZADOCAP],
			[PAGOREALIZADOINT],
			[PAGOREALIZADOCOM],
			[PAGOREALIZADOFALTAPAG],
			[PAGOREALIZADOIVAOTR],
			[CODIGOCREDITOREESTRUCTURADO],
			[INDGAR],
			[DIASATRASO],
			[PORPAGO],
			[ATR],
			[MAXATR],
			[MESES_DESDE_ULT_ATR_BK],
			[DELIP],
			[TIPO_MET],
			[INDICADOR_SIC],
			[MESES_CONSULTA_SIC],
			[ETAPADETERIORO],
			[TASA_COBRADA],
			[ANTIGUEDADACREDITADO],
			[MONEDA],
			[CAT],
			[MONTOPAGARSIC],
			[MESSIC],
			[RVAS_ADICIONALES],
			[SALDO_REP_SIC],
			[HIPOTESISPRESREF],
			[FECHABAJA],
			[TIPOBAJA],
			[TIPOCREDITO],
			[FECHA_INFO],
			[FACT_AJUSTE_GENERO],
			[CAMBIO_ETAPA],
			[PERIODICIDAD_CAPITAL]
			)
	SELECT
		CR.COD_CLIENTE AS ID_PERSONA,
		CR.ID_EXTERNO AS ID_CREDITO,
		'CPLA04' AS PRODUCTO, -- Valor fijo definido por operación
		CR.FEC_APERTURA AS FECHAALTA,
		IIF(CR.IND_ESTADO_CRED = 'E', CR.FEC_APERTURA, CR.FEC_ULT_DESEMBOLSO) AS FECHADISPREEST,
		CR.FEC_VENCIMIENTO AS FECHAFINREEST,
		CR.FEC_VENCIMIENTO AS FECHAVENCIMIENTO,
		0 AS EDO_CTA_PAPEL, -- Valor de catálogo fijo (no hay estado de cuenta en papel)
		50 AS TIPOREESTRUCTURA, -- Valor de catálogo fijo para consumo (estructura simple)
		ISNULL((SELECT COUNT(DISTINCT A.ID_SECUENCIA)
			FROM BRONZE.SAF.PR_ENCABEZADO_PAGO A 
			INNER JOIN BRONZE.SAF.PR_DETALLE_PAGO B  ON A.ID_SECUENCIA = B.ID_SECUENCIA
			WHERE TIP_TRANSACCION = 3 AND COD_ESTADO = 'A' AND CAST(A.NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT)
			GROUP BY A.NUM_CREDITO),0) AS NUMDISPOSICIONES, 
		CASE
			WHEN CR.PER_CUOTA = 'SE' THEN 10
			WHEN CR.PER_CUOTA = 'ME' THEN 50
			ELSE 50
		END AS PERIODICIDADFACTURACION, -- catálogo
		CR.PLAZO_CREDITO AS PLAZOTOTAL, 
		DATEDIFF(MONTH, @FechaSistema, CR.FEC_VENCIMIENTO) AS PLAZOREMANENTE,
		CR.MON_CREDITO AS IMPORTEORIGINAL,
		CR.TASA_INTERES AS TASAINTERES,
		1 AS MEDIOADQUISICION, -- Valor fijo. Catálogo Medioadquisicion
		ISNULL(COND.MONTO_CONDONADO, 0) AS QUITASCONDONACIONES,
		ISNULL(BON.MONTO_BONIFICADO, 0) AS BONIFICACIONESDESCUENTOS, -- Desde subconsulta de bonificaciones
		ISNULL(IND.SALDO_CAPITAL_VIGENTE, 0) AS SALDOCAPITALVIGENTE,
		ISNULL(IND.SALDO_INTERES_VIGENTE, 0) AS SALDOINTERESVIGENTE,
		ISNULL(IND.SALDO_CAPITAL_VENCIDO, 0) AS SALDOCAPITALVENCIDO,
		ISNULL(IND.SALDO_INTERES_VENCIDO, 0) AS SALDOINTERESVENCIDO,
		ISNULL(IND.INTERES_CARTERA_VENCIDA, 0) AS INTERESCARTERAVENCIDA,
		EOMONTH(@FechaSistema) AS FECHACORTE,
		ISNULL((
			SELECT ISNULL(SUM(SAL_RUBRO),0) + ISNULL(SUM(SAL_IVA),0)
			FROM BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO
			WHERE IND_ESTADO = 'P'
			  AND FEC_COBRO < @FechaFin
			  AND CAST(NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT)
			GROUP BY NUM_CREDITO
		), 0) AS MONTOEXIGIBLE,
	
		ISNULL((
			SELECT SUM(SAL_RUBRO)
			FROM BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO
			WHERE IND_ESTADO = 'P'
			  AND FEC_COBRO < @FechaFin
			  AND CAST(NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT)
			  AND COD_RUBRO = 'PRINCIPAL'
			GROUP BY NUM_CREDITO
		), 0) AS MONTOEXIGIBLECAP,

		ISNULL((
			SELECT SUM(SAL_RUBRO)
			FROM BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO
			WHERE IND_ESTADO = 'P'
			  AND FEC_COBRO < @FechaFin
			  AND CAST(NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT)
			  AND COD_RUBRO = 'INTERESES'
			GROUP BY NUM_CREDITO
		), 0) AS MONTOEXIGIBLEINT,

		ISNULL((
			SELECT SUM(SAL_RUBRO)
			FROM BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO
			WHERE IND_ESTADO = 'P'
			  AND FEC_COBRO < @FechaFin
			  AND CAST(NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT)
			  AND COD_RUBRO IN ('SEGURO_DANOS', 'SEGURO_VIDA'
				  /*'ASE_FINANCIERA', 'COM_PRORROGA', 'COM_REESTRUCTURA',
				  'OPI_VALOR', 'VISITA_EXTRA'*/)
			GROUP BY NUM_CREDITO
		), 0)AS MONTOEXIGIBLECOM,

		ISNULL((
			SELECT SUM(SAL_RUBRO)
			FROM BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO
			WHERE IND_ESTADO = 'P'
			  AND FEC_COBRO < @FechaFin
			  AND CAST(NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT)
			  AND COD_RUBRO IN ('COMISION_ADMINISTRACION', 'MORATORIOS')
			GROUP BY NUM_CREDITO
		), 0) AS MONTOEXIGIBLEFALTAPAG,
	---- Monto exigible de IVA otros
		ISNULL((
			SELECT SUM(SAL_IVA)
			FROM BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO
			WHERE IND_ESTADO = 'P'
			  AND FEC_COBRO < @FechaFin
			  AND CAST(NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT)
			GROUP BY NUM_CREDITO
		), 0) AS MONTOEXIGIBLEIVAOTROS,

		---- Pago realizado total
		ISNULL((
			SELECT SUM(B.MON_PAGO)
			FROM BRONZE.SAF.PR_ENCABEZADO_PAGO A
			INNER JOIN BRONZE.SAF.PR_DETALLE_PAGO B
				ON A.ID_SECUENCIA = B.ID_SECUENCIA
			WHERE 
				CAST(A.NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT)
				AND A.ID_CUENTA NOT IN (
					'140103009001', '140103009002', '140103009003',
					'1401030090010000', '1401030090020000',
					'5050260901000300', '2073','2074','2068','2061','2056'
				)
				AND NOT (TIP_TRANSACCION = 4 AND SUBTIP_TRANSAC = 5)
				AND FEC_MOVIMIENTO >= @FechaIni
				AND FEC_MOVIMIENTO < @FechaFin
			GROUP BY A.NUM_CREDITO
		), 0) AS PAGOREALIZADO,

		---- Pago a capital
		--truncate table [SILVER].[RR].[001_ENT_CONSUMO]
		ISNULL((
			SELECT SUM(B.MON_PAGO)
			FROM BRONZE.SAF.PR_ENCABEZADO_PAGO A
			INNER JOIN BRONZE.SAF.PR_DETALLE_PAGO B
				ON A.ID_SECUENCIA = B.ID_SECUENCIA
			WHERE 
				CAST(NUM_CREDITO AS INT) = CR.NUM_CREDITO
				AND A.ID_CUENTA NOT IN ('140103009001', '140103009002', '140103009003',
					'1401030090010000', '1401030090020000',
					'5050260901000300', '2073','2074','2068','2061','2056')
				AND NOT (TIP_TRANSACCION = 4 AND SUBTIP_TRANSAC = 5)
				AND COD_CONCEPTO = 'PRINCIPAL'
				AND FEC_MOVIMIENTO >= @FechaIni
				AND FEC_MOVIMIENTO < @FechaFin
			GROUP BY A.NUM_CREDITO
		), 0) AS PAGOREALIZADOCAP,

		---- Pago a intereses
		--truncate table [SILVER].[RR].[001_ENT_CONSUMO]
		ISNULL((
			SELECT SUM(B.MON_PAGO)
			FROM BRONZE.SAF.PR_ENCABEZADO_PAGO A
			INNER JOIN BRONZE.SAF.PR_DETALLE_PAGO B
				ON A.ID_SECUENCIA = B.ID_SECUENCIA
			WHERE 
				CAST(A.NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT)
				AND A.ID_CUENTA NOT IN ('140103009001', '140103009002', '140103009003',
					'1401030090010000', '1401030090020000',
					'5050260901000300', '2073','2074','2068','2061','2056')
				AND NOT (TIP_TRANSACCION = 4 AND SUBTIP_TRANSAC = 5)
				AND COD_CONCEPTO = 'INTERESES'
				AND FEC_MOVIMIENTO >= @FechaIni
				AND FEC_MOVIMIENTO < @FechaFin
			GROUP BY A.NUM_CREDITO
		), 0) AS PAGOREALIZADOINT,
		--truncate table [SILVER].[RR].[001_ENT_CONSUMO] 
		-- Pago a comisiones
		ISNULL((
			SELECT SUM(B.MON_PAGO)
			FROM BRONZE.SAF.PR_ENCABEZADO_PAGO A
			INNER JOIN BRONZE.SAF.PR_DETALLE_PAGO B
				ON A.ID_SECUENCIA = B.ID_SECUENCIA
			WHERE 
				CAST(A.NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT)
				AND A.ID_CUENTA NOT IN ('140103009001', '140103009002', '140103009003',
					'1401030090010000', '1401030090020000',
					'5050260901000300', '2073','2074','2068','2061','2056')
				AND NOT (TIP_TRANSACCION = 4 AND SUBTIP_TRANSAC = 5)
				AND COD_CONCEPTO IN ('SEGURO_VIDA','SEGURO_DANOS'
					/*'ASE_FINANCIERA', 'COM_PRORROGA',
					'COM_REESTRUCTURA', 'OPI_VALOR', 'VISITA_EXTRA'*/
				)
				AND FEC_MOVIMIENTO >= @FechaIni
				AND FEC_MOVIMIENTO < @FechaFin
			GROUP BY A.NUM_CREDITO
		), 0) AS PAGOREALIZADOCOM,
		---- Pago realizado por falta de pago
		ISNULL((
			SELECT SUM(B.MON_PAGO)
			FROM BRONZE.SAF.PR_ENCABEZADO_PAGO A
			INNER JOIN BRONZE.SAF.PR_DETALLE_PAGO B
				ON A.ID_SECUENCIA = B.ID_SECUENCIA
			WHERE 
				CAST(A.NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT)
				AND A.ID_CUENTA NOT IN ('140103009001', '140103009002', '140103009003',
					'1401030090010000', '1401030090020000',
					'5050260901000300', '2073','2074','2068','2061','2056')
				AND NOT (TIP_TRANSACCION = 4 AND SUBTIP_TRANSAC = 5)
				AND COD_CONCEPTO IN ('COMISION_ADMINISTRACION', 'MORATORIOS')
				AND FEC_MOVIMIENTO >= @FechaIni
				AND FEC_MOVIMIENTO < @FechaFin
			GROUP BY A.NUM_CREDITO
		), 0) AS PAGOREALIZADOFALTAPAG,

		---- Pago de IVA (de MON_PAGO_IVA, sin filtro)
		CAST(ISNULL((
			SELECT SUM(B.MON_PAGO_IVA)
			FROM BRONZE.SAF.PR_ENCABEZADO_PAGO A
			INNER JOIN BRONZE.SAF.PR_DETALLE_PAGO B
				ON A.ID_SECUENCIA = B.ID_SECUENCIA
			WHERE 
				CAST(A.NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT)
				AND A.ID_CUENTA NOT IN ('140103009001', '140103009002', '140103009003',
					'1401030090010000', '1401030090020000',
					'5050260901000300', '2073','2074','2068','2061','2056')
				AND NOT (TIP_TRANSACCION = 4 AND SUBTIP_TRANSAC = 5)
				AND FEC_MOVIMIENTO >= @FechaIni
				AND FEC_MOVIMIENTO < @FechaFin
			GROUP BY A.NUM_CREDITO
		), 0) AS NUMERIC(11,2)) AS PAGOREALIZADOIVAOTR,

		-- Código de crédito reestructurado
		IIF(
			CR.IND_ESTADO_CRED = 'E',
			(SELECT ID_EXTERNO FROM BRONZE.SAF.PR_CREDITOS WHERE NUM_CREDITO = CR.NUM_CONTRATO),
			''
		) AS CODIGOCREDITOREESTRUCTURADO,
		1 AS INDGAR, -- Valor fijo según catálogo. 1 = Con garantía
		-- Días de atraso
		ISNULL((
			SELECT DATEDIFF(DAY, MIN(FEC_COBRO), @FechaFin)-1
			FROM BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO
			WHERE IND_ESTADO = 'P'
			  AND FEC_COBRO < @FechaFin
			  AND NUM_CREDITO = CR.NUM_CREDITO
			  AND IND_POSPOSICION = 'N'
			GROUP BY NUM_CREDITO
		), 0) AS DIASATRASO,

		0 AS PORPAGO, -- Se llena con información proporcionada por Bajaware
		0 AS ATR, -- Se llena con información proporcionada por Bajaware
		0 AS MAXATR, -- Se llena con información proporcionada por Bajaware
		0 AS MESES_DESDE_ULT_ATR_BK, -- Se llena con información proporcionada por Bajaware
		0 AS DELIP, -- valor fijo por cartera
		0 AS TIPO_MET, -- valor fijo de catálogo
		1 AS INDICADOR_SIC, -- valor fijo de catálogo
		0 AS MESES_CONSULTA_SIC, -- Se llena con información proporcionada por Bajaware
		CR.IND_ETAPA AS ETAPADETERIORO,
		CR.TASA_INTERES AS TASA_COBRADA, -- DEFINIDO POR RIESGOS
		DATEDIFF(
			MONTH,
			IIF(
				CR.IND_ESTADO_CRED = 'E',
				(SELECT FEC_APERTURA FROM BRONZE.SAF.PR_CREDITOS WHERE NUM_CREDITO = CR.NUM_CONTRATO),
				CR.FEC_APERTURA
			),
			@FechaFin
		) AS ANTIGUEDADACREDITADO,
		CASE CR.COD_MONEDA
			WHEN 1 THEN 1 -- Moneda nacional
			WHEN 2 THEN 3 -- Dólares
			WHEN 3 THEN 3 -- Euros tratados como dólares
			WHEN 9 THEN 2 -- Udis
			ELSE 1        -- Valor por defecto
		END AS MONEDA,
		CR.MON_CAT AS CAT,
		0 AS MONTOPAGARSIC, -- se llena con información proporcionada por Bajaware
		0 AS MESESIC, -- se llena con información proporcionada por Bajaware
		CASE
			WHEN CR.IND_ESTADO_CRED = 'E' AND CR.IND_ETAPA = 3 THEN ISNULL(RAH.MONTO_RESERVAS,0)
			ELSE 0
		END AS RVAS_ADICIONALES,
		0 AS SALDO_REP_SIC, -- se llena con información proporcionada por Bajaware
		0 AS HIPOTESISPRESREF, -- valor de catálogo no aplica
		CR.FEC_CANCELACION AS FECHABAJA,
		ISNULL(RR.TIP_BAJA,0) AS TIPOBAJA, -- Descomentar cuando se liberen cambios

		CASE 
			WHEN CR.IND_ESTADO_CRED <> 'E' THEN 50
			ELSE
				CASE
					WHEN CR.TASA_INTERES = (
						SELECT TASA_INTERES
						FROM BRONZE.SAF.PR_CREDITOS
						WHERE NUM_CREDITO = CR.NUM_CONTRATO
					) THEN 500
					WHEN CR.TASA_INTERES < (
						SELECT TASA_INTERES
						FROM BRONZE.SAF.PR_CREDITOS
						WHERE NUM_CREDITO = CR.NUM_CONTRATO
					) THEN 515
					ELSE 50
				END
		END AS TIPOCREDITO,
		EOMONTH(@FechaSistema) AS FECHA_INFO,

		CASE 
			WHEN PF.IND_SEXO = 'F' THEN 1
			WHEN PF.IND_SEXO = 'M' THEN 0
			ELSE 0
		END AS FACT_AJUSTE_GENERO,

		ISNULL(IND.CAMBIO_ETAPA, 0)AS CAMBIO_ETAPA,
		50 AS PERIODICIDAD_CAPITAL -- valor fijo: mensual
	FROM BRONZE.SAF.PR_CREDITOS AS CR

	-- Cierre individual para campos de saldo, etapa y estado del crédito
	LEFT JOIN (SELECT * FROM BRONZE.CIERRE.IND_REP WHERE PERIODO = EOMONTH(@FechaSistema) AND CARTERA = 'ION') AS IND ON CR.ID_EXTERNO = IND.CLAVE

	-- Datos de género
	LEFT JOIN BRONZE.SAF.CL_PERSONAS_FISICAS AS PF ON PF.COD_CLIENTE = CR.COD_CLIENTE

	-- Condonaciones
	LEFT JOIN (
		SELECT A.NUM_CREDITO, SUM(B.MON_PAGO) AS MONTO_CONDONADO
		FROM BRONZE.SAF.PR_ENCABEZADO_PAGO A 
		INNER JOIN BRONZE.SAF.PR_DETALLE_PAGO B 
			ON A.ID_SECUENCIA = B.ID_SECUENCIA
		WHERE 
			((TIP_TRANSACCION = 4 AND SUBTIP_TRANSAC = 5)
			OR (TIP_TRANSACCION = 4 AND A.ID_CUENTA IN ('5050260901000300')))
			AND FEC_MOVIMIENTO >= DATEFROMPARTS(YEAR(@FechaSistema), MONTH(@FechaSistema), 1)
			AND FEC_MOVIMIENTO < DATEADD(MONTH, 1, DATEFROMPARTS(YEAR(@FechaSistema), MONTH(@FechaSistema), 1))
		GROUP BY A.NUM_CREDITO
	) AS COND 
		ON CAST(COND.NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT)

	-- Bonificaciones
	LEFT JOIN (
		SELECT A.NUM_CREDITO, SUM(B.MON_PAGO) AS MONTO_BONIFICADO
		FROM BRONZE.SAF.PR_ENCABEZADO_PAGO A 
		INNER JOIN BRONZE.SAF.PR_DETALLE_PAGO B 
			ON A.ID_SECUENCIA = B.ID_SECUENCIA
		WHERE 
			A.ID_CUENTA IN ('140103009001','140103009002')
			AND FEC_MOVIMIENTO >= DATEFROMPARTS(YEAR(@FechaSistema), MONTH(@FechaSistema), 1)
			AND FEC_MOVIMIENTO < DATEADD(MONTH, 1, DATEFROMPARTS(YEAR(@FechaSistema), MONTH(@FechaSistema), 1))
		GROUP BY A.NUM_CREDITO
	) AS BON 
		ON CAST(BON.NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT)
	
	LEFT JOIN BRONZE.SAF.PR_INF_RR_X_CREDITO RR ON CR.NUM_CREDITO = RR.NUM_CREDITO
	LEFT JOIN BRONZE.RR.RESERVAS_ADICIONALES_HISTORICO RAH ON CR.ID_EXTERNO = RAH.ID_EXTERNO
	WHERE CR.NUM_CREDITO IN
	(
	SELECT MAX(NUM_CREDITO) FROM BRONZE.SAF.PR_CREDITOS  -- SE OBTIENE UN SOLO CRÉDITO POR CLIENTE PARA EVITAR DUPLICIDAD
	WHERE 
	(IND_ESTADO = 'D' OR 
			(IND_ESTADO = 'C' 
			AND (FEC_CANCELACION >= @FechaIni
			AND FEC_CANCELACION < @FechaFin))
		) 
	AND CR.NUM_DESC_TIPO_CREDITO IN (1) -- CONSUMO
	AND IND_LINEA = 'N'
	GROUP BY COD_CLIENTE)



		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

# [dbo].[003_ENT_REL_CONSUMO_GARANTIA]

```sql
/****** Object:  StoredProcedure [dbo].[003_ENT_REL_CONSUMO_GARANTIA]    Script Date: 01/06/2026 20:11:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
--	Modificado: Alberto Vazquez
CREATE PROCEDURE [dbo].[003_ENT_REL_CONSUMO_GARANTIA] 
	@CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[003_ENT_REL_CONSUMO_GARANTIA]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[003_ENT_REL_CONSUMO_GARANTIA]
				WHERE [FECHA_REPORTE] >= @FechaIni
					AND [FECHA_REPORTE] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[003_ENT_REL_CONSUMO_GARANTIA]
			WHERE [FECHA_REPORTE] >= @FechaIni
				AND [FECHA_REPORTE] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END

		INSERT INTO [SILVER].[RR].[003_ENT_REL_CONSUMO_GARANTIA] (
			CODIGOGARANTIA,
			CODIGOCREDITO,
			PORCENTAJE,
			FECHA_REPORTE
			)
		SELECT 
			B.NUM_GARANTIA AS CODIGOGARANTIA,
			A.NUM_CREDITO AS CODIGOCREDITO,
			CASE 
				WHEN B.MON_GARANTIA IS NULL THEN NULL
				WHEN B.MON_GARANTIA = 0 THEN NULL -- Handle division by zero
				WHEN (A.MON_SALDO / B.MON_GARANTIA) * 100 > 9999.999999 THEN 9999.999999 -- Cap maximum value
				ELSE CAST((A.MON_SALDO / B.MON_GARANTIA) * 100 AS NUMERIC(10, 6))
			END AS PORCENTAJE,
			--DATEADD(dd, - (DAY(A.[FEC_APERTURA]) - 1), A.[FEC_APERTURA]) AS FECHA_REPORTE -- First day of month
			EOMONTH(@FechaSistema)
			FROM [BRONZE].SAF.PR_CREDITOS A WITH (NOLOCK)
			INNER JOIN [BRONZE].SAF.PR_GARANTIAS B WITH (NOLOCK) ON A.COD_EMPRESA = B.COD_EMPRESA
			AND A.COD_CLIENTE = B.COD_CLIENTE
			WHERE 
			(A.IND_ESTADO = 'D' OR 
				(A.IND_ESTADO = 'C' 
				AND (A.FEC_CANCELACION >= @FechaIni
				AND A.FEC_CANCELACION < @FechaFin))
			)
			AND A.IND_LINEA = 'N'
			AND A.NUM_DESC_TIPO_CREDITO = 1


		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

# [dbo].[004_PAGOS_EXIGIBLES]

```sql
/****** Object:  StoredProcedure [dbo].[004_PAGOS_EXIGIBLES]    Script Date: 01/06/2026 20:11:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[004_PAGOS_EXIGIBLES] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[004_PAGOS_EXIGIBLES]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[004_PAGOS_EXIGIBLES]
				WHERE [Fecha] >= @FechaIni
					AND [Fecha] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[004_PAGOS_EXIGIBLES]
			WHERE [Fecha] >= @FechaIni
				AND [Fecha] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		WITH pagos
		AS (
			SELECT 
				C.ID_EXTERNO AS CodigoCredito,
				'PAGO' AS TipoMovimiento,
				B.MON_PAGO + B.MON_PAGO_IVA AS Monto,
				A.Fec_movimiento AS FECHA,
				Q.num_secuencia AS Num_pago
			FROM [BRONZE].[SAF].PR_ENCABEZADO_PAGO A WITH (NOLOCK)
			INNER JOIN [BRONZE].[SAF].PR_DETALLE_PAGO B WITH (NOLOCK) ON A.ID_SECUENCIA = B.ID_SECUENCIA
			INNER JOIN [BRONZE].[SAF].PR_CREDITOS C WITH (NOLOCK) ON A.COD_EMPRESA = C.COD_EMPRESA
				AND A.NUM_CREDITO = C.NUM_CREDITO
			INNER JOIN [BRONZE].[SAF].PR_RUBRO_COBRO_X_CREDITO Q WITH (NOLOCK) ON Q.Num_Secuencia = B.Id_referencia
				AND Q.Num_Credito = A.Num_credito
				AND Q.Num_Det_Secuencia = B.Num_det_secuencia
				AND Q.Cod_Rubro = B.Cod_concepto
				AND Q.Ind_Estado != 'N'
			WHERE C.IND_ESTADO IN ('D', 'C')
				AND C.IND_LINEA = 'N'
				AND A.mon_movimiento > 0
				AND C.[ID_EXTERNO] IS NOT NULL
				AND A.FEC_MOVIMIENTO >= @FechaIni
				AND A.FEC_MOVIMIENTO < @FechaFin
				AND C.NUM_DESC_TIPO_CREDITO IN (/*Consumo*/ 1, /*Hip*/ 12,13,14,15,28,3,4,16,17)
			),
		exigibles
		AS (
			SELECT 
				C.ID_EXTERNO AS CODIGOCREDITO,
				'MONEXI' AS TipoMovimiento,
				Q.Mon_Rubro + Q.Mon_Iva AS Monto,
				Q.FEC_COBRO AS FECHA,
				Q.num_secuencia AS Num_pago
			FROM [BRONZE].[SAF].PR_CREDITOS C WITH (NOLOCK)
			INNER JOIN [BRONZE].[SAF].PR_RUBRO_COBRO_X_CREDITO Q WITH (NOLOCK) ON Q.Num_Credito = C.Num_credito
				AND Q.Ind_Estado != 'N'
			WHERE C.IND_ESTADO IN ('D', 'C')
				AND C.IND_LINEA = 'N'
				AND C.[ID_EXTERNO] IS NOT NULL
				AND Q.FEC_COBRO >= @FechaIni
				AND Q.FEC_COBRO < @FechaFin
				AND C.NUM_DESC_TIPO_CREDITO IN (/*Consumo*/ 1, /*Hip*/ 12,13,14,15,28,3,4,16,17)

			)
		INSERT INTO [SILVER].[RR].[004_PAGOS_EXIGIBLES] (
			[CodigoCredito],
			[TipoMovimiento],
			[Monto],
			[Fecha],
			[NumPago]
			)
		SELECT CodigoCredito,
			TipoMovimiento,
			SUM(Monto) AS Monto,
			Fecha,
			Num_pago
		FROM pagos
		GROUP BY [CodigoCredito],
			[TipoMovimiento],
			[Fecha],
			Num_pago
		
		UNION ALL
		
		SELECT [CodigoCredito],
			[TipoMovimiento],
			SUM(Monto) AS Monto,
			[FecHA],
			Num_pago
		FROM exigibles
		GROUP BY [CodigoCredito],
			[TipoMovimiento],
			[Fecha],
			Num_pago;

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;


GO
```

# [dbo].[007_ENT_CLIENTES_PRODUCTO]

```sql
/****** Object:  StoredProcedure [dbo].[007_ENT_CLIENTES_PRODUCTO]    Script Date: 01/06/2026 20:11:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO


CREATE PROCEDURE [dbo].[007_ENT_CLIENTES_PRODUCTO] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @FechaIni DATE,
		@FechaFin DATE,
		@Fecha DATE,
		@TipoRep INT,
		@MesIni INT
	DECLARE @NombreJob NVARCHAR(128) = '[007_ENT_CLIENTES_PRODUCTO]';

	BEGIN TRY
		SET @TipoRep = 3 --meses del periodo
		SET @MesIni = FLOOR((month(@FechaSistema) + @TipoRep - 1) / @TipoRep) * @TipoRep - @TipoRep + 1
		SET @FechaIni = DATEfromparts(year(@FechaSistema), @MesIni, 1)
		SET @FechaFin = DATEadd(month, @TipoRep, @FechaIni)
		--------------------------------------------------------------------------------
		-- QUERY
		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[007_ENT_CLIENTES_PRODUCTO]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
					AND IND_EXTRACCION = 'S'
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[007_ENT_CLIENTES_PRODUCTO]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin
				AND IND_EXTRACCION = 'S';

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END
		INSERT INTO [SILVER].[RR].[007_ENT_CLIENTES_PRODUCTO] (
			[PRODUCTO],
			[MUNICIPIO],
			[TIPO_CLIENTE],
			[PERSONALIDAD_JURIDICA],
			[CODIGO_CLIENTE],
			[MONEDA],
			[TIPO_INSTRUMENTO],
			[MONTO_OPERACIONES],
			[NUMERO_OPERACIONES],
			[FECHA_INFO],
			[IND_EXTRACCION]
			)

		SELECT 
			'03' AS PRODUCTO,
			ISNULL((SELECT TOP 1 CONCAT(
								IIF(E.COD_PAIS IS NULL, '00', E.COD_PAIS),
								IIF(E.COD_PROVINCIA IS NULL, '000', E.COD_PROVINCIA)) 
			FROM BRONZE.SAF.CL_DIR_CLIENTES E WHERE E.IND_ESTADO= 'A' AND E.COD_DIRECCION = 1 AND E.COD_CLIENTE = C.COD_CLIENTE),'00000')
			AS MUNICIPIO,
			CASE 
				WHEN D.IND_PERSONA = 'F' THEN 1
				WHEN D.IND_PERSONA = 'J' THEN 3
			END AS TIPO_CLIENTE,
			CASE 
				WHEN D.IND_PERSONA = 'F' THEN 1
				WHEN D.IND_PERSONA = 'J' THEN 2
			END AS PERSONALIDAD_JURIDICA,
			C.COD_CLIENTE,
			'2' AS MONEDA,
			'99' TIPO_INSTRUMENTO,
			SUM(B.MON_PAGO + B.MON_PAGO_IVA) AS MONTO_OPERACIONES,
			COUNT(DISTINCT A.ID_SECUENCIA) AS NUMERO_OPERACIONES,
			A.[Fec_movimiento] AS [FECHA_INFO],
			'S' AS IND_EXTRACCION
		FROM [BRONZE].[SAF].PR_ENCABEZADO_PAGO A WITH (NOLOCK)
		INNER JOIN [BRONZE].[SAF].PR_DETALLE_PAGO B WITH (NOLOCK) ON A.ID_SECUENCIA = B.ID_SECUENCIA
		INNER JOIN [BRONZE].[SAF].PR_CREDITOS C WITH (NOLOCK) ON A.COD_EMPRESA = C.COD_EMPRESA
			AND A.NUM_CREDITO = C.NUM_CREDITO
		LEFT JOIN [BRONZE].[SAF].CL_CLIENTES D WITH (NOLOCK) ON C.COD_EMPRESA = D.COD_EMPRESA
			AND C.COD_CLIENTE = D.COD_CLIENTE
		WHERE A.[Fec_movimiento] >= @FechaIni
			AND A.[Fec_movimiento] < @FechaFin
			AND A.ID_CUENTA NOT IN ('140103009001', '140103009002', '140103009003','1401030090010000','1401030090020000','5050260901000300','2073','2074','2068','2061','2056')
			AND A.TIP_TRANSACCION !=3
			AND NOT (TIP_TRANSACCION = 4 AND SUBTIP_TRANSAC = 5)
			AND (C.IND_ESTADO = 'D' OR     -- CREDITOS ACTIVOS O CANCELADOS DENTRO DEL PERIODO A REPORTAR  
				(C.IND_ESTADO = 'C' 
				AND (FEC_CANCELACION >= @FechaIni
				AND FEC_CANCELACION < @FechaFin))
				)
			AND C.NUM_DESC_TIPO_CREDITO IN (/*COM*/12,13,14,15,28,2,5,18,27,/*IND*/1,3,4,9,10,11,16,17)
			AND B.MON_PAGO <> 0
			AND A.mon_movimiento > 0
		GROUP BY D.IND_PERSONA,
			C.COD_CLIENTE,
			A.[Fec_movimiento]
		ORDER BY C.COD_CLIENTE;
		
		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

# [dbo].[008_ENT_GARANTIA]

```sql
/****** Object:  StoredProcedure [dbo].[008_ENT_GARANTIA]    Script Date: 01/06/2026 20:11:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[008_ENT_GARANTIA] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @FechaIni DATE,
		@FechaFin DATE;
	DECLARE @NombreJob NVARCHAR(128) = '[008_ENT_GARANTIA]';

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[008_ENT_GARANTIA]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[008_ENT_GARANTIA]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END

		INSERT INTO [SILVER].[RR].[008_ENT_GARANTIA] (
			[ID_GARANTIA],
			[TIPO_GARANTIA],
			[FECHA_ULT_ACTUALIZACION],
			[MONEDA_GAR],
			[PRELACION_GAR],
			[MONTO_GAR],
			[FECHA_ULT_AVALUO],
			[LOCALIDAD_GARANTIA_INM],
			[DESCRIPCION_GARANTIA],
			[GRADO_REISGP],
			[FECHA_INFO]
			)
		SELECT 
			A.NUM_GARANTIA AS ID_GARANTIA,
			CASE 
				WHEN A.TIP_GARANTIA = 'HIP' THEN 53
				WHEN A.TIP_GARANTIA = 'FID' THEN 55
			END AS TIPO_GARANTIA, --CATALOGO
			CASE 
				WHEN A.FEC_AVALUO IS NOT NULL THEN A.FEC_AVALUO 
				ELSE 0
			END
			AS FECHA_ULT_ACTUALIZACION, 
			CASE
				WHEN A.COD_MONEDA = 1 THEN 0
				WHEN A.COD_MONEDA = 9 THEN 200
			END AS MONEDA_GAR, -- CATALOGO
			CASE
				WHEN C.PRELACION = 1 THEN 1
				WHEN C.PRELACION = 2 THEN 2
				ELSE 0
			END AS PRELACION_GAR, -- CATALOGO
			ISNULL(A.MON_GARANTIA,0) AS MONTO_GAR,
			CASE 
				WHEN A.FEC_AVALUO IS NOT NULL THEN A.FEC_AVALUO 
				ELSE 0
			END AS FECHA_ULT_AVALUO,
			CONCAT('484',
				IIF(D.COD_PAIS IS NULL, '00', D.COD_PAIS),
				IIF(D.COD_PROVINCIA IS NULL, '000', D.COD_PROVINCIA),
				'0001') AS LOCALIDAD_GARANTIA_INM,
			NULL AS DESCRIPCION_GARANTIA,
			NULL AS GRADO_REISGP,
			EOMONTH(@FECHASISTEMA)AS FECHA_INFO
		FROM BRONZE.SAF.PR_GARANTIAS A
		LEFT JOIN BRONZE.SAF.PR_CREDITOS B ON A.COD_CLIENTE = B.COD_CLIENTE
		LEFT JOIN BRONZE.INF.CREDITO_EMPRESARIAL C ON B.ID_EXTERNO = C.IDCREDITO
		LEFT JOIN BRONZE.SAF.PR_GARANTIAS_HIPOTECARIAS D ON A.NUM_GARANTIA = D.NUM_GARANTIA
		WHERE A.FEC_AVALUO IS NOT NULL
		AND B.IND_ESTADO = 'D' 
		AND B.IND_LINEA = 'N'
		AND B.NUM_DESC_TIPO_CREDITO IN (/*COM*/12,13,14,15,28,2,5,18,27)
		AND A.IND_ESTADO = 'A'
		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log
	
	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;


GO
```

# [dbo].[009_ENT_R04A_SALDOS]

```sql
/****** Object:  StoredProcedure [dbo].[009_ENT_R04A_SALDOS]    Script Date: 01/06/2026 20:11:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[009_ENT_R04A_SALDOS] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @FechaIni DATE,
		@FechaFin DATE;
	DECLARE @NombreJob NVARCHAR(128) = '[009_ENT_R04A_SALDOS]';

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni)

		--------------------------------------------------------------------------------
		-- QUERY
		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[009_ENT_R04A_SALDOS]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[009_ENT_R04A_SALDOS]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END --COMPLETAR

		INSERT INTO [SILVER].[RR].[009_ENT_R04A_SALDOS] (
			[CUENTA_CREDITO],
			[SITUACION_CREDITO],
			[SALDO_PRINCIPAL],
			[INTERES_DEVENGADO],
			[INTERESES_COBRADOS],
			[COMISIONES_COBRADAS],
			[SALDO_PROMEDIO],
			[MONEDA],
			[FECHA_INFO]
			)
		SELECT 
			CASE
				WHEN A.NUM_DESC_TIPO_CREDITO IN (12,13,14,15,28,2,5,18,27) THEN ISNULL(E1.CLAVE,'')
				WHEN A.NUM_DESC_TIPO_CREDITO IN (1,3,4,9,10,11,16,17) THEN ISNULL(E2.CLAVE,'')
				ELSE ''
			END AS CUENTA_CREDITO,
			CASE
				WHEN A.NUM_DESC_TIPO_CREDITO IN (12,13,14,15,28,2,5,18,27) THEN ISNULL(B.ETAPA_RIESGO,0)
				WHEN A.NUM_DESC_TIPO_CREDITO IN (1,3,4,9,10,11,16,17) THEN ISNULL(C.ETAPA_RIESGO,0)
				ELSE 0
			END AS SITUACION_CREDITO,
			ISNULL(D.MON_SALDO,0) AS SALDO_PRINCIPAL,
			ISNULL(D.MON_INT_ACUMULADO,0) AS INTERES_DEVENGADO,
			ISNULL(D.MON_PAGADO_INTERESES,0) AS INTERESES_COBRADOS,
			CASE
				WHEN A.NUM_DESC_TIPO_CREDITO IN (12,13,14,15,28,2,5,18,27) THEN /*ISNULL(B.COMISIONES_PAGADAS,0)*/ ISNULL(B.PAGADO_COMISIONES,0) + ISNULL(B.PAGADO_COMISIONES_IVA,0)
				WHEN A.NUM_DESC_TIPO_CREDITO IN (1,3,4,9,10,11,16,17) THEN ISNULL(C.COMISIONES_PAGADAS,0)
				ELSE 0
			END AS COMISIONES_COBRADAS,
			CASE 
				WHEN A.TIP_CALENDARIO = 2 THEN ISNULL(D.MON_SALDO/CAST(DAY(EOMONTH(@FechaSistema)) AS INT),0) 
				WHEN A.TIP_CALENDARIO = 3 THEN ISNULL(D.MON_SALDO/30,0)
			END AS SALDO_PROMEDIO,
			CASE
				WHEN A.COD_MONEDA = 1 THEN 0
				WHEN A.COD_MONEDA = 2 THEN 1
				WHEN A.COD_MONEDA = 3 THEN 124
				WHEN A.COD_MONEDA = 9 THEN 200
				ELSE 0
			END AS MONEDA,
			EOMONTH(@FechaSistema) AS FECHA_INFO
		FROM BRONZE.SAF.PR_CREDITOS A
		LEFT JOIN (SELECT * FROM BRONZE.CIERRE.CMR_REP WHERE FECHA_SISTEMA = EOMONTH(@FechaSistema)) B ON A.ID_EXTERNO = B.ID_EXTERNO
		LEFT JOIN (SELECT * FROM BRONZE.CIERRE.IND_REP WHERE PERIODO = EOMONTH(@FechaSistema) AND CARTERA = 'ION') C ON A.ID_EXTERNO = C.CLAVE
		LEFT JOIN (SELECT * FROM BRONZE.SAF.PR_CREDITOS_HISTORICO WHERE FEC_CORTE = EOMONTH(@FechaSistema)) D ON A.ID_EXTERNO = D.ID_EXTERNO
		LEFT JOIN BRONZE.RR.Catalogo_Cuenta_R04A_V1 E1 ON CONCAT(ISNULL(B.CREDITO_TIPO,''),'/',ISNULL(B.CARTERA_TIPO,''),'/',ISNULL(B.ETAPA_RIESGO,''),'/')=E1.DESCRIPCION
		LEFT JOIN BRONZE.RR.Catalogo_Cuenta_R04A_V1 E2 ON CONCAT(ISNULL(C.CLASIFICACION_DE_VIVIENDA,''),'/',ISNULL(C.SIN_RESTRICCION_RESTRINGIDOS,''),'/',ISNULL(C.ETAPA_RIESGO,''),'/')=E2.DESCRIPCION
		WHERE A.NUM_CREDITO IN
		(
		SELECT MAX(NUM_CREDITO) FROM BRONZE.SAF.PR_CREDITOS  -- SE OBTIENE UN SOLO CRÉDITO POR CLIENTE PARA EVITAR DUPLICIDAD
		WHERE 
		(IND_ESTADO = 'D' OR 
				(IND_ESTADO = 'C' 
				AND (FEC_CANCELACION >= @FechaIni
				AND FEC_CANCELACION < @FechaFin))
			) 
		AND A.NUM_DESC_TIPO_CREDITO IN (/*COM*/12,13,14,15,28,2,5,18,27,/*IND*/1,3,4,9,10,11,16,17)
		AND IND_LINEA = 'N'
		GROUP BY COD_CLIENTE
		)


		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = ' Proceso completado.Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = ' Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH

	-- Preparar el mensaje detallado para la alerta
	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos ';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = ' ALERTA: Error en ' + ISNULL(@NombreJob, ' Job Desconocido ');
		SET @Cuerpo = ' Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + ' Detalles del Job :' + CHAR(13) + CHAR(10) + ' - Nombre del Job: ' + ISNULL(@NombreJob, ' No especificado ') + CHAR(13) + CHAR(10) + ' - Programado por: ' + ISNULL(@ProgramadorJob, ' No especificado ') + CHAR(13) + CHAR(10) + ' - Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + ' - Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + ' - Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + ' Detalles de la Ejecución :' + CHAR(13) + CHAR(10) + ' Mensaje de Error :' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + ' Log de Ejecución :' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = ' TEXT ',
				@importance = ' High ';

			SET @LogMessage = ' Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = ' Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END

	-- Registrar en la tabla de log
	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = ' Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

# [dbo].[010_ENT_PERSONA]

```sql
/****** Object:  StoredProcedure [dbo].[010_ENT_PERSONA]    Script Date: 01/06/2026 20:11:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[010_ENT_PERSONA] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @FechaIni DATE,
		@FechaFin DATE;
	DECLARE @NombreJob NVARCHAR(128) = '[010_ENT_PERSONA]';

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[010_ENT_PERSONA]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[010_ENT_PERSONA]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END

		INSERT INTO [SILVER].[RR].[010_ENT_PERSONA] (
			[ID_PERSONA],
			[NOMBRE_PERSONA],
			[APELLIDO_PER_FISICA],
			[NOMBRE_CNBV],
			[RFC],
			[CURP],
			[LOCALIDAD],
			[NACIONALIDAD],
			[ACTIVIDAD_ECONOMICA],
			[GRUPO_RIESGO_COMUN],
			[PERSONA_RELACIONADA],
			[TIPO_CARTERA],
			[CREDITO_REP_SIC],
			[FOLIO_CONSULTA_BURO],
			[RC_GRUPORIESGO],
			--[RC_TABLAADEUDO],
			--[RC_GRADORIESGO],
			--[RC_ESCALACALIF],
			[GENERO],
			[ESTADO_CIVIL],
			[DOMICILIO],
			[EJECUTIVO],
			[TELEFONO],
			[CORREO],
			[SUCURSAL],
			[CODIGO_CLIENTE_IPAB],
			[NUMERO_EMPLEADOS],
			[EDAD_ACREDITADO],
			[APORTA_SUBCUENTA],
			[INGRESOS_MENSUALES],
			[COMPROBANTE_INGRESOS],
			[SECTOR_LABORAL],
			[CODIGOCLIENTEBC],
			[PERSONALIDADJURIDICA],
			[FECHANACIMIENTO],
			[SECTORECONOMICO],
			[CODIGOPOSTAL],
			[PI_PERSONA_CNBV],
			[TIPO_CTE_GAR],
			[FECHA_INFO]
			)
		SELECT 
			A.COD_CLIENTE AS ID_PERSONA,
			IIF(C.PRIMER_NOMBRE IS NOT NULL, C.PRIMER_NOMBRE, ISNULL(D.RAZON_SOCIAL,'')) AS NOMBRE_PERSONA,
			ISNULL(C.PRIMER_APELLIDO,'') AS APELLIDO_PER_FISICA,
			ISNULL(A.NOMBRECNBV,'') AS NOMBRE_CNBV,
			ISNULL((SELECT ISNULL(LEFT(NUM_ID,13),'') FROM BRONZE.SAF.CL_ID_CLIENTES WHERE COD_TIPO_ID = 'RFC' AND COD_CLIENTE = A.COD_CLIENTE),'') AS RFC,
			ISNULL((SELECT ISNULL(LEFT(NUM_ID,18),'') FROM BRONZE.SAF.CL_ID_CLIENTES WHERE COD_TIPO_ID = 'CURP' AND COD_CLIENTE = A.COD_CLIENTE),'') AS CURP,
			CONCAT('484',ISNULL(E.COD_PAIS,'00'),ISNULL(E.COD_PROVINCIA,'000'),'0001') AS LOCALIDAD,
			ISNULL(CN.CLAVE,0) AS NACIONALIDAD, -- catalogo
			IIF(C.COD_SUBACTIV IS NOT NULL, C.COD_SUBACTIV, ISNULL(D.COD_SUBACTIV,'')) AS ACTIVIDAD_ECONOMICA,
			ISNULL(F.GrupoDeRiesgo,'') AS GRUPO_RIESGO_COMUN,
			CASE
				WHEN F.AcreditadoRelacionado = 'Persona No Relacionada' THEN 8
				WHEN F.AcreditadoRelacionado = 'Persona Relacionada Relevante' THEN 9
				WHEN F.AcreditadoRelacionado = 'Persona Relacionada No Relevante' THEN 10
				ELSE 8
			END AS PERSONA_RELACIONADA,
			CASE
				WHEN B.NUM_DESC_TIPO_CREDITO IN (12,13,14,15,28,2,5,18,27) THEN '310' -- comercial
				WHEN B.NUM_DESC_TIPO_CREDITO IN (1) THEN '320' --consumo
				WHEN B.NUM_DESC_TIPO_CREDITO IN (3,4,9,10,11,16,17) THEN '320' --hipotecario
				ELSE 0
			END AS TIPO_CARTERA,
			IIF(F.NumeroConsultaSociedadInformacionCrediticia IS NOT NULL, 700, 750) AS CREDITO_REP_SIC,
			ISNULL(F.NumeroConsultaSociedadInformacionCrediticia,'') AS FOLIO_CONSULTA_BURO,
			CASE
				WHEN B.NUM_DESC_TIPO_CREDITO IN (12,13,14,15,28,2,5,18,27) AND B.ID_EXTERNO NOT LIKE '%PY%'THEN 90
				WHEN B.NUM_DESC_TIPO_CREDITO IN (12,13,14,15,28,2,5,18,27) AND B.ID_EXTERNO LIKE '%PY%'THEN 71
				ELSE 0
			END AS RC_GRUPORIESGO,
			CASE
				WHEN C.IND_SEXO= 'M' THEN 2
				WHEN C.IND_SEXO= 'F' THEN 1
				ELSE 0
			END AS GENERO,
			CASE
				WHEN C.EST_CIVIL= 'S' THEN 1
				WHEN C.EST_CIVIL= 'C' THEN 2
				ELSE 0
			END AS ESTADO_CIVIL,
			LEFT(CONCAT(DR.CALLE,' ',DR.NUMERO_EXTERIOR,' ',DR.NUMERO_INTERIOR,' ',DR.COLONIA_POBLACION,' ',DR.DELEGACION_MUNICIPIO,' ',
			DR.CIUDAD,' ',DR.ESTADO,' ',DR.PAIS,' ',DR.CP),50) AS DOMICILIO, -- archivo externo
			'' AS EJECUTIVO,
			ISNULL(DR.TELEFONO,'')  AS TELEFONO, -- archivo externo
			ISNULL(DR.EMAIL,'') AS CORREO, -- archivo externo
			'' AS SUCURSAL,
			'' AS CODIGO_CLIENTE_IPAB,
			ISNULL(F.NumeroEmpleados,0) AS NUMERO_EMPLEADOS,
			ISNULL(DATEDIFF(YEAR,C.FEC_NACIMIENTO,EOMONTH(@FechaSistema)),0) AS EDAD_ACREDITADO,
			CASE
				WHEN DC.MONTO_SUBCUENTA <> 0 THEN 1
				ELSE 2
			END AS APORTA_SUBCUENTA, -- archivo externo
			ISNULL(DR.INGRESOS_MENSUALES_BRUTOS,0)  AS INGRESOS_MENSUALES, -- archivo externo
			CASE
				WHEN DR.COMPROBACION_INGRESOS = 'NO ASALARIADO' THEN 1
				WHEN DR.COMPROBACION_INGRESOS = 'ASALARIADO' THEN 2
				ELSE 0
			END AS COMPROBANTE_INGRESOS, -- archivo externo
			CASE
				WHEN DR.SECTOR_LABORAL = 'SECTOR PUBLICO ESTATAL' THEN 1
				WHEN DR.SECTOR_LABORAL = 'SECTOR PUBLICO MUNICIPAL' THEN 2
				WHEN DR.SECTOR_LABORAL = 'SECTOR PUBLICO FEDERAL' THEN 3
				WHEN DR.SECTOR_LABORAL = 'SECTOR PRIVADO' THEN 4
				WHEN DR.SECTOR_LABORAL = 'NO ASALARIADO' THEN 5
				ELSE 0
			END AS SECTOR_LABORAL, -- archivo externo
			ISNULL((SELECT ISNULL(NUM_ID,'') FROM BRONZE.SAF.CL_ID_CLIENTES WHERE COD_TIPO_ID = 'RFC' AND COD_CLIENTE = A.COD_CLIENTE),'') AS CODIGOCLIENTEBC,
			/*ISNULL(A.PERSONALIDAD_JURIDICA,0)*/ 0 AS PERSONALIDADJURIDICA,
			ISNULL(C.FEC_NACIMIENTO,'') AS FECHANACIMIENTO,
			IIF(C.COD_SECTOR IS NOT NULL, C.COD_SECTOR, ISNULL(D.COD_SECTOR,0)) AS SECTORECONOMICO, --CRUZAR
			ISNULL(E.COD_POSTAL,'') AS CODIGOPOSTAL,
			(SELECT IIF(SUM(BANDERA)=0,0,10) FROM
			(SELECT COD_CLIENTE, IIF(IND_ETAPA=3 OR FEC_VENCIMIENTO < EOMONTH(@FechaSistema),1,0) AS BANDERA
				FROM BRONZE.SAF.PR_CREDITOS)B WHERE B.COD_CLIENTE = A.COD_CLIENTE
			GROUP BY COD_CLIENTE ) AS PI_PERSONA_CNBV,
			CASE
				WHEN I.TIP_GARANTE = 'A' THEN
					CASE
						WHEN B.NUM_DESC_TIPO_CREDITO IN (12,13,14,15,28,2,5,18,27) THEN '1'
						WHEN B.NUM_DESC_TIPO_CREDITO IN (1) THEN '2'
						WHEN B.NUM_DESC_TIPO_CREDITO IN (3,4,9,10,11,16,17) THEN '3'
						ELSE '0'
					END
				WHEN I.TIP_GARANTE = 'AV' THEN '4'
				WHEN I.TIP_GARANTE = 'O' THEN '6'
				WHEN I.TIP_GARANTE = 'G' THEN '7'
				ELSE '0'
			END AS TIPO_CTE_GAR,
			EOMONTH(@FechaSistema) AS FECHA_INFO
		FROM BRONZE.SAF.CL_CLIENTES A
		JOIN (SELECT * FROM ( SELECT *, ROW_NUMBER() OVER(PARTITION BY COD_CLIENTE ORDER BY IND_ESTADO DESC, NUM_CREDITO DESC ) AS RN
		FROM BRONZE.SAF.PR_CREDITOS) AS T WHERE RN = 1) B ON A.COD_CLIENTE = B.COD_CLIENTE
		LEFT JOIN BRONZE.SAF.CL_PERSONAS_FISICAS C ON A.COD_CLIENTE = C.COD_CLIENTE
		LEFT JOIN BRONZE.SAF.CL_PERSONAS_JURIDICAS D ON A.COD_CLIENTE = D.COD_CLIENTE
		LEFT JOIN (SELECT * FROM BRONZE.SAF.CL_DIR_CLIENTES WHERE TIP_DIRECCION = 'C' AND IND_ESTADO= 'A' AND COD_DIRECCION = 1) E ON A.COD_CLIENTE = E.COD_CLIENTE
		LEFT JOIN BRONZE.INF.CREDITO_EMPRESARIAL F ON B.ID_EXTERNO = F.IDCREDITO 
		--LEFT JOIN (SELECT * FROM BRONZE.CIERRE.CMR_REP WHERE FECHA_SISTEMA = EOMONTH(@FechaSistema)) G ON B.ID_EXTERNO = G.ID_EXTERNO
		--LEFT JOIN (SELECT * FROM BRONZE.CIERRE.IND_REP WHERE FECHA_SISTEMA = EOMONTH(@FechaSistema)) H ON B.ID_EXTERNO = H.ID_EXTERNO
		LEFT JOIN (SELECT NUM_CREDITO, COD_CLIENTE, MAX(TIP_GARANTE)TIP_GARANTE FROM BRONZE.[SAF].[PR_CLIENTES_X_CREDITO]
		GROUP BY NUM_CREDITO, COD_CLIENTE) I ON A.COD_CLIENTE = I.COD_CLIENTE AND B.NUM_CREDITO = I.NUM_CREDITO
		--LEFT JOIN BRONZE.[RR].[ActividadEconomica_V2] J1 ON C.COD_SUBACTIV = J1.CODIGO
		--LEFT JOIN BRONZE.[RR].[ActividadEconomica_V2] J2 ON D.COD_SUBACTIV = J2.CODIGO
		LEFT JOIN BRONZE.SAF.CL_NACIONALIDADES NAC ON C.NACIONALIDAD = NAC.COD_NACIONALIDAD AND C.COD_EMPRESA = NAC.COD_EMPRESA
		LEFT JOIN BRONZE.RR.NACIONALIDADES CN ON NAC.DES_NACIONALIDAD = CN.PAIS_NORMALIZADO
		--LEFT JOIN BRONZE.LMDA.DATOS_CLIENTE DR ON DR.ID_CREDITO = B.ID_EXTERNO
		LEFT JOIN (SELECT * FROM ( SELECT *, ROW_NUMBER() OVER(PARTITION BY ID_CREDITO ORDER BY ID_CREDITO DESC) AS RN
		FROM BRONZE.LMDA.DATOS_CLIENTE) AS T WHERE RN = 1) DR ON DR.ID_CREDITO = B.ID_EXTERNO
		LEFT JOIN BRONZE.LMDA.DATOS_CREDITO DC ON DC.ID_CREDITO = B.ID_EXTERNO
		WHERE 
		A.IND_ESTADO = 'A' -- CLIENTES ACTIVOS
		/*B.NUM_CREDITO IN
		(
		SELECT MAX(NUM_CREDITO) FROM BRONZE.SAF.PR_CREDITOS  -- SE OBTIENE UN SOLO CRÉDITO POR CLIENTE PARA EVITAR DUPLICIDAD
		WHERE 
		(IND_ESTADO = 'D' OR 
				(IND_ESTADO = 'C' 
				AND (FEC_CANCELACION >= @FechaIni
				AND FEC_CANCELACION < @FechaFin))
			) 
		AND B.NUM_DESC_TIPO_CREDITO IN (/*COM*/12,13,14,15,28,2,5,18,27,/*IND*/1,3,4,9,10,11,16,17)
		AND IND_LINEA = 'N'
		GROUP BY COD_CLIENTE
		)*/

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;

GO
```

# [dbo].[011_ENT_REL_CREDITO_GARANTIA]

```sql
/****** Object:  StoredProcedure [dbo].[011_ENT_REL_CREDITO_GARANTIA]    Script Date: 01/06/2026 20:11:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[011_ENT_REL_CREDITO_GARANTIA] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @FechaIni DATE,
		@FechaFin DATE;
	DECLARE @NombreJob NVARCHAR(128) = '[011_ENT_REL_CREDITO_GARANTIA]';

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[011_ENT_REL_CREDITO_GARANTIA]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[011_ENT_REL_CREDITO_GARANTIA]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END

		INSERT INTO [SILVER].[RR].[011_ENT_REL_CREDITO_GARANTIA] (
			[ID_CREDITO],
			[ID_GARANTIA],
			[PORCENTAJE_GARANTIA],
			[FECHA_INFO]
			)
		SELECT 
			ID_EXTERNO,
			B.NUM_GARANTIA,
			CASE 
				WHEN A.MON_SALDO = 0
					THEN 0 -- Handle division by zero
				WHEN B.MON_GARANTIA = 0
					THEN 0
				WHEN B.MON_GARANTIA > 1000000000
					THEN 9999.999999 -- Límite en valor absoluto de garantía
				WHEN A.MON_SALDO < 0.001
					THEN 9999.999999 -- Límite en valor mínimo del saldo
				WHEN (B.MON_GARANTIA / A.MON_SALDO) * 100 > 9999.999999
					THEN 9999.999999 -- Cap maximum value
				ELSE (B.MON_GARANTIA / A.MON_SALDO) * 100
				END AS PORCENTAJE_GARANTIA_PERS,
			EOMONTH(@FechaSistema) AS FECHA_INFO -- First day of month
		FROM [BRONZE].SAF.PR_CREDITOS A WITH (NOLOCK)
		INNER JOIN [BRONZE].SAF.PR_GARANTIAS B WITH (NOLOCK) ON A.COD_EMPRESA = B.COD_EMPRESA
			AND A.COD_CLIENTE = B.COD_CLIENTE
		WHERE 
		(A.IND_ESTADO = 'D' OR 
				(A.IND_ESTADO = 'C' 
				AND (A.FEC_CANCELACION >= @FechaIni
				AND A.FEC_CANCELACION < @FechaFin))
			)
		AND A.NUM_DESC_TIPO_CREDITO IN (/*COM*/12,13,14,15,28,2,5,18,27)
		AND A.ID_EXTERNO IS NOT NULL

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

# [dbo].[012_ENT_REL_CREDITO_GARANTIA_PER]

```sql
/****** Object:  StoredProcedure [dbo].[012_ENT_REL_CREDITO_GARANTIA_PER]    Script Date: 01/06/2026 20:11:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[012_ENT_REL_CREDITO_GARANTIA_PER] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @FechaIni DATE,
		@FechaFin DATE;
	DECLARE @NombreJob NVARCHAR(128) = '[012_ENT_REL_CREDITO_GARANTIA_PER]';

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[012_ENT_REL_CREDITO_GARANTIA_PER]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[012_ENT_REL_CREDITO_GARANTIA_PER]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END

		INSERT INTO [SILVER].[RR].[012_ENT_REL_CREDITO_GARANTIA_PER] (
			[ID_CREDITO],
			[ID_PERSONA],
			[PORCENTAJE_GARANTIA_PERS],
			[FECHA_INFO]
			)
		SELECT 
			A.ID_EXTERNO,
			B.COD_CLIENTE,
			/*CASE 
				WHEN A.MON_SALDO = 0
					THEN 0 -- Handle division by zero
				WHEN B.MON_GARANTIA = 0
					THEN 0
				WHEN B.MON_GARANTIA > 1000000000
					THEN 9999.999999 -- Límite en valor absoluto de garantía
				WHEN A.MON_SALDO < 0.001
					THEN 9999.999999 -- Límite en valor mínimo del saldo
				WHEN (B.MON_GARANTIA / A.MON_SALDO) * 100 > 9999.999999
					THEN 9999.999999 -- Cap maximum value
				ELSE (B.MON_GARANTIA / A.MON_SALDO) * 100
			END*/ 0 AS PORCENTAJE_GARANTIA_PERS, -- SE ASIGNA 0 DE ACUERDO A COMENTARIOS DE RIESGOS
			EOMONTH(@FechaSistema) AS FECHA_INFO
		FROM [BRONZE].SAF.PR_CREDITOS A WITH (NOLOCK)
		INNER JOIN [BRONZE].SAF.PR_GARANTIAS B WITH (NOLOCK) ON A.COD_EMPRESA = B.COD_EMPRESA
			AND A.COD_CLIENTE = B.COD_CLIENTE
		WHERE 
		(A.IND_ESTADO = 'D' OR   -- CREDITOS ACTIVOS O CANCELADOS DURANTE EL PERIODO A REPORTAR
				(A.IND_ESTADO = 'C' 
				AND (A.FEC_CANCELACION >= @FechaIni
				AND A.FEC_CANCELACION < @FechaFin))
			)
		AND A.NUM_DESC_TIPO_CREDITO IN (/*COM*/12,13,14,15,28,2,5,18,27)
		AND A.ID_EXTERNO IS NOT NULL


		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

# [dbo].[013_ENT_CREDITO]

```sql
/****** Object:  StoredProcedure [dbo].[013_ENT_CREDITO]    Script Date: 01/06/2026 20:11:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[013_ENT_CREDITO] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @FechaIni DATE,
		@FechaFin DATE;
	DECLARE @NombreJob NVARCHAR(128) = '[013_ENT_CREDITO]';

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[013_ENT_CREDITO]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[013_ENT_CREDITO]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END

		INSERT INTO [RR].[013_ENT_CREDITO] (
			[ID_PERSONA],
			[ID_CREDITO],
			[MONTO_CONDONACION],
			[MONTO_QUITAS],
			[MONTO_CASTIGOS],
			[MONTO_BONIFICACION],
			[MONTO_DESCUENTO],
			[MONTO_DACION],
			[CLASIFICACIONCONTABLE],
			[CONCURSOMERCANTIL],
			[ID_DISPOSICION],
			[FECHA_DISPOSICION],
			[FECHA_VENC_DISPOSICION],
			[MONEDA_DISPOSICION],
			[TIPO_TASA_DISP],
			[DIFER_TASA_DISP],
			[OPER_DIF_TASA_DISP],
			[FRECUENCIAREVISIONTASA_DISP],
			[NOMBRE_FACTORADO],
			[RFC_FACTORADO],
			[SALDO_CAPITAL_VIGENTE],
			[SALDO_CAPITAL_VENCIDO],
			[SALDO_INTERES_VIGENTE],
			[SALDO_INTERES_VENCIDO],
			[TASA_INTERES_ORDINARIA],
			[TASA_INTERES_MORATORIA],
			[MONTO_DISPUESTO_LINEA],
			[MONTO_PAGO_EXIGIBLE],
			[MONTO_CAPITAL_PAGADO],
			[MONTO_INTERES_ORDINARIO_PAGADO],
			[COMISIONES_PAGADAS],
			[MONTO_INTERES_MORATORIO],
			[OTROS_MONTOS_PAGADOS],
			[OTROS_MONTOS_PRINCIPAL],
			[DIAS_CALCULO_INTERES],
			[ETAPA_DETERIORO],
			[DIAS_ATRASO],
			[FECHA_ULTIMO_PAGO_EX_REC],
			[MONTO_FONDEADO_BD],
			[INSTITUTO_FONDEA],
			[TASA_INTERES_EFECTIVA],
			[HIPOTESIS_PRESUNCION],
			[RESERVAS_ADICIONALES],
			[MITIGANTE],
			[ID_LINEACREDITO],
			[FECHA_INFO],
			[ID_PROYECTO],
			[VALOR_MITIGANTE],
			[MONEDA_MITIGANTE],
			[SALDO_BASE_INTERES],
			[INTERESES_RESULTANTES],
			[INTERES_FUERA_BALANCE],
			[CODIGOCREDITOREESTRUCTURA],
			[PERIODICIDADCAPITAL],
			[FECHA_ETAPA3]
			)
		SELECT 
			CR.COD_CLIENTE AS ID_PERSONA,
			CR.ID_EXTERNO AS ID_CREDITO,
			ISNULL(COM.MONTO_CONDONACION,0) AS MONTO_CONDONACION,
			ISNULL(COM.MONTO_QUITAS,0) AS MONTO_QUITAS,
			ISNULL(COM.MONTO_CASTIGOS,0) AS MONTO_CASTIGOS,
			ISNULL(COM.MONTO_BONIFICACION,0) AS MONTO_BONIFICACION,
			ISNULL(COM.MONTO_DESCUENTO,0) AS MONTO_DESCUENTO,
			ISNULL(COM.MONTO_DAC,0) AS MONTO_DACION,
			ISNULL(CC.CLAVE,0) AS CLASIFICACIONCONTABLE, -- CRUZAR CON CATALOGO TABLA
			CASE 
				WHEN CLI.IND_CONCURSO_MERCANTIL = 'S' THEN 1
				WHEN CLI.IND_CONCURSO_MERCANTIL = 'N' THEN 2
				ELSE 2
			END AS CONCURSOMERCANTIL, -- CATALOGO
			ISNULL((SELECT COUNT(DISTINCT FORMAT(FEC_MOVIMIENTO, 'yyyyMM'))
			FROM BRONZE.SAF.PR_ENCABEZADO_PAGO A 
			INNER JOIN BRONZE.SAF.PR_DETALLE_PAGO B  ON A.ID_SECUENCIA = B.ID_SECUENCIA
			WHERE TIP_TRANSACCION = 3 AND COD_ESTADO = 'A' AND CAST(CR.NUM_CREDITO AS INT) = CAST(A.NUM_CREDITO AS INT) 
			GROUP BY A.NUM_CREDITO),0) AS ID_DISPOSICION,
			CR.FEC_ULT_DESEMBOLSO AS FECHA_DISPOSICION, --duda, validar
			IIF(CR.FEC_APERTURA = CR.FEC_ULT_DESEMBOLSO, CR.FEC_VENCIMIENTO, CR.FEC_ULT_DESEMBOLSO)  AS FECHA_VENC_DISPOSICION, --DUDA. VALIDAR
			CASE 
				WHEN CR.COD_MONEDA = 1 THEN 0
				WHEN CR.COD_MONEDA = 2 THEN 1
				WHEN CR.COD_MONEDA = 3 THEN 124
				WHEN CR.COD_MONEDA = 9 THEN 200
				ELSE 0 --CAT
			END AS MONEDA_DISPOSICION,
			CASE 
				WHEN CR.TIP_TASA = 'F' THEN 1
				WHEN CR.TIP_TASA = 'V' THEN 3
				ELSE 1
			END AS TIPO_TASA_DISP,
			CR.VAL_VARIACION_BASE AS DIFER_TASA_DISP,
			110 AS OPER_DIF_TASA_DISP, -- VALOR DE CATALOGO FIJO ACTUAL
			CASE 
				WHEN CR.TIP_TASA = 'F' THEN 0
				WHEN CR.TIP_TASA = 'V' THEN 30
				ELSE 0
			END AS FRECUENCIAREVISIONTASA_DISP,
			'' AS NOMBRE_FACTORADO, -- ACTUALMENTE NO APLICA
			'' AS RFC_FACTORADO, -- ACTUALMENTE NO APLICA
			ISNULL(COM.SALDO_CAPITAL_VIGENTE,0) AS SALDO_CAPITAL_VIGENTE,
			ISNULL(COM.AMORTIZACION_VENCIDA,0) /*ISNULL(COM.SALDO_CAPITAL_VENCIDO,0)*/ AS SALDO_CAPITAL_VENCIDO,
			ISNULL(COM.SALDO_INTERES_VIGENTE,0) AS SALDO_INTERES_VIGENTE,
			ISNULL(COM.INTERESES_EXIGIBLES,0) /*ISNULL(COM.SALDO_INTERES_VENCIDO,0)*/ AS SALDO_INTERES_VENCIDO,
			ISNULL(COM.TASA_INTERES,0) AS TASA_INTERES_ORDINARIA,
			ISNULL(COM.TASA_INTERES_MORATORIA,0) AS TASA_INTERES_MORATORIA,
			ISNULL(COM.ANTICIPOS_MINISTRACIONES,0) /*ISNULL(COM.MONTO_DISPUESTO_LINEA,0)*/ AS MONTO_DISPUESTO_LINEA,
			ISNULL(COM.MONTO_PAGO_EXIGIBLE,0) AS MONTO_PAGO_EXIGIBLE,
			ISNULL(COM.PAGADO_CAPITAL,0) AS MONTO_CAPITAL_PAGADO,
			ISNULL(COM.PAGADO_INTERESES,0) AS MONTO_INTERES_ORDINARIO_PAGADO,
			ISNULL(COM.PAGADO_COMISIONES,0) + ISNULL(COM.PAGADO_COMISIONES_IVA,0)/*ISNULL(COM.COMISIONES_PAGADAS,0)*/ AS COMISIONES_PAGADAS,
			ISNULL(COM.PAGADO_MORATORIOS,0) + ISNULL(COM.PAGADO_MORATORIOS_IVA,0)/*ISNULL(COM.MONTO_INTERES_MORATORIO,0)*/ AS MONTO_INTERES_MORATORIO,
			ISNULL(COM.OTROS_MONTOS_PAGADOS,0) AS OTROS_MONTOS_PAGADOS,
			-1 * (ISNULL(COM.CARTERA_SHF_70,0) + ISNULL(COM.SALDO_INI_AJUSTE,0) + ISNULL(COM.MONTO_CASTIGOS,0))/*ISNULL(COM.OTROS_MONTOS_PRINCIPAL,0)*/ AS OTROS_MONTOS_PRINCIPAL,
			CASE 
				-- Si el crédito no es nuevo ni liquidado
				WHEN CR.FEC_APERTURA < @FechaIni AND (CR.FEC_CANCELACION IS NULL OR CR.FEC_CANCELACION >= @FechaFin) 
				THEN 30

				-- Si el crédito se dispone y se liquida en el mismo mes
				WHEN CR.FEC_APERTURA >= @FechaIni AND CR.FEC_APERTURA < @FechaFin
					 AND CR.FEC_CANCELACION IS NOT NULL
					 AND CR.FEC_CANCELACION >= @FechaIni AND CR.FEC_CANCELACION < @FechaFin
				THEN DATEDIFF(DAY, CR.FEC_APERTURA, CR.FEC_CANCELACION) -1 --DIA DE LA APERTURA

				-- Si se dispone en el mes y no se liquida
				WHEN CR.FEC_APERTURA >= @FechaIni AND CR.FEC_APERTURA < @FechaFin
					 AND (CR.FEC_CANCELACION IS NULL OR CR.FEC_CANCELACION >= @FechaFin)
				THEN DATEDIFF(DAY, CR.FEC_APERTURA, @FechaFin) - 2 --DIA DE APERTURA Y PRIMER DIA DEL MES SIG

				-- Si se liquida en el mes pero fue dispuesto antes
				WHEN CR.FEC_APERTURA < @FechaIni 
					 AND CR.FEC_CANCELACION IS NOT NULL 
					 AND CR.FEC_CANCELACION >= @FechaIni AND CR.FEC_CANCELACION < @FechaFin
				THEN DATEDIFF(DAY, @FechaIni, CR.FEC_CANCELACION)

				ELSE 0
			END AS DIAS_CALCULO_INTERES,
			IIF(CR.IND_ETAPA = 0, 1,CR.IND_ETAPA) AS ETAPA_DETERIORO,
			
			/*ISNULL((SELECT DATEDIFF(DAY,MIN(FEC_COBRO),@FechaFin) -1 FROM BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO
			WHERE IND_ESTADO= 'P'AND FEC_COBRO >= @FechaIni AND FEC_COBRO < @FechaFin
			AND CAST(NUM_CREDITO AS INT)= CAST(CR.NUM_CREDITO AS INT)
			GROUP BY NUM_CREDITO),0)*/ ISNULL(COM.DIAS_VENCIDOS,0) AS DIAS_ATRASO, -- SI NO ES CORRECTO, TOMAR DEL CIERRE
			
			ISNULL((SELECT MAX(FEC_COBRO) FROM BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO
			WHERE IND_ESTADO= 'A'
			AND CAST(NUM_CREDITO AS INT)= CAST(CR.NUM_CREDITO AS INT)
			GROUP BY NUM_CREDITO),0) AS FECHA_ULTIMO_PAGO_EX_REC,
			IIF(CR.COD_ORIGEN != '008',ISNULL(COM.SALDO_PRINCIPAL_FIN,0) + ISNULL(COM.SALDO_INTERESES_BALANCE,0),0) /*ISNULL(COM.MONTO_FONDEADO_BD,0)*/ AS MONTO_FONDEADO_BD,
			CASE
				WHEN COM.CEDIDO LIKE '%SHF%' THEN '037168' -- Sociedad Hipotecaria Federal
				WHEN COM.CEDIDO LIKE '%NAFIN%' THEN '037135' -- Nacional Financiera
				WHEN COM.CEDIDO LIKE '%BANCOMEXT%' OR COM.CEDIDO LIKE '%COMERCIO EXTERIOR%' THEN '037006' -- Banco Nacional de Comercio Exterior
				WHEN COM.CEDIDO LIKE '%BANOBRAS%' OR COM.CEDIDO LIKE '%OBRAS%' THEN '037009' -- Banco Nacional de Obras y Servicios Públicos
				WHEN COM.CEDIDO LIKE '%BANJERCITO%' OR COM.CEDIDO LIKE '%EJÉRCITO%' THEN '037019' -- Banco Nacional del Ejército, fuerza Aérea y Armada
				WHEN COM.CEDIDO LIKE '%BANSEFI%' OR COM.CEDIDO LIKE '%AHORRO NACIONAL%' THEN '037166' -- Banco del Ahorro Nacional y Servicios Financieros
				WHEN COM.CEDIDO LIKE '%FIFOMI%' OR COM.CEDIDO LIKE '%FOMENTO MINERO%' THEN '031013' -- Fideicomiso de Fomento Minero
				WHEN COM.CEDIDO LIKE '%FIRA%' OR COM.CEDIDO LIKE '%FINANCIAMIENTOS AGRO%' THEN '031012' -- Fondo Especial para Financiamientos Agropecuarios
				WHEN COM.CEDIDO LIKE '%AGRIC%' THEN '031004' -- Fondo de Garantía y Fomento para la Agricultura Ganadería y Avicultura
				WHEN COM.CEDIDO LIKE '%PESQUER%' THEN '031010' -- Fondo de Garantía y Fomento para las Actividades Pesqueras
				WHEN COM.CEDIDO LIKE '%ASISTENCIA%' THEN '031011' -- Fondo Especial de Asistencia Técnica y Garantía de Créditos Agropecuarios
				WHEN COM.CEDIDO LIKE '%DESARROLLO COMERCIAL%' THEN '031002' -- Fondo para el Desarrollo Comercial
				WHEN COM.CEDIDO LIKE '%VIVIENDA%' OR COM.CEDIDO LIKE '%DAMNIFICADOS%' THEN '031001' -- Fondo de Operación y Financiamiento Bancario a la Vivienda
				WHEN COM.CEDIDO LIKE '%FINANCIERA NACIONAL%' THEN '050001' -- Financiera Nacional de Desarrollo Agropecuario, Rural, Forestal y Pesquero
				ELSE '0' -- Sin fondeo por parte de la Banca de Desarrollo o Fondo de Fomento
			END AS INSTITUTO_FONDEA, -- CATALOGO
			ISNULL(TE.TASA_EFECTIVA,0) AS TASA_INTERES_EFECTIVA,
			0 AS HIPOTESIS_PRESUNCION, --ACTUALMENTE NO APLICA
			CASE
				WHEN CR.IND_ESTADO_CRED = 'E' AND CR.IND_ETAPA = 3 THEN ISNULL(RAH.MONTO_RESERVAS,0)
				ELSE 0
			END AS RESERVAS_ADICIONALES,
			0 AS MITIGANTE,
			CR.ID_EXTERNO AS ID_LINEACREDITO,
			EOMONTH(@FechaSistema)FECHA_INFO,
			PRO.COD_TIPO_PROYECTO AS ID_PROYECTO,
			CR.MON_SALDO /*ISNULL(COM.MONTO_FONDEADO_BD,0)*/ AS VALOR_MITIGANTE, -- MONTO_FONDEADO = SALDO_CONTABLE 
			0 AS MONEDA_MITIGANTE,
			ISNULL((SELECT TOP 1 MON_BASE FROM BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO WHERE
			CAST(NUM_CREDITO AS INT)= CAST(CR.NUM_CREDITO AS INT) AND FEC_COBRO >= @FechaIni AND FEC_COBRO < @FechaFin
			AND COD_RUBRO = 'INTERESES'),0) AS SALDO_BASE_INTERES,
			ISNULL((SELECT TOP 1 MON_RUBRO FROM BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO WHERE
			CAST(NUM_CREDITO AS INT)= CAST(CR.NUM_CREDITO AS INT) AND FEC_COBRO >= @FechaIni AND FEC_COBRO < @FechaFin
			AND COD_RUBRO = 'INTERESES'),0) AS INTERESES_RESULTANTES,
			ISNULL(COM.SALDO_INTERESES_ORDEN,0) AS INTERES_FUERA_BALANCE,
			CASE
				WHEN CR.IND_ESTADO_CRED= 'E' THEN (SELECT TOP 1 ID_EXTERNO FROM BRONZE.SAF.PR_CREDITOS WHERE CR.NUM_CONTRATO = NUM_CREDITO AND IND_ESTADO_CRED != 'E')
				ELSE ''
			END AS CODIGOCREDITOREESTRUCTURA,
			CASE
				WHEN CR.PER_CUOTA = 'SE' THEN 2  -- Pagos semanales
				WHEN CR.PER_CUOTA = 'ME' THEN 5  -- Pagos mensuales
				WHEN CR.PER_CUOTA = 'TR' THEN 7  -- Pagos trimestrales
				ELSE 10                         -- Pago con varias periodicidades (default)
			END AS PERIODICIDADCAPITAL, -- VALOR DE CATALOGO - MES
			(SELECT fec_reclasificacion
			FROM BRONZE.SAF.PR_RECLASIFICACION_ETAPA_RIESGO
			where CAST(NUM_CREDITO AS INT)= CAST(CR.NUM_CREDITO AS INT)
				and ind_etapa_act = 3
				and fec_reclasificacion In (
					SELECT max(fec_reclasificacion)
					FROM BRONZE.SAF.PR_RECLASIFICACION_ETAPA_RIESGO
					where CAST(NUM_CREDITO AS INT)= CAST(CR.NUM_CREDITO AS INT) and not (ind_etapa_ant = 3 )
				)) AS FECHA_ETAPA3
		--FROM	BRONZE.SAF.PR_CREDITOS CR
		FROM (SELECT * FROM ( SELECT *, ROW_NUMBER() OVER(PARTITION BY COD_CLIENTE ORDER BY IND_ESTADO DESC, NUM_CREDITO DESC ) AS RN
		FROM BRONZE.SAF.PR_CREDITOS) AS T WHERE RN = 1) CR 
		LEFT JOIN (SELECT * FROM BRONZE.CIERRE.CMR_REP WHERE FECHA_SISTEMA = EOMONTH(@FechaIni))COM ON CR.ID_EXTERNO = COM.ID_EXTERNO
		LEFT JOIN BRONZE.SAF.CL_CLIENTES CLI ON CR.COD_CLIENTE = CLI.COD_CLIENTE
		LEFT JOIN BRONZE.SAF.PR_INF_PROYECTO PRO ON CR.NUM_CREDITO = PRO.NUM_CREDITO
		LEFT JOIN (SELECT * FROM BRONZE.INF.CONSOLIDADO_MENSUAL WHERE FECHA >= @FechaIni AND FECHA < @FechaFin) TE ON CR.ID_EXTERNO = TE.ID_EXTERNO
		LEFT JOIN BRONZE.RR.CLASIFICACION_CONTABLE CC ON CONCAT(COM.CREDITO_TIPO, ', ', COM.CARTERA_TIPO, ', ', COM.ETAPA_RIESGO)= CONCAT(CC.REFERENCIA_PADRE,', ' ,CC.SITUACION_CARTERA)
		LEFT JOIN BRONZE.RR.RESERVAS_ADICIONALES_HISTORICO RAH ON CR.ID_EXTERNO = RAH.ID_EXTERNO
		WHERE CR.NUM_CREDITO IN
		(
		SELECT MAX(NUM_CREDITO) FROM BRONZE.SAF.PR_CREDITOS  -- SE OBTIENE UN SOLO CRÉDITO POR CLIENTE PARA EVITAR DUPLICIDAD
		WHERE 
		(IND_ESTADO = 'D' OR 
				(IND_ESTADO = 'C' 
				AND (FEC_CANCELACION >= @FechaIni
				AND FEC_CANCELACION < @FechaFin))
			) 
		AND NUM_DESC_TIPO_CREDITO IN (/*COM*/12,13,14,15,28,2,5,18,27/*,IND 1,3,4,9,10,11,16,17*/)
		--AND IND_LINEA != 'N'
		GROUP BY COD_CLIENTE
		)


		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[014_ENT_LINEACREDITO]    Script Date: 01/06/2026 20:11:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[014_ENT_LINEACREDITO] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @FechaIni DATE,
		@FechaFin DATE;
	DECLARE @NombreJob NVARCHAR(128) = '[014_ENT_LINEACREDITO]';

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[014_ENT_LINEACREDITO]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[014_ENT_LINEACREDITO]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));
			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[014_ENT_LINEACREDITO] (
			[ID_LINEACREDITO],
			[ID_PERSONA],
			[DESTINOCREDITOCC],
			[MONTOLINEAAUT],
			[FECHA_OTORG_LINEA],
			[FECHA_VENC_LINEA],
			[FECHA_MAX_DISP],
			[DISPOSICION_CREDITO],
			[REVOCABLE_IRREVOCABLE],
			[PRELACION],
			[INSTITUTO_FONDEA],
			[TIPO_ALTA],
			[MONEDA_LINEA],
			[ES_MULTIMONEDA],
			[TIPO_TASA],
			[DIFER_TASA],
			[OPER_DIF_TASA],
			[FRECUENCIAREVISIONTASA],
			[PERIODICIDADCAPITAL],
			[PERIODICIDADINTERES],
			[MESESGRACIACAP],
			[MESESGRACIAINTERESES],
			[COMISIONAPERT_TASA],
			[COMISIONAPERT_MONTO],
			[COMISIONDISP_TASA],
			[COMISIONDISP_MONTO],
			[CAT],
			[MONTOLINEASINA],
			[MONTOPRIMASANUALES],
			[LOCALIDADDESTINO],
			[ACTIVIDADECONOMICADESTINO],
			[TIPO_BAJA],
			[TIPO_BAJA_OTRA],
			[FECHA_INFO],
			[FOLIO_CONSULTA_BURO],
			[CODIGOCREDITOREESTRUCTURA]
			)
		SELECT
			A.ID_EXTERNO AS ID_LINEACREDITO,
			A.COD_CLIENTE AS ID_PERSONA,
			CASE
				WHEN B.DestCredito = 'Adquisición de terreno para vivienda' THEN 1
				WHEN B.DestCredito = 'Adquisición de terreno y construcción simultánea' THEN 2
				WHEN B.DestCredito = 'Construcción de vivienda propia' THEN 3
				WHEN B.DestCredito = 'Adquisición de vivienda nueva' THEN 4
				WHEN B.DestCredito = 'Adquisición de vivienda usada' THEN 5
				WHEN B.DestCredito = 'Mejoras, ampliaciones y/o remodelaciones' THEN 6
				WHEN B.DestCredito = 'Pago de Pasivos Hipotecarios' THEN 7
				WHEN B.DestCredito = 'Crédito para Liquidez' THEN 8
				WHEN B.DestCredito = 'Créditos a exempleados de la Entidad' THEN 11
				ELSE 0
			END AS DESTINOCREDITOCC,
			ISNULL(A.MON_CREDITO,0) AS MONTOLINEAAUT,
			A.FEC_APERTURA AS FECHA_OTORG_LINEA,
			A.FEC_VENCIMIENTO AS FECHA_VENC_LINEA,
			CASE
				WHEN A.NUM_DESC_TIPO_CREDITO IN (2,5,18,27) THEN A.FEC_VENCIMIENTO
				ELSE DATEADD(MONTH,1,A.FEC_APERTURA)
			END AS FECHA_MAX_DISP,
			CASE
				WHEN A.NUM_DESC_TIPO_CREDITO = 7 THEN 104 --SINDICADO
				WHEN A.IND_LINCRED_REVOLVENTE = 'N' THEN 102 --No Revolvente en múltiples disposiciones
				WHEN A.IND_LINCRED_REVOLVENTE IS NULL THEN 103 --Revolvente
				ELSE 101
			END AS DISPOSICION_CREDITO,
			CASE
				WHEN B.LineaCreditoRevocableOIrrevocable = 'Línea de Crédito Revocable' THEN 1
				WHEN B.LineaCreditoRevocableOIrrevocable = 'Línea de Crédito Irrevocable' THEN 2
				WHEN B.LineaCreditoRevocableOIrrevocable = 'Carta de Crédito' THEN 181
				ELSE 1
			END AS REVOCABLE_IRREVOCABLE,
			CASE
				WHEN B.PrelacionDePago = 'Prelación de Pago Preferente' THEN 1
				WHEN B.PrelacionDePago = 'Prelación de Pago Subordinado' THEN 2
				WHEN B.PrelacionDePago = 'Carta de Crédito' THEN 181
				ELSE 1
			END AS PRELACION,
			IIF(A.COD_ORIGEN = '008', 0,ISNULL(IFO.CLAVE,0)) AS INSTITUTO_FONDEA,
			ISNULL(RR.TIP_ALTA,0) AS TIPO_ALTA,
			CASE
				WHEN A.COD_MONEDA = 1 THEN 0
				WHEN A.COD_MONEDA = 2 THEN 1
				WHEN A.COD_MONEDA = 3 THEN 124
				WHEN A.COD_MONEDA = 9 THEN 200
			END AS MONEDA_LINEA,
			0 AS ES_MULTIMONEDA,
			CASE
				WHEN A.TIP_TASA = 'F' THEN 1
				WHEN A.TIP_TASA = 'V' THEN 3
				ELSE 2 --MIXTA
			END AS TIPO_TASA,
			A.VAL_VARIACION_BASE AS DIFER_TASA, 
			CASE
				WHEN B.OperacionDiferencialTasaReferencia = 'Suma del diferencial sobre la tasa de referencia' THEN 110
				WHEN B.OperacionDiferencialTasaReferencia = 'Multiplicación del diferencial sobre la tasa de referencia' THEN 111
				WHEN B.OperacionDiferencialTasaReferencia = 'Resta del diferencial sobre la tasa de referencia' THEN 112
				WHEN B.OperacionDiferencialTasaReferencia = 'Otra operación sobre la tasa de referencia' THEN 119
				ELSE 119
			END AS OPER_DIF_TASA,
			CASE
				WHEN A.PER_REV_TASA = 'ME' THEN 30
				WHEN A.PER_REV_TASA = 'TR' THEN 90
				ELSE 30
			END AS FRECUENCIAREVISIONTASA,
			CASE
				WHEN B.PeriodicidadPagosCapital =  'Amortización única al vencimiento' THEN 1
				WHEN B.PeriodicidadPagosCapital = 'Pagos semanales' THEN 2
				WHEN B.PeriodicidadPagosCapital = 'Pagos quincenales' THEN 4
				WHEN B.PeriodicidadPagosCapital = 'Pagos mensuales' THEN 5
				WHEN B.PeriodicidadPagosCapital = 'Pagos bimestrales' THEN 6
				WHEN B.PeriodicidadPagosCapital = 'Pagos trimestrales' THEN 7
				WHEN B.PeriodicidadPagosCapital = 'Pagos semestrales' THEN 8
				WHEN B.PeriodicidadPagosCapital = 'Pagos anuales' THEN 9
				ELSE 5
			END AS PERIODICIDADCAPITAL,
			CASE
				WHEN B. PeriodicidadPagosIntereses = 'Pago al vencimiento' THEN 1
				WHEN B. PeriodicidadPagosIntereses = 'Pagos semanales' THEN 2
				WHEN B. PeriodicidadPagosIntereses = 'Pagos quincenales' THEN 4 
				WHEN B. PeriodicidadPagosIntereses = 'Pagos mensuales' THEN 5
				WHEN B. PeriodicidadPagosIntereses = 'Pagos bimestrales' THEN 6
				WHEN B. PeriodicidadPagosIntereses = 'Pagos trimestrales' THEN 7
				WHEN B. PeriodicidadPagosIntereses = 'Pagos semestrales' THEN 8
				WHEN B. PeriodicidadPagosIntereses = 'Pagos anuales' THEN 9
				ELSE 5
			END AS PERIODICIDADINTERES,
			ISNULL(B.MesesGraciaAmortizarCapital,0) AS MESESGRACIACAP,
			ISNULL(B.MesesGraciaPagoIntereses,0) AS MESESGRACIAINTERESES,
			ISNULL(B.ComisionAperturaTasa,0) AS COMISIONAPERT_TASA,
			ISNULL(B.ComisionAperturaMonto,0) AS COMISIONAPERT_MONTO,
			ISNULL(B.ComisionDisposicionTasa,0) AS COMISIONDISP_TASA,
			ISNULL(B.ComisionDisposicionMonto,0) AS COMISIONDISP_MONTO,
			ISNULL(A.MON_CAT,0) AS CAT,
			A.MON_CREDITO AS MONTOLINEASINA,
			ISNULL(B.MontoPrimasAnualesSegurosObligatorios,0) AS MONTOPRIMASANUALES,
			ISNULL(CONCAT('484',
			IIF(C.COD_PAIS IS NULL, '00', C.COD_PAIS),
			IIF(C.COD_PROVINCIA IS NULL, '000', C.COD_PROVINCIA),
			'0001'),'484000000001') AS LOCALIDADDESTINO,
			ISNULL(LEFT(D.CODIGO,8),'') AS ACTIVIDADECONOMICADESTINO,
			ISNULL(RR.TIP_BAJA,0) AS TIPO_BAJA,
			'0' AS TIPO_BAJA_OTRA,
			EOMONTH(@FechaSistema) AS FECHA_INFO,
			B.NumeroConsultaSociedadInformacionCrediticia AS FOLIO_CONSULTA_BURO,
			CASE
				WHEN A.IND_ESTADO_CRED= 'E' THEN (SELECT TOP 1 ID_EXTERNO FROM BRONZE.SAF.PR_CREDITOS WHERE NUM_CREDITO = A.NUM_CONTRATO AND IND_ESTADO_CRED != 'E')
				ELSE ''
			END AS CODIGOCREDITOREESTRUCTURA
		FROM (SELECT * FROM ( SELECT *, ROW_NUMBER() OVER(PARTITION BY ID_EXTERNO ORDER BY IND_ESTADO DESC, NUM_CREDITO DESC) AS RN
		FROM BRONZE.SAF.PR_CREDITOS) AS T WHERE RN = 1) A
		LEFT JOIN BRONZE.INF.CREDITO_EMPRESARIAL B ON A.ID_EXTERNO = B.IDCREDITO
		LEFT JOIN (SELECT * FROM BRONZE.SAF.CL_DIR_CLIENTES WHERE IND_ESTADO= 'A' AND COD_DIRECCION = 1) C ON A.COD_CLIENTE = C.COD_CLIENTE
		LEFT JOIN BRONZE.[RR].[ActividadEconomica_V2] D ON B.ActividadEconomicaDestinoCredito = D.DESCRIPCION
		LEFT JOIN (SELECT * FROM BRONZE.SAF.PR_CREDITOS_HISTORICO WHERE FEC_CORTE = DATEADD(MONTH,-1,EOMONTH(@FechaSistema)))E ON A.ID_EXTERNO = E.ID_EXTERNO
		AND A.NUM_CREDITO = E.NUM_CREDITO
		LEFT JOIN BRONZE.SAF.PR_ORIGEN_FONDOS OFO ON A.COD_ORIGEN = OFO.COD_ORIGEN
		LEFT JOIN BRONZE.RR.INSTITUCIONES_FONDEADORAS IFO ON OFO.DES_ORIGEN LIKE IFO.DESC_NORMALIZADOS
		LEFT JOIN BRONZE.SAF.PR_INF_RR_X_CREDITO RR ON A.NUM_CREDITO = RR.NUM_CREDITO
		WHERE 
		(A.IND_ESTADO != 'N' AND (A.FEC_APERTURA >= @FechaIni AND A.FEC_APERTURA < @FechaFin)) -- CREDITOS NUEVOS
		OR (A.IND_ESTADO = 'C' AND (A.FEC_CANCELACION >= @FechaIni AND A.FEC_CANCELACION < @FechaFin)) -- CREDITOS CANCELADOS
		AND A.NUM_DESC_TIPO_CREDITO IN (/*COM*/12,13,14,15,28,2,5,18,27,/*IND*/1,3,4,9,10,11,16,17)
	


		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[016_ENT_ANEXO19]    Script Date: 01/06/2026 20:11:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[016_ENT_ANEXO19] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[016_ENT_ANEXO19]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE,		
		@FechaIni2 DATE,
		@FechaFin2 DATE,
		@Fecha DATE,
		@TipoRep INT,
		@MesIni INT;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni)
		SET @TipoRep = 3 --meses del periodo
		SET @MesIni = FLOOR((month(@FechaSistema) + @TipoRep - 1) / @TipoRep) * @TipoRep - @TipoRep + 1
		SET @FechaIni2 = DATEfromparts(year(@FechaSistema), @MesIni, 1)
		SET @FechaFin2 = DATEadd(month, @TipoRep, @FechaIni)--------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[016_ENT_ANEXO19]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[016_ENT_ANEXO19]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[016_ENT_ANEXO19] (
			[ID_PERSONA],
			[TIPO_PROYECTO],
			[FECHA_INICIO_OPERACION],
			[POR_EXPOS_29_INST_BAN],
			[VP_FEGP],
			[ETAPA_PROYECTO],
			[ANALISIS_ESTRES],
			[ESTRUCTURA_FINANCIERA],
			[RIES_POL_ENT_REG],
			[TASA_DE_DESCUENTO_VP],
			[RIESGO_EVENTOS_FUERZA_MAY],
			[ADQ_APOYOS_APROBAC],
			[CUMPLIMIENTO_CONTRATOS],
			[RIESGO_CONSTRUCCION],
			[TIPO_CONTRATO_CONSTRUCCION],
			[RIESGO_OPERATIVO],
			[PRENDA_ACTIVOS],
			[CONTROL_INST_FLUJO_EFECTIVO],
			[FONDOS_RESERVA],
			[HISTORIAL_PATROCINADOR],
			[COBERTURA_SEGUROS],
			[ACTIVO_BRGR],
			[SOBRECOSTO],
			[INGRESOS_PROPIOS_PROY],
			[GASTOS_TOTALES_PROY],
			[IMPACTO_FISCAL_PROY],
			[PAGO_DEDUDA_12MESES],
			[FLUJO_EFECTIVO_ADICIONAL],
			[FECHA_INFO],
			[ID_PROYECTO],
			[VENTAS_NETAS_INGRESOS],
			[RESERVA_PROYECTO],
			[VP_FEGP_12MESES],
			[VALOR_ACTIVO_SUBY],
			[FECHA_INFO_FIN],
			[FECHA_INFO_BURO]
			)
		SELECT 
			A.COD_CLIENTE AS ID_PERSONA,
			CASE
				WHEN B.TipoProyecto ='Financiamiento de Proyectos.' THEN 'FP'
				WHEN B.TipoProyecto ='Financiamiento de Proyectos Sin contrato de Gtia o Cob Precios' THEN 'FPSC'
				WHEN B.TipoProyecto ='Financiamiento de Bienes' THEN 'FA'
				WHEN B.TipoProyecto ='Financiamiento de Mercancías o productos básicos.' THEN 'FC'
				WHEN B.TipoProyecto ='Bienes Raíces Generadores de Rentas' THEN 'BRGR'
				ELSE 'BRGR'
			END AS TIPO_PROYECTO,
			ISNULL(B.FechaFirma,0) AS FECHA_INICIO_OPERACION,
			CAST(NULL AS NUMERIC(10,6)) AS POR_EXPOS_29_INST_BAN,
			ISNULL(CAST(F.VP_FEGP AS NUMERIC(21,2)),0) AS VP_FEGP,
			ISNULL(E.CLAVE,0) AS ETAPA_PROYECTO,
			ISNULL(B.AnalisisEstres,0) AS ANALISIS_ESTRES,
			ISNULL(B.EstructuraFinanciera,0) AS ESTRUCTURA_FINANCIERA,
			CASE
				WHEN B.RiesgoPolitico = 'Nula exposición' THEN 1
				WHEN B.RiesgoPolitico = 'Baja exposición' THEN 2
				WHEN B.RiesgoPolitico = 'Moderada exposición' THEN 3
				WHEN B.RiesgoPolitico = 'Alta exposición' THEN 4
				WHEN B.RiesgoPolitico = 'Sin información' THEN 0
			ELSE 0
			END AS RIES_POL_ENT_REG,
			ISNULL(F.TASA_DE_DESCUENTO_VP,0),
			CASE
				WHEN B.RiesgoEventosFuerzaMayor = 'No se ve afectado' THEN 1
				WHEN B.RiesgoEventosFuerzaMayor = 'Se encuentra o es proceso a riesgo' THEN 2
				WHEN B.RiesgoEventosFuerzaMayor = 'Sin información' THEN 0
			ELSE 0 END AS RIESGO_EVENTOS_FUERZA_MAY,
			CASE
				WHEN B.ImpactoAmbiental = 'Cuenta con estudio impacto ambiental' THEN 1
				WHEN B.ImpactoAmbiental = 'No cuenta con estudio impacto ambiental' THEN 2
				WHEN B.ImpactoAmbiental = 'Sin información' THEN 0
			ELSE 0
			END AS ADQ_APOYOS_APROBAC,
			CASE
				WHEN B.CumplimientoContratos = 'Cuenta con estudio impacto ambiental' THEN 1
				WHEN B.CumplimientoContratos = 'No cuenta con estudio impacto ambiental' THEN 2
				WHEN B.CumplimientoContratos = 'Sin información' THEN 0
			ELSE 0
			END AS CUMPLIMIENTO_CONTRATOS,
			ISNULL(B.RiesgoConstruccion,0) AS RIESGO_CONSTRUCCION,
			ISNULL(B.ContratoConstruccion,0) AS TIPO_CONTRATO_CONSTRUCCION,
			ISNULL(B.ContratoOperacion,0) AS RIESGO_OPERATIVO,
			ISNULL(B.PrendaActivos,0) AS PRENDA_ACTIVOS,
			ISNULL(B.ControlFlujoEfectivo,0) AS CONTROL_INST_FLUJO_EFECTIVO,
			ISNULL(B.FondoReserva,0) AS FONDOS_RESERVA,
			ISNULL(B.HistorialPatrocinador,0) AS HISTORIAL_PATROCINADOR,
			ISNULL(B.CoberturaSeguros,0) AS COBERTURA_SEGUROS,
			ISNULL(B.BRGR,0) AS ACTIVO_BRGR,
			ISNULL(F.SOBRECOSTO,0),
			0.0 AS INGRESOS_PROPIOS_PROY,
			0.0 AS GASTOS_TOTALES_PROY,
			0.0 AS IMPACTO_FISCAL_PROY,
			F.PAGO_DEDUDA_12MESES,
			F.FLUJO_EFECTIVO_ADICIONAL,
			EOMONTH(@FechaSistema) AS FECHA_INFO,
			ISNULL(A.ID_EXTERNO,'') AS ID_PROYECTO,
			0.0 AS VENTAS_NETAS_INGRESOS,
			0.0 AS RESERVA_PROYECTO,
			ISNULL(F.VP_FEGP_12MESES,0),
			0 AS VALOR_ACTIVO_SUBY,
			CAST(NULL AS DATE) AS FECHA_INFO_FIN,
			ISNULL(B.FechaConsultaBuro,0) AS FECHA_INFO_BURO
		FROM BRONZE.SAF.PR_CREDITOS A
		LEFT JOIN BRONZE.INF.CREDITO_EMPRESARIAL B ON A.ID_EXTERNO = B.IDCREDITO
		LEFT JOIN (SELECT * FROM BRONZE.PO.SAF_CIERRE_CMR_REP WHERE FECHA_SISTEMA = EOMONTH(@FechaSistema)) C ON A.ID_EXTERNO = C.ID_EXTERNO
		LEFT JOIN (SELECT * FROM BRONZE.LMDA.FLUJOS WHERE FECHA_VPN BETWEEN @FechaIni2 AND @FechaFin2 ) F ON A.ID_EXTERNO = F.ID_EXTERNO 
		LEFT JOIN BRONZE.RR.FLUJO_ETAPAS E ON E.DESCRIPCION = F.ETAPA_PROYECTO
		WHERE A.NUM_CREDITO IN
		(
		SELECT MAX(NUM_CREDITO) FROM BRONZE.SAF.PR_CREDITOS  -- SE OBTIENE UN SOLO CRÉDITO POR CLIENTE PARA EVITAR DUPLICIDAD
		WHERE 
		(IND_ESTADO = 'D' OR 
				(IND_ESTADO = 'C' 
				AND (FEC_CANCELACION >= @FechaIni
				AND FEC_CANCELACION < @FechaFin))
			)
		AND TIP_CREDITO = 16 
		AND A.NUM_DESC_TIPO_CREDITO IN (12,13,14,15,28,2,5,18,27)
		AND IND_LINEA = 'N'
		GROUP BY COD_CLIENTE
		)

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[018_ENT_ANEXO21]    Script Date: 01/06/2026 20:11:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[018_ENT_ANEXO21] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[018_ENT_ANEXO21]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[018_ENT_ANEXO21]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[018_ENT_ANEXO21]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[018_ENT_ANEXO21] (
			[ID_PERSONA],
			[QUITAS_CASTIGO_REEST],
			[VENTAS_NETAS_INGRESOS],
			[POR_PAGO_TIEMPO_INSTCRED],
			[POR_PAGO_TIEMPO_INSTNOBAN],
			[CREDITO_PYME],
			[FECHA_INFO_BURO],
			[FECHA_INFO_FIN],
			[HITSIC],
			[PI100],
			[GARANTIA_LEY_FED],
			[LUGAR_RADICA],
			[DOMICILIO_EXT],
			[ES_GARANTE],
			[FECHA_INFO]
			)
		SELECT 
			A.COD_CLIENTE AS ID_PERSONA,
			0 AS QUITAS_CASTIGO_REEST, --RELLENA BW
			ISNULL(C.VentasNetasTotalesAnuales,0) AS VENTAS_NETAS_INGRESOS,
			0 AS POR_PAGO_TIEMPO_INSTCRED, --RELLENA BW
			0 AS POR_PAGO_TIEMPO_INSTNOBAN, --RELLENA BW
			1 AS CREDITO_PYME, --CATALOGO
			CAST(NULL AS DATE) AS FECHA_INFO_BURO, --RELLENA BW
			C.FechaEstadosFinancieros AS FECHA_INFO_FIN, 
			0 AS HITSIC, --RELLENA BW
			(SELECT IIF(SUM(BANDERA)=0,0,10) FROM
				(SELECT COD_CLIENTE, IIF(IND_ETAPA=3 OR FEC_VENCIMIENTO < EOMONTH(@FechaSistema),1,0) AS BANDERA
				FROM BRONZE.SAF.PR_CREDITOS)B WHERE B.COD_CLIENTE = A.COD_CLIENTE
			GROUP BY COD_CLIENTE ) AS PI100, --CATALOGO
			790 AS GARANTIA_LEY_FED, -- CATALOGO
			CASE
				WHEN MAX(B.COD_PAIS) IS NULL THEN 1 -- DEFAULT MEXICO SI NO HAY INFORMACIÓN
				WHEN MAX(B.COD_PAIS) BETWEEN '01' AND '32' THEN 1 -- MEXICO
				ELSE 2 -- EXTRANJERO
			END AS LUGAR_RADICA, --CATALOGO
			CAST(NULL AS INT) AS DOMICILIO_EXT,
			CASE 
				WHEN C.TipoDeGarante = 'Garantía de Primeras Pérdidas' THEN 1
				WHEN C.TipoDeGarante = 'Garantía de Paso y Medida' THEN 2
				WHEN C.TipoDeGarante = 'Garantía Personal Aval' THEN 3
				WHEN C.TipoDeGarante = 'Garantía Personal Obligado Solidario' THEN 4
				WHEN C.TipoDeGarante = 'Garantía Personal Factorado con Recurso' THEN 5
				WHEN C.TipoDeGarante = 'No Aplica' THEN 0
				ELSE CAST(NULL AS INT)
			END AS ES_GARANTE,-- CATALOGO
			EOMONTH(@FechaSistema) AS FECHA_INFO
		FROM BRONZE.SAF.PR_CREDITOS A
		LEFT JOIN (SELECT * FROM BRONZE.SAF.CL_DIR_CLIENTES WHERE TIP_DIRECCION = 'C' AND IND_ESTADO= 'A' AND COD_DIRECCION = 1) B ON A.COD_CLIENTE = B.COD_CLIENTE
		LEFT JOIN BRONZE.INF.CREDITO_EMPRESARIAL C ON A.ID_EXTERNO = C.IDCREDITO
		LEFT JOIN BRONZE.PO.SAF_CIERRE_CMR_REP D ON A.ID_EXTERNO = D.ID_EXTERNO
		WHERE 
			A.NUM_CREDITO IN
			(
			SELECT MAX(NUM_CREDITO) FROM BRONZE.SAF.PR_CREDITOS  -- SE OBTIENE UN SOLO CRÉDITO POR CLIENTE PARA EVITAR DUPLICIDAD
			WHERE 
			(IND_ESTADO = 'D' OR 
					(IND_ESTADO = 'C' 
					AND (FEC_CANCELACION >= @FechaIni
					AND FEC_CANCELACION < @FechaFin))
				) 
			GROUP BY COD_CLIENTE)
			AND A.IND_LINEA = 'N'
			AND A.NUM_DESC_TIPO_CREDITO IN (12,13,14,15,28,2,5,18,27) -- CREDITOS COMERCIALES
			AND A.TIP_CREDITO NOT IN (16, 26) -- TIPOS DE CREDITO DESCARTADOS 
			--AND C.VentasNetasTotalesAnuales < (14000000 * (SELECT VAL_UDIS FROM BRONZE.SAF.CL_UDIS WHERE COD_EMPRESA= '001' AND FEC_UDIS = @FechaSistema)) --VENTAS
		GROUP BY A.COD_CLIENTE, C.VentasNetasTotalesAnuales, C.FechaEstadosFinancieros, C.TipoDeGarante

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log
	
	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[020_ENT_SWAP_MEREC_CONVO]    Script Date: 01/06/2026 20:11:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[020_ENT_SWAP_MEREC_CONVO] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[020_ENT_SWAP_MEREC_CONVO]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[020_ENT_SWAP_MEREC_CONVO]
				WHERE [FE_CORTE] >= @FechaIni
					AND [FE_CORTE] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[020_ENT_SWAP_MEREC_CONVO]
			WHERE [FE_CORTE] >= @FechaIni
				AND [FE_CORTE] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [SILVER].[RR].[020_ENT_SWAP_MEREC_CONVO] (
			[PIZARRA],
			[CONTRAT_VIG],
			[ACT_OPE],
			[MON_ACT_OPE],
			[PAS_OPE],
			[MON_PAS_OPE],
			[DUR_ACT],
			[DUR_PAS],
			[ID_GAR],
			[NUM_ID],
			[PRO_TER],
			[CLIENTE],
			[FE_CON_OPE],
			[POS_OPER],
			[OBJ_OPE],
			[SOCIO_LIQ],
			[FE_CORTE],
			[ACT_OPE_VAL],
			[PAS_OPE_VAL]
			)
		SELECT [PIZARRA],
			[CONTRAT_VIG],
			[ACT_OPE],
			[MON_ACT_OPE],
			[PAS_OPE],
			[MON_PAS_OPE],
			[DUR_ACT],
			[DUR_PAS],
			[ID_GAR],
			[NUM_ID],
			[PRO_TER],
			[CLIENTE],
			[FE_CON_OPE],
			[POS_OPER],
			[OBJ_OPE],
			[SOCIO_LIQ],
			[FE_CORTE],
			[ACT_OPE_VAL],
			[PAS_OPE_VAL]
		FROM [SILVER].[RR].[020_ENT_SWAP_MEREC_CONVO] --Debe cambiar por la consulta de extraccion
		WHERE [FE_CORTE] >= @FechaIni
			AND [FE_CORTE] < @FechaFin

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[025_ENT_HIPOTECARIO]    Script Date: 01/06/2026 20:11:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[025_ENT_HIPOTECARIO] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @INPC2 DECIMAL(18,6);
	DECLARE @NombreJob NVARCHAR(128) = '[025_ENT_HIPOTECARIO]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) 
		SET @INPC2 = (SELECT  VALOR FROM BRONZE.lmda.INPC_Historico 
			WHERE YEAR(FechaIndicador) = YEAR(@FechaSistema) AND MONTH(FechaIndicador) = MONTH(@FechaSistema))--------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[025_ENT_HIPOTECARIO]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[025_ENT_HIPOTECARIO]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[025_ENT_HIPOTECARIO] (
			[ID_PERSONA],
			[ID_CREDITO_H],
			[DESTINOCREDITO],
			[CLASIFICACIONCONTABLE_H],
			[TIPO_BAJA_H],
			[TIPO_BAJA_OTRA_H],
			[MONTO_CONDONACION],
			[MONTO_QUITAS],
			[MONTO_CASTIGOS],
			[MONTO_BONIFICACION],
			[MONTO_DESCUENTO],
			[MONTO_DACION],
			[PRODUCTO_HIPOTECARIO],
			[MONTO_CRED_AUT],
			[FECHA_OTORG_CRED],
			[FECHA_VENC_CRED],
			[PROG_SUBSIDIO_FRENTE],
			[MONTO_SUBSIDIO_FRENTE],
			[ENTIDADCOFINANCIADORA],
			[MONTOSUBCUENTA],
			[MONTOCOFINANCIADOR],
			[TIPO_ALTA_H],
			[MOTIVO_REESTRUCTURA],
			[PORTABILIDAD],
			[ENTIDAD_PORTABILIDAD],
			[MONTO_FONDEADO_BD_H],
			[INSTITUTO_FONDEA_H],
			[MONTO_ORIGINAL],
			[VALOR_VIVIENDA_ORIG],
			[MONEDA_CREDITO],
			[TIPO_TASA_CRED],
			[TASA_REF_HIP],
			[DIFER_TASA_HIP],
			[OPER_DIF_TASA_HIP],
			[PERIODICIDADCAPITAL_HIP],
			[PERIODICIDADINTERES_HIP],
			[PERIODO_FACTURACION],
			[SEGUROS_ACREDITADO],
			[COMISIONES_OTORG],
			[CUENTA_COACREDITADO],
			[NOMBRE_COACREDITADO],
			[RFC_COACREDITADO],
			[CURP_COACREDITADO],
			[GENERO_COACREDITADO],
			[EDAD_COACREDITADO],
			[ESTADO_CIVIL_COACREDITADO],
			[LOCALIDAD_COACREDITADO],
			[NACIONALIDAD_COACREDITADO],
			[APORTA_SUBCUENTA_COACREDITADO],
			[INGRESOS_MENSUALES_COACREDITADO],
			[COMPROBACION_INGRESOS_COACREDITADO],
			[SECTOR_LABORAL_COACREDITADO],
			[CONSULTA_SIC_COACREDITADO],
			[VALOR_AVALUO],
			[NUMERO_AVALUO],
			[LOCALIDAD_VIVIENDA],
			[CLAVE_CUV],
			[TIPO_SEGURO],
			[ENTIDAD_SEGURO],
			[PORCENTAJE_ASEGURADO],
			[SUBCUENTA_VIV],
			[VALOR_GARANTIAS],
			[SALDO_CAPITAL_VIGENTE],
			[SALDO_CAPITAL_VENCIDO],
			[SALDO_INTERES_VIGENTE],
			[SALDO_INTERES_VENCIDO],
			[TASA_INTERES_ORDINARIA],
			[TASA_INTERES_MORATORIA],
			[MONTO_PAGO_EXIGIBLE],
			[MONTO_CAPITAL_PAGADO],
			[MONTO_INTERES_ORDINARIO_PAGADO],
			[COMISIONES_PAGADAS],
			[MONTO_INTERES_MORATORIO],
			[OTROS_MONTOS_PAGADOS],
			[DIAS_CALCULO_INTERES],
			[ETAPA_DETERIORO],
			[FECHA_ULTIMO_PAGO_EX_REC],
			[SDAÑOS],
			[DIAS_ATRASO],
			[ATR],
			[MAXATR_4],
			[MAXATR_7],
			[TIPO_MET_RESERVAS],
			[PROM_PAGO_4],
			[PROM_PAGO_7],
			[CLTV],
			[MESES_DESDE_ULT_ATR_BK],
			[VECES],
			[PROMEDIORET],
			[SDESI],
			[SVIDA],
			[QX],
			[CURAS],
			[FACTOR_A],
			[FACTOR_B],
			[FACTOR_C],
			[REGION_VIVIENDA],
			[CONVENIO],
			[VALORVIVIENDA],
			[FACTOR_VIVIENDA],
			[TIPO_ACTUALIZACION],
			[NUMERO_REAVALUO],
			[TASA_INTERES_EFECTIVA],
			[HIPOTESIS_PRESUNCION],
			[RESERVAS_ADICIONALES],
			[FECHA_INFO],
			[CODIGOCREDITOREESTRUCTURA],
			[TIPOGARANTIA],
			[HITSIC],
			[SUMA_PROGRAMADOS],
			[INTERES_FUERA_BALANCE],
			[SALDO_BASE_INTERES]
			)
		SELECT --TOP 2
			CR.COD_CLIENTE  AS ID_PERSONA,
			CR.ID_EXTERNO  AS ID_CREDITO_H,
			CASE
				WHEN CR.COD_PLAN_INVERSION IN (4,5) THEN CAST(CR.COD_PLAN_INVERSION AS DECIMAL(2,0))
				WHEN CR.TIP_CREDITO IN (2,3,11,12,13,14) THEN 8 -- LIQUIDEZ
				ELSE 0
			END AS DESTINOCREDITO,
			ISNULL(CAST(CATCC.CLAVE AS DECIMAL(12,0)),0) AS CLASIFICACIONCONTABLE_H, 
			ISNULL(CAST(INF.TIP_BAJA AS DECIMAL(3,0)),0) AS TIPO_BAJA_H,
			0 AS TIPO_BAJA_OTRA_H,
			ISNULL(CAST(IND.MONTO_CONDONACION AS DECIMAL(21,2)),0) AS MONTO_CONDONACION,
			ISNULL(CAST(IND.MONTO_QUITAS AS DECIMAL(21,2)),0) AS MONTO_QUITAS,
			ISNULL(CAST(IND.MONTO_CASTIGOS AS DECIMAL(21,2)),0) AS MONTO_CASTIGOS,
			ISNULL(CAST(IND.MONTO_BONIFICACION AS DECIMAL(21,2)),0) AS MONTO_BONIFICACION,
			ISNULL(CAST(IND.MONTO_DESCUENTO AS DECIMAL(21,2)),0) AS MONTO_DESCUENTO,
			ISNULL(CASE
				WHEN CR.IND_ESTADO = 'C' THEN 
					CAST((SELECT SUM(B.MON_PAGO)
					FROM BRONZE.SAF.PR_ENCABEZADO_PAGO A
					INNER JOIN BRONZE.SAF.PR_DETALLE_PAGO B ON A.ID_SECUENCIA = B.ID_SECUENCIA
					INNER JOIN BRONZE.SAF.PR_CREDITOS C ON A.COD_EMPRESA = C.COD_EMPRESA AND A.NUM_CREDITO = C.NUM_CREDITO
					WHERE 
					CAST(A.NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT)
					AND A.FEC_MOVIMIENTO = CR.FEC_CANCELACION 
					and A.mon_movimiento > 0 -- no se consideran cancelaciones
					AND A.Id_cuenta IN ('2073')
					GROUP BY A.NUM_CREDITO)AS DECIMAL(21,2))
				ELSE 0
			END,0) AS MONTO_DACION,
			CASE 
				WHEN DC.TIPO_CREDITO = 'H. GARANTIZADA' THEN '68059AMX'
				WHEN DC.TIPO_CREDITO = 'ADQUISICION' THEN '68059AMX'
				WHEN DC.TIPO_CREDITO = 'COFINAVIT' THEN '68059CMX'
				WHEN DC.TIPO_CREDITO = 'LIQUIDEZ' THEN '68059LMX'
				ELSE ''
			END AS PRODUCTO_HIPOTECARIO,
			CAST(CR.MON_CREDITO AS DECIMAL(21,2)) AS MONTO_CRED_AUT,
			CR.FEC_APERTURA AS FECHA_OTORG_CRED,
			CR.FEC_VENCIMIENTO AS FECHA_VENC_CRED,
			'0' AS PROG_SUBSIDIO_FRENTE, --valor fijo de catalogo
			0 AS MONTO_SUBSIDIO_FRENTE,
			LEFT(ISNULL(CAST(DC.ENTIDAD_COF AS VARCHAR(6)),''),6)  AS ENTIDADCOFINANCIADORA,
			ISNULL(CAST(DC.MONTO_SUBCUENTA AS DECIMAL(21,2)),0) AS MONTOSUBCUENTA,
			ISNULL(CAST(DC.MONTO_COFINANCIAMIENTO AS DECIMAL(21,2)),0) AS MONTOCOFINANCIADOR,
			ISNULL(CAST(INF.TIP_ALTA AS DECIMAL(3,0)),0) AS TIPO_ALTA_H,
			CASE
				WHEN CR.IND_ESTADO_CRED = 'E' THEN 31 -- VALOR FIJO DE CATALOGO
				ELSE 0
			END AS MOTIVO_REESTRUCTURA,
			2 AS PORTABILIDAD, --VALOR FIJO DE CATALOGO
			'0' AS ENTIDAD_PORTABILIDAD, --VALOR FIJO DE CATALOGO
			ISNULL(CAST(TCV.MON_PRINCIPAL_FONDEADOR AS DECIMAL(21,2)),0) AS MONTO_FONDEADO_BD_H,
			ISNULL(CAST(CR.COD_ENTE_FONDEO AS VARCHAR(6)),'')  AS INSTITUTO_FONDEA_H,
			CAST(CR.MON_CREDITO AS DECIMAL(21,2)) AS MONTO_ORIGINAL,
			ISNULL(CAST(G.MON_AVALUO AS DECIMAL(21,2)),0) AS VALOR_VIVIENDA_ORIG,
			14 AS MONEDA_CREDITO, --VALOR FIJO DE CATALOGO
			CASE
				WHEN DC.TIPO_TASA = 'Fija' THEN 1
				WHEN DC.TIPO_TASA = 'Variable' THEN 2
				WHEN DC.TIPO_TASA = 'Mixta' THEN 3
				ELSE 0
			END AS TIPO_TASA_CRED, --VALOR FIJO DE CATALOGO
			600 AS TASA_REF_HIP, -- VALOR FIJO DE CATALOGO
			0 AS DIFER_TASA_HIP,
			110 AS OPER_DIF_TASA_HIP, -- VALOR FIJO CATALOGO
			5 AS PERIODICIDADCAPITAL_HIP, -- VALOR FIJO CATALOGO
			5 AS PERIODICIDADINTERES_HIP, -- VALOR FIJO CATALOGO
			50 AS PERIODO_FACTURACION, -- VALOR FIJO CATALOGO
			CASE
				WHEN EXISTS (SELECT COD_RUBRO FROM BRONZE.SAF.PR_RUBRO_X_CREDITO WHERE CAST(NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT) AND COD_RUBRO = 'SEGURO_VIDA')
				AND EXISTS (SELECT COD_RUBRO FROM BRONZE.SAF.PR_RUBRO_X_CREDITO WHERE CAST(NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT) AND COD_RUBRO = 'SEGURO_DANOS') 
				THEN 15 -- TIENE LOS DOS SEGUROS
				WHEN EXISTS (SELECT COD_RUBRO FROM BRONZE.SAF.PR_RUBRO_X_CREDITO WHERE CAST(NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT) AND COD_RUBRO = 'SEGURO_VIDA')
				AND NOT EXISTS (SELECT COD_RUBRO FROM BRONZE.SAF.PR_RUBRO_X_CREDITO WHERE CAST(NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT) AND COD_RUBRO = 'SEGURO_DANOS') 
				THEN 11
				WHEN NOT EXISTS (SELECT COD_RUBRO FROM BRONZE.SAF.PR_RUBRO_X_CREDITO WHERE CAST(NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT) AND COD_RUBRO = 'SEGURO_VIDA')
				AND EXISTS (SELECT COD_RUBRO FROM BRONZE.SAF.PR_RUBRO_X_CREDITO WHERE CAST(NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT) AND COD_RUBRO = 'SEGURO_DANOS') 
				THEN 13
				ELSE 2
			END AS SEGUROS_ACREDITADO,
			ISNULL(CAST((
					SELECT SUM(SAL_RUBRO)
					FROM BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO
					WHERE IND_ESTADO = 'P'
					  AND FEC_COBRO >= @FechaIni AND FEC_COBRO < @FechaFin
					  AND CAST(NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT)
					  AND COD_RUBRO IN ('COMISION_ADMINISTRACION','MORATORIOS','COMISION_ANTICIPADA')
					GROUP BY NUM_CREDITO
				)AS DECIMAL(21,2)), 0) AS COMISIONES_OTORG,
			CASE 
				WHEN DR.PARTICIPACION = 'COACREDITADO' THEN 1
				ELSE 2
			END AS CUENTA_COACREDITADO,
			CASE
				WHEN DR.PARTICIPACION = 'COACREDITADO' THEN DR.NOMBRE
				ELSE ''
			END AS NOMBRE_COACREDITADO,
			CASE
				WHEN DR.PARTICIPACION  = 'COACREDITADO' THEN DR.RFC
				ELSE ''
			END	AS RFC_COACREDITADO,
			CASE
				WHEN DR.PARTICIPACION  = 'COACREDITADO' THEN DR.CURP
				ELSE ''
			END	AS CURP_COACREDITADO,
			CASE
				WHEN DR.PARTICIPACION = 'COACREDITADO' THEN
			CASE
				WHEN DR.SEXO = 'FEMENINO' THEN 1
				WHEN DR.SEXO = 'MASCULINO' THEN 2
				ELSE 0
			END
				ELSE 0
			END AS GENERO_COACREDITADO,
			CASE 
				WHEN DR.PARTICIPACION = 'COACREDITADO' THEN
				 DATEDIFF(year, DR.FECHA_DE_NACIMIENTO, GETDATE())-
			CASE
				WHEN (MONTH(GETDATE()) < MONTH (DR.FECHA_DE_NACIMIENTO)) OR
					 (MONTH(GETDATE()) = MONTH (DR.FECHA_DE_NACIMIENTO) AND 
					 DAY(GETDATE()) < DAY(DR.FECHA_DE_NACIMIENTO)) THEN 1
				ELSE 0
			END 
				ELSE 0
			END AS EDAD_COACREDITADO,
			CASE 
				WHEN DR.PARTICIPACION = 'COACREDITADO' THEN
			CASE
				WHEN DR.ESTADO_CIVIL = 'SOLTERO' THEN 1
				WHEN DR.ESTADO_CIVIL = 'CASADO' THEN 2
				ELSE 0
			END
				ELSE 0
			END AS ESTADO_CIVIL_COACREDITADO,
			CASE 
				WHEN DR.PARTICIPACION = 'COACREDITADO' THEN 
				(SELECT TOP(1) LEFT(CLAVE_DE_LA_LOCALIDAD_CNBV,5) FROM BRONZE.RR.LOCALIDAD 
				WHERE DESC_DEL_PAIS COLLATE SQL_Latin1_General_CP1_CI_AI = DR.PAIS
				AND DESC_DEL_ESTADO COLLATE SQL_Latin1_General_CP1_CI_AI = DR.ESTADO
				AND DESC_DEL_MUNICIPIO COLLATE SQL_Latin1_General_CP1_CI_AI = DR.DELEGACION_MUNICIPIO 
				ORDER BY CLAVE_DE_LA_LOCALIDAD ASC)
				ELSE 0
			END AS LOCALIDAD_COACREDITADO,
			CASE
				WHEN DR.PARTICIPACION = 'COACREDITADO' 
				THEN 
				(SELECT CLAVE FROM BRONZE.RR.NACIONALIDADES WHERE DESCRIPCION_LARGA_SUPREPORTE = DR.NACIONALIDAD)
				ELSE 0
			END AS NACIONALIDAD_COACREDITADO,
			CASE
				WHEN DR.PARTICIPACION = 'COACREDITADO' THEN
			CASE
				WHEN DC.MONTO_SUBCUENTA <> 0 THEN 1
				WHEN DC.MONTO_SUBCUENTA = 0 THEN 2
				ELSE 0
			END
				ELSE 0
			END AS APORTA_SUBCUENTA_COACREDITADO,
			CASE 
				WHEN DR.PARTICIPACION = 'COACREDITADO' THEN CAST(DR.INGRESOS_MENSUALES_BRUTOS AS DECIMAL(21,2)) 
				ELSE 0
			END AS INGRESOS_MENSUALES_COACREDITADO,
			CASE
				WHEN DR.PARTICIPACION = 'COACREDITADO' THEN
			CASE
				WHEN DR.COMPROBACION_INGRESOS = 'NO ASALARIADO' THEN 1
				WHEN DR.COMPROBACION_INGRESOS = 'ASALARIADO' THEN 2
				ELSE 0
			END
				ELSE 0
			END AS COMPROBACION_INGRESOS_COACREDITADO,
			CASE
				WHEN DR.PARTICIPACION = 'COACREDITADO' THEN
			CASE
				WHEN DR.SECTOR_LABORAL = 'SECTOR PUBLICO ESTATAL' OR DR.SECTOR_LABORAL = 'Sector público estatal' THEN 1
				WHEN DR.SECTOR_LABORAL = 'SECTOR PUBLICO MUNICIPAL' THEN 2
				WHEN DR.SECTOR_LABORAL = 'SECTOR PUBLICO FEDERAL' OR DR.SECTOR_LABORAL = 'Sector público federal' 
				OR DR.SECTOR_LABORAL = 'PÚBLICO' OR DR.SECTOR_LABORAL = 'PUBLICO' THEN 3
				WHEN DR.SECTOR_LABORAL = 'SECTOR PRIVADO' OR DR.SECTOR_LABORAL = 'PRIVADO' OR DR.SECTOR_LABORAL = 'ASALARIADO'  THEN 4
				WHEN DR.SECTOR_LABORAL = 'NO ASALARIADO' OR DR.SECTOR_LABORAL = 'SECTOR INFORMAL' THEN 5
				ELSE 0
			END
				ELSE 0
			END AS SECTOR_LABORAL_COACREDITADO,
			CASE
				WHEN DR.PARTICIPACION = 'COACREDITADO' THEN CAST(DR.NUM_CONSULTA AS VARCHAR(16))
				ELSE ''
			END AS CONSULTA_SIC_COACREDITADO,
			ISNULL(CAST(G.MON_AVALUO AS DECIMAL(21,2)),0) AS VALOR_AVALUO,
			LEFT(ISNULL(CAST(G.NUM_AVALUO AS VARCHAR(17)),''),17) AS NUMERO_AVALUO,
			CONCAT('484',
				IIF(GH.COD_PAIS IS NULL, '00', GH.COD_PAIS),
				IIF(GH.COD_PROVINCIA IS NULL, '000', GH.COD_PROVINCIA),
				'0001') AS LOCALIDAD_VIVIENDA,
			CAST(DC.CUV AS VARCHAR(16)) AS CLAVE_CUV,
			CASE
				WHEN DC.TIPO_SEGURO_CRED = 'Sin Seguro de Crédito a la Vivienda' THEN 0
				WHEN DC.TIPO_SEGURO_CRED = 'Primeras Pérdidas' THEN 1
				WHEN DC.TIPO_SEGURO_CRED = 'Pari Passu' THEN 2
				WHEN DC.TIPO_SEGURO_CRED = 'Seguro de Crédito a la Vivienda' THEN 3
				ELSE 0
			END AS TIPO_SEGURO,
			0 AS ENTIDAD_SEGURO,
			0 AS PORCENTAJE_ASEGURADO,
			CASE 
				WHEN DR.PARTICIPACION = 'COACREDITADO' THEN 
			CASE
				WHEN DC.MONTO_SUBCUENTA <> 0 THEN CAST(DC.MONTO_SUBCUENTA AS DECIMAL(21,2))
				ELSE 0
			END
				ELSE 0
			END AS SUBCUENTA_VIV,
			0 AS VALOR_GARANTIAS, -- valor fijo de operaciones
			ISNULL(CAST(IND.SALDO_CAPITAL_VIGENTE AS DECIMAL(21,2)),0) AS SALDO_CAPITAL_VIGENTE,
			ISNULL(CAST(IND.SALDO_CAPITAL_VENCIDO AS DECIMAL(21,2)),0) AS SALDO_CAPITAL_VENCIDO,
			ISNULL(CAST(IND.SALDO_INTERES_VIGENTE AS DECIMAL(21,2)),0) AS SALDO_INTERES_VIGENTE,
			ISNULL(CAST(IND.SALDO_INTERES_VENCIDO AS DECIMAL(21,2)),0) AS SALDO_INTERES_VENCIDO,
			CAST(CR.TASA_INTERES AS DECIMAL(10,6)) AS TASA_INTERES_ORDINARIA,
			CAST(CR.TASA_MORA AS DECIMAL(10,6)) AS TASA_INTERES_MORATORIA,

			ISNULL(CAST((SELECT SUM(B.MON_PAGO)
				FROM BRONZE.SAF.PR_ENCABEZADO_PAGO A 
				INNER JOIN BRONZE.SAF.PR_DETALLE_PAGO B ON A.ID_SECUENCIA = B.ID_SECUENCIA
				LEFT JOIN BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO Q WITH (NOLOCK) ON Q.Num_Secuencia = B.Id_referencia
										AND Q.Num_Credito = A.Num_credito AND Q.Num_Det_Secuencia = B.Num_det_secuencia AND Q.Cod_Rubro = B.Cod_concepto AND Q.Ind_Estado != 'N'
				WHERE 
				ID_CUENTA NOT IN ('140103009001', '140103009002', '140103009003','1401030090010000','1401030090020000','5050260901000300','2073','2074','2068','2061','2056')
				AND NOT (TIP_TRANSACCION = 4 AND SUBTIP_TRANSAC = 5)
				AND FEC_MOVIMIENTO >= @FechaIni AND FEC_MOVIMIENTO < @FechaFin
				AND FEC_COBRO >= @FechaIni AND FEC_COBRO < @FechaFin
				AND CAST(A.NUM_CREDITO AS INT) = CAST(CR.NUM_CREDITO AS INT) 
				GROUP BY A.NUM_CREDITO)AS DECIMAL(21,2)),0) AS MONTO_PAGO_EXIGIBLE,

			ISNULL(CAST((
					SELECT SUM(B.MON_PAGO)
					FROM BRONZE.SAF.PR_ENCABEZADO_PAGO A
					INNER JOIN BRONZE.SAF.PR_DETALLE_PAGO B
						ON A.ID_SECUENCIA = B.ID_SECUENCIA
					WHERE 
						CAST(NUM_CREDITO AS INT) =  CAST(CR.NUM_CREDITO AS INT)
						AND A.ID_CUENTA NOT IN ('140103009001', '140103009002', '140103009003',
							'1401030090010000', '1401030090020000',
							'5050260901000300', '2073','2074','2068','2061','2056')
						AND NOT (TIP_TRANSACCION = 4 AND SUBTIP_TRANSAC = 5)
						AND COD_CONCEPTO = 'PRINCIPAL'
						AND FEC_MOVIMIENTO >= @FechaIni
						AND FEC_MOVIMIENTO < @FechaFin
					GROUP BY A.NUM_CREDITO
				)AS DECIMAL(21,2)), 0) AS MONTO_CAPITAL_PAGADO,
			ISNULL(CAST(CR.MON_PAGADO_INTERESES AS DECIMAL(21,2)),0) AS MONTO_INTERES_ORDINARIO_PAGADO,
			ISNULL(CAST(EC.MON_COMISION_COBRADA AS DECIMAL(21,2)),0) AS COMISIONES_PAGADAS,
			ISNULL(CAST(IND.MORA AS DECIMAL(21,2)),0) AS MONTO_INTERES_MORATORIO,
			ISNULL(CAST((
					SELECT SUM(B.MON_PAGO)
					FROM BRONZE.SAF.PR_ENCABEZADO_PAGO A
					INNER JOIN BRONZE.SAF.PR_DETALLE_PAGO B
						ON A.ID_SECUENCIA = B.ID_SECUENCIA
					WHERE 
						CAST(NUM_CREDITO AS INT) =  CAST(CR.NUM_CREDITO AS INT)
						AND A.ID_CUENTA NOT IN ('140103009001', '140103009002', '140103009003',
							'1401030090010000', '1401030090020000',
							'5050260901000300', '2073','2074','2068','2061','2056')
						AND NOT (TIP_TRANSACCION = 4 AND SUBTIP_TRANSAC = 5)
						AND COD_CONCEPTO IN ('SEGURO_DANOS','SEGURO_VIDA')
						AND FEC_MOVIMIENTO >= @FechaIni
						AND FEC_MOVIMIENTO < @FechaFin
					GROUP BY A.NUM_CREDITO
				)AS DECIMAL(21,2)), 0) AS OTROS_MONTOS_PAGADOS,

			ISNULL(CAST((SELECT TOP 1 DIAS_CALCULO FROM BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO WHERE
					CAST(NUM_CREDITO AS INT)= CAST(CR.NUM_CREDITO AS INT) AND FEC_COBRO >= @FechaIni AND FEC_COBRO < @FechaFin
					AND COD_RUBRO = 'INTERESES')AS DECIMAL(4,0)),0) AS DIAS_CALCULO_INTERES,

			CAST(CR.IND_ETAPA AS DECIMAL(1,0)) AS ETAPA_DETERIORO,
			ISNULL((
					SELECT /*FORMAT(*/MAX(FEC_MOVIMIENTO)/*, 'yyyy-MM')*/
					FROM BRONZE.SAF.PR_ENCABEZADO_PAGO A
					INNER JOIN BRONZE.SAF.PR_DETALLE_PAGO B ON A.ID_SECUENCIA = B.ID_SECUENCIA
					WHERE 
						CAST(NUM_CREDITO AS INT) =  CAST(CR.NUM_CREDITO AS INT)
						AND A.ID_CUENTA NOT IN ('140103009001', '140103009002', '140103009003',
							'1401030090010000', '1401030090020000',
							'5050260901000300', '2073','2074','2068','2061','2056')
						AND NOT (TIP_TRANSACCION = 4 AND SUBTIP_TRANSAC = 5)
						AND FEC_MOVIMIENTO >= @FechaIni AND FEC_MOVIMIENTO < @FechaFin
					GROUP BY A.NUM_CREDITO
				), /*FORMAT(*/CAST('' AS DATE)/*,'yyyy-MM')*/)AS FECHA_ULTIMO_PAGO_EX_REC,
			0 AS SDAÑOS, -- valor fijo , corregir campo en tabla

			ISNULL(CAST((SELECT DATEDIFF(DAY,MIN(FEC_COBRO),@FechaFin) -1 FROM BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO
					WHERE IND_ESTADO= 'P' AND IND_POSPOSICION = 'N' AND FEC_COBRO >= @FechaIni AND FEC_COBRO < @FechaFin
					AND CAST(NUM_CREDITO AS INT)= CAST(CR.NUM_CREDITO AS INT)
					GROUP BY NUM_CREDITO)AS DECIMAL(8,0)) ,0)AS DIAS_ATRASO,
			CASE
				WHEN CR.IND_ESTADO = 'C' THEN CAST((SELECT COUNT(DISTINCT NUM_SECUENCIA) FROM BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO
					WHERE CAST(NUM_CREDITO AS INT)= CAST(CR.NUM_CREDITO AS INT) AND IND_ESTADO = 'P' AND FEC_COBRO < @FechaSistema)AS DECIMAL(11,0))
				WHEN CR.IND_ESTADO = 'D' THEN CAST((SELECT COUNT(DISTINCT NUM_SECUENCIA) FROM BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO
					WHERE CAST(NUM_CREDITO AS INT)= CAST(CR.NUM_CREDITO AS INT) AND IND_ESTADO = 'P' AND FEC_COBRO >= DATEADD(MONTH,7,@FechaSistema) AND FEC_COBRO <= @FechaSistema )AS DECIMAL(11,0))
			END AS ATR,
			CAST((SELECT COUNT(DISTINCT NUM_SECUENCIA) FROM BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO
					WHERE CAST(NUM_CREDITO AS INT)= CAST(CR.NUM_CREDITO AS INT) AND IND_ESTADO = 'P' AND FEC_COBRO >= DATEADD(MONTH,4,@FechaSistema) AND FEC_COBRO <= @FechaSistema ) AS DECIMAL(8,0)) AS MAXATR_4,
			CAST((SELECT COUNT(DISTINCT NUM_SECUENCIA) FROM BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO
					WHERE CAST(NUM_CREDITO AS INT)= CAST(CR.NUM_CREDITO AS INT) AND IND_ESTADO = 'P' AND FEC_COBRO >= DATEADD(MONTH,7,@FechaSistema) AND FEC_COBRO <= @FechaSistema ) AS DECIMAL(8,0)) AS MAXATR_7,
			3 AS TIPO_MET_RESERVAS, -- VALOR FIJO DE CATALOGO
			ISNULL(CAST((SELECT ISNULL(SUM(B.MON_PAGO)/4,0)
			FROM BRONZE.SAF.PR_ENCABEZADO_PAGO A 
			INNER JOIN BRONZE.SAF.PR_DETALLE_PAGO B ON A.ID_SECUENCIA = B.ID_SECUENCIA
			LEFT JOIN BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO Q WITH (NOLOCK) ON Q.Num_Secuencia = B.Id_referencia
									AND Q.Num_Credito = A.Num_credito AND Q.Num_Det_Secuencia = B.Num_det_secuencia AND Q.Cod_Rubro = B.Cod_concepto AND Q.Ind_Estado != 'N'
			WHERE 
			ID_CUENTA NOT IN ('140103009001', '140103009002', '140103009003','1401030090010000','1401030090020000','5050260901000300','2073','2074','2068','2061','2056')
			AND NOT (TIP_TRANSACCION = 4 AND SUBTIP_TRANSAC = 5)
			AND FEC_MOVIMIENTO >= @FechaIni AND FEC_MOVIMIENTO < @FechaFin
			AND FEC_COBRO >= DATEADD(MONTH,4,@FechaSistema) AND FEC_COBRO < @FechaSistema
			AND CAST(A.NUM_CREDITO AS INT)= CAST(CR.NUM_CREDITO AS INT)
			GROUP BY A.NUM_CREDITO)AS DECIMAL(10,6)),0)  AS PROM_PAGO_4,

			ISNULL(CAST((SELECT ISNULL(SUM(B.MON_PAGO)/7,0)
			FROM BRONZE.SAF.PR_ENCABEZADO_PAGO A 
			INNER JOIN BRONZE.SAF.PR_DETALLE_PAGO B ON A.ID_SECUENCIA = B.ID_SECUENCIA
			LEFT JOIN BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO Q WITH (NOLOCK) ON Q.Num_Secuencia = B.Id_referencia
									AND Q.Num_Credito = A.Num_credito AND Q.Num_Det_Secuencia = B.Num_det_secuencia AND Q.Cod_Rubro = B.Cod_concepto AND Q.Ind_Estado != 'N'
			WHERE 
			ID_CUENTA NOT IN ('140103009001', '140103009002', '140103009003','1401030090010000','1401030090020000','5050260901000300','2073','2074','2068','2061','2056')
			AND NOT (TIP_TRANSACCION = 4 AND SUBTIP_TRANSAC = 5)
			AND FEC_MOVIMIENTO >= @FechaIni AND FEC_MOVIMIENTO < @FechaFin
			AND FEC_COBRO >= DATEADD(MONTH,7,@FechaSistema) AND FEC_COBRO < @FechaSistema
			AND CAST(A.NUM_CREDITO AS INT)= CAST(CR.NUM_CREDITO AS INT)
			GROUP BY A.NUM_CREDITO)AS DECIMAL(10,6)),0)  AS PROM_PAGO_7,

			ISNULL (100 * CAST ((IND.RESPONSABILIDAD) / 
			(( @INPC2 / (SELECT VALOR FROM BRONZE.lmda.INPC_Historico 
			WHERE YEAR(FechaIndicador) = YEAR(CR.FEC_APERTURA) AND MONTH(FechaIndicador) = MONTH(CR.FEC_APERTURA))) 
			* (G.MON_AVALUO)) AS DECIMAL(10,6) ),0) AS CLTV,
			0 AS MESES_DESDE_ULT_ATR_BK,
			CAST(CR.MON_CREDITO/ISNULL((SELECT MAX(NUM_SECUENCIA) 
			FROM BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO WHERE CAST(NUM_CREDITO AS INT)= CAST(CR.NUM_CREDITO AS INT)),1 )AS DECIMAL(8,0))AS VECES,
			0 AS PROMEDIORET, -- VALOR FIJO DE OPERACIONES
			0 AS SDESI, -- VALOR FIJO DE OPERACIONES

			ISNULL(CAST((
					SELECT SUM(B.MON_PAGO)
					FROM BRONZE.SAF.PR_ENCABEZADO_PAGO A
					INNER JOIN BRONZE.SAF.PR_DETALLE_PAGO B
						ON A.ID_SECUENCIA = B.ID_SECUENCIA
					WHERE 
						CAST(NUM_CREDITO AS INT) = CR.NUM_CREDITO
						AND A.ID_CUENTA NOT IN ('140103009001', '140103009002', '140103009003',
							'1401030090010000', '1401030090020000',
							'5050260901000300', '2073','2074','2068','2061','2056')
						AND NOT (TIP_TRANSACCION = 4 AND SUBTIP_TRANSAC = 5)
						AND COD_CONCEPTO = 'SEGURO_VIDA'
						AND FEC_MOVIMIENTO >= @FechaIni
						AND FEC_MOVIMIENTO < @FechaFin
					GROUP BY A.NUM_CREDITO
				)AS DECIMAL(21,2)), 0)  AS SVIDA,
   
		   CASE
				WHEN PF.IND_SEXO = 'M' THEN CAST((SELECT QX_HOMBRE FROM BRONZE.RR.TABLA_QX WHERE 
				EDAD = DATEDIFF(YEAR, PF.FEC_NACIMIENTO, @FechaSistema) - 
				IIF(MONTH(PF.FEC_NACIMIENTO) > MONTH(@FechaSistema) OR MONTH(PF.FEC_NACIMIENTO) = MONTH(@FechaSistema) AND DAY(PF.FEC_NACIMIENTO) > DAY(@FechaSistema),1,0)) AS DECIMAL(8,4))
		
				WHEN PF.IND_SEXO = 'F' THEN CAST((SELECT QX_MUJER FROM BRONZE.RR.TABLA_QX WHERE 
				EDAD = (DATEDIFF(YEAR, PF.FEC_NACIMIENTO, @FechaSistema) - 
				IIF(MONTH(PF.FEC_NACIMIENTO) > MONTH(@FechaSistema) OR MONTH(PF.FEC_NACIMIENTO) = MONTH(@FechaSistema) AND DAY(PF.FEC_NACIMIENTO) > DAY(@FechaSistema),1,0))) AS DECIMAL(8,4))
				ELSE 0
			END AS QX,
			CASE
				WHEN IND.PV >=  0 AND IND.PV < 10	THEN 27--.50
				WHEN IND.PV >= 10 AND IND.PV < 16	THEN 16--.20
				WHEN IND.PV >= 16 AND IND.PV < 22	THEN 8--.60
				WHEN IND.PV >= 22 AND IND.PV < 28	THEN 3--.50
				ELSE 0--.00
			END AS CURAS,
			CASE 
				WHEN IND.PV >= 0  AND IND.PV < 4  THEN 0.4235
				WHEN IND.PV >= 4  AND IND.PV < 10 THEN 0.3933
				WHEN IND.PV >= 10 AND IND.PV < 16 THEN 0.3617
				WHEN IND.PV >= 16 AND IND.PV < 22 THEN 0.3317
				WHEN IND.PV >= 22 AND IND.PV < 28 THEN 0.3020
				WHEN IND.PV >= 28 AND IND.PV < 34 THEN 0.2751
				WHEN IND.PV >= 34 AND IND.PV < 40 THEN 0.2523
				WHEN IND.PV >= 40 AND IND.PV < 46 THEN 0.2335
				WHEN IND.PV >= 46 AND IND.PV < 52 THEN 0.2140
				WHEN IND.PV >= 52 AND IND.PV < 58 THEN 0.1984
				WHEN IND.PV >= 58 AND IND.PV < 60 THEN 0.1758
				ELSE 0.0000
			END AS FACTOR_A,
			0.8588 AS FACTOR_B, -- VALOR FIJO DE CATALOGO
			0 AS FACTOR_C, -- VALOR FIJO DE CATALOGO
			ISNULL(CAST(RV.CLAVE AS DECIMAL(3,0)),0)  AS REGION_VIVIENDA,
			50 AS CONVENIO, -- VALOR FIJO DE CATALOGO
			ISNULL(CAST(G.MON_AVALUO AS DECIMAL(21,2)),0)  AS VALORVIVIENDA,
			ISNULL(CAST (( @INPC2 / (SELECT VALOR FROM BRONZE.lmda.INPC_Historico 
			WHERE YEAR(FechaIndicador) = YEAR(CR.FEC_APERTURA) AND MONTH(FechaIndicador) = MONTH(CR.FEC_APERTURA)))
			AS DECIMAL(10,4)),0) AS FACTOR_VIVIENDA,
			3 AS TIPO_ACTUALIZACION, -- VALOR FIJO DE CATALOGO
			CASE
				WHEN CR.IND_ESTADO_CRED = 'E' 
				THEN ISNULL(CAST(DC.NUMERO_AVALUO AS VARCHAR(17)),'')   
				ELSE '0'
			END AS NUMERO_REAVALUO,
			CAST(CR.TASA_INTERES AS DECIMAL(3,0)) AS TASA_INTERES_EFECTIVA,
			0 AS HIPOTESIS_PRESUNCION, -- VALOR FIJO DE CATALOGO
			CASE
				WHEN CR.IND_ESTADO_CRED = 'E' AND CR.IND_ETAPA = 3 THEN ISNULL(CAST(RAH.MONTO_RESERVAS AS DECIMAL(21,2)),0) 
				ELSE 0
			END AS RESERVAS_ADICIONALES,  
			EOMONTH(@FechaSistema) AS FECHA_INFO,
			CASE
				WHEN CR.IND_ESTADO_CRED= 'E' THEN CAST((SELECT TOP 1 ID_EXTERNO FROM BRONZE.SAF.PR_CREDITOS WHERE CR.NUM_CONTRATO = NUM_CREDITO AND IND_ESTADO_CRED != 'E') AS VARCHAR(22))
				ELSE ''
			END AS CODIGOCREDITOREESTRUCTURA,
			0 AS TIPOGARANTIA, -- VALOR FIJO DE CATALOGO
			2 AS HITSIC, -- VALOR NO YA QUE NO APLICA PARA HIPOTECARIO
			ISNULL(CAST((SELECT MAX(NUM_SECUENCIA) 
			FROM BRONZE.SAF.PR_RUBRO_COBRO_X_CREDITO WHERE CAST(NUM_CREDITO AS INT)= CAST(CR.NUM_CREDITO AS INT)) AS DECIMAL(21,2)),1 )  AS SUMA_PROGRAMADOS,
			ISNULL(CAST(IND.CUENTA_EN_ORDEN_INTERES AS DECIMAL(21,2)),0)  AS INTERES_FUERA_BALANCE,
			ISNULL(CAST(IND.MONTO_BASE_INTERES AS DECIMAL(21,2)),0)  AS SALDO_BASE_INTERES
		FROM BRONZE.SAF.PR_CREDITOS CR
		LEFT JOIN BRONZE.SAF.PR_INF_RR_X_CREDITO INF ON CR.NUM_CREDITO = INF.NUM_CREDITO AND CR.COD_EMPRESA = INF.COD_EMPRESA AND CR.COD_AGENCIA = INF.COD_AGENCIA
		LEFT JOIN (SELECT * FROM BRONZE.CIERRE.IND_REP WHERE PERIODO = EOMONTH(@FechaSistema) AND CARTERA = 'ION') IND ON CR.ID_EXTERNO = IND.CLAVE
		LEFT JOIN (SELECT A.NUM_CREDITO, A.MON_PRINCIPAL_FONDEADOR FROM BRONZE.SAF.PR_TRASPASO_CARTERA_VENCIDA A
			JOIN (SELECT NUM_CREDITO, MAX(FEC_VENCIDO) FEC_VENCIDO FROM BRONZE.SAF.PR_TRASPASO_CARTERA_VENCIDA
			GROUP BY NUM_CREDITO) B ON A.NUM_CREDITO = B.NUM_CREDITO AND A.FEC_VENCIDO = B.FEC_VENCIDO) TCV ON CR.NUM_CREDITO = TCV.NUM_CREDITO
		LEFT JOIN (SELECT A.* FROM BRONZE.SAF.PR_GARANTIAS A
			JOIN (SELECT COD_CLIENTE, MAX(NUM_GARANTIA) NUM_GARANTIA FROM BRONZE.SAF.PR_GARANTIAS
			WHERE IND_ESTADO = 'A'
			AND TIP_GARANTIA = 'HIP'
			GROUP BY COD_CLIENTE) B ON A.COD_CLIENTE = B.COD_CLIENTE AND A.NUM_GARANTIA = B.NUM_GARANTIA) G ON CR.COD_CLIENTE = G.COD_CLIENTE 
		LEFT JOIN BRONZE.SAF.PR_GARANTIAS_HIPOTECARIAS GH ON G.NUM_GARANTIA = GH.NUM_GARANTIA
		LEFT JOIN (SELECT * FROM BRONZE.SAF.PR_ESTADO_CUENTA_ENC WHERE FEC_FIN_CORTE = EOMONTH(@FechaIni)) EC ON CR.NUM_CREDITO = EC.NUM_CREDITO
		LEFT JOIN BRONZE.RR.Clasificacion_Contable_CC_PE_H CATCC ON IND.CLASIFICACION_DE_VIVIENDA = CATCC.DESCRIPCION 
		AND IND.SIN_RESTRICCION_RESTRINGIDOS = CATCC.RESTRINGIDO AND  IND.ETAPA_RIESGO = CATCC.ETAPA
		LEFT JOIN BRONZE.SAF.CL_PERSONAS_FISICAS PF ON CR.COD_CLIENTE = PF.COD_CLIENTE
		LEFT JOIN BRONZE.RR.REGION_VIVIENDA RV ON GH.COD_PAIS = RV.COD_PAIS_SAF
		LEFT JOIN (SELECT * FROM BRONZE.SAF.PR_CREDITOS_HISTORICO WHERE FEC_CORTE = EOMONTH(DATEADD(MONTH,-1,@FechaFin))) CRH ON CR.NUM_CREDITO = CRH.NUM_CREDITO 
		LEFT JOIN BRONZE.RR.RESERVAS_ADICIONALES_HISTORICO RAH ON CR.ID_EXTERNO = RAH.ID_EXTERNO
		LEFT JOIN BRONZE.LMDA.DATOS_CREDITO DC ON DC.ID_CREDITO = CR.ID_EXTERNO
		--LEFT JOIN BRONZE.LMDA.DATOS_CLIENTE DR ON DR.ID_CREDITO = CR.ID_EXTERNO
		LEFT JOIN (SELECT * FROM ( SELECT *, ROW_NUMBER() OVER(PARTITION BY ID_CREDITO ORDER BY ID_CREDITO DESC) AS RN
		FROM BRONZE.LMDA.DATOS_CLIENTE) AS T WHERE RN = 1) DR ON DR.ID_CREDITO = CR.ID_EXTERNO
		WHERE CR.NUM_CREDITO IN
				(
				SELECT MAX(NUM_CREDITO) FROM BRONZE.SAF.PR_CREDITOS  -- SE OBTIENE UN SOLO CRÉDITO POR CLIENTE PARA EVITAR DUPLICIDAD
				WHERE 
				(IND_ESTADO = 'D' OR 
						(IND_ESTADO = 'C' 
						AND (FEC_CANCELACION >= @FechaIni
						AND FEC_CANCELACION < @FechaFin))
					) 
				AND NUM_DESC_TIPO_CREDITO IN (/*Hip*/ 12,13,14,15,28,3,4,16,17)
				--AND IND_LINEA != 'N'
				GROUP BY COD_CLIENTE
				)

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

# [dbo].[026_ENT_INTEGRACION_ACCIONARIA]

```sql
/****** Object:  StoredProcedure [dbo].[026_ENT_INTEGRACION_ACCIONARIA]    Script Date: 01/06/2026 20:11:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[026_ENT_INTEGRACION_ACCIONARIA] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[026_ENT_INTEGRACION_ACCIONARIA]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE,
		@Fecha DATE,
		@TipoRep INT,
		@MesIni INT

	BEGIN TRY
		SET @TipoRep = 3 --meses del periodo
		SET @MesIni = FLOOR((month(@FechaSistema) + @TipoRep - 1) / @TipoRep) * @TipoRep - @TipoRep + 1
		SET @FechaIni = DATEfromparts(year(@FechaSistema), @MesIni, 1)
		SET @FechaFin = DATEadd(month, @TipoRep, @FechaIni) 
		--------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[026_ENT_INTEGRACION_ACCIONARIA]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[026_ENT_INTEGRACION_ACCIONARIA]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[026_ENT_INTEGRACION_ACCIONARIA] (
			[AGENTECOLOCADOR],
			[NOMBREACCIONISTA],
			[RFC_ACCIONISTA],
			[PERSONALIDAD_JURIDICA],
			[TIPO_ACCIONISTA],
			[NUMERO_ACCIONISTAS],
			[RELACIONACCIONISTA],
			[TIPO_SERIE],
			[CLASE_ACCIONES],
			[NUMEROACCIONES],
			[VALOR_NOMINAL_X_ACCION],
			[VALOR_ACCIONES],
			[PORC_CAPITAL_SOCIAL],
			[FECHA_INFO]
			)
		SELECT 
			IA.AGENTECOLOCADOR AS AGENTECOLOCADOR,
			IA.NOMBREACCIONISTA AS NOMBREACCIONISTA,
			IA.RFC_ACCIONISTA AS RFC_ACCIONISTA,
			IA.PERSONALIDAD_JURIDICA AS PERSONALIDAD_JURIDICA,
			IA.TIPO_ACCIONISTA AS TIPO_ACCIONISTA,
			IA.NUMERO_ACCIONISTAS AS NUMERO_ACCIONISTAS,
			IA.RELACIONACCIONISTA AS RELACIONACCIONISTA,
			CAST(IA.TIPO_SERIE AS VARCHAR(2)) AS TIPO_SERIE,
			IA.CLASE_ACCIONES AS CLASE_ACCIONES,
			IA.NUMEROACCIONES AS NUMEROACCIONES,
			IA.VALOR_NOMINAL_X_ACCION AS VALOR_NOMINAL_X_ACCION,
			IA.VALOR_ACCIONES AS VALOR_ACCIONES,
			IA.PORC_CAPITAL_SOCIAL AS PORC_CAPITAL_SOCIAL,
			IA.FECHA_INFO AS FECHA_INFO
		FROM BRONZE.INF.INTEGRACION_ACCIONARIA IA
		WHERE FECHA_INFO >= @FechaIni AND FECHA_INFO < @FechaFin

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[029_ENT_FLUJOS]    Script Date: 01/06/2026 20:11:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[029_ENT_FLUJOS] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[029_ENT_FLUJOS]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE,
		@Fecha DATE,
		@TipoRep INT,
		@MesIni INT

	BEGIN TRY
		SET @TipoRep = 3 --meses del periodo
		SET @MesIni = FLOOR((month(@FechaSistema) + @TipoRep - 1) / @TipoRep) * @TipoRep - @TipoRep + 1
		SET @FechaIni = DATEfromparts(year(@FechaSistema), @MesIni, 1)
		SET @FechaFin = DATEadd(month, @TipoRep, @FechaIni)
		--------------------------------------------------------------------------------
		-- QUERY
		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[029_ENT_FLUJOS]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[029_ENT_FLUJOS]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[029_ENT_FLUJOS] (
			[CUENTA_CNBV_R13],
			[IMPORTE],
			[TIPO_SALDO],
			[FECHA_INFO]
			)
		SELECT 
			CUENTA_CNBV_R13,
			IMPORTE,
			TIPO_SALDO,
			FECHA_INFO
		FROM BRONZE.LMDA.FLUJOS_VARIACIONES
		WHERE FECHA_INFO >= @FechaIni
		AND FECHA_INFO < @FechaFin


		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[039_ENT_EVENTOS_ROP]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[039_ENT_EVENTOS_ROP] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[039_ENT_EVENTOS_ROP]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(year, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[039_ENT_EVENTOS_ROP]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[039_ENT_EVENTOS_ROP]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[039_ENT_EVENTOS_ROP] (
			[FOLIO_2812],
			[PRODUCTO],
			[PROCESO],
			[LINEA_NEGOCIO],
			[TIP_RIESGO_OPERACIONAL],
			[CALIFICACION_RIESGO],
			[DESCRIPCION_DEL_RIESGO_OPERACIONAL],
			[FECHA_INFO]
			)
		SELECT
			ER.FOLIO_2812,
			ISNULL(CAST(CAT_P.CLAVE AS VARCHAR(4)),'') AS PRODUCTO,
			ISNULL(CAST(CAT_R.CLAVE AS VARCHAR(3)),'') AS PROCESO,
			ISNULL(CAST(CAT_L.CLAVE AS VARCHAR(3)),'')  AS LINEA_NEGOCIO,
			ISNULL(LEFT(CAST(CAT_T.CLAVE AS VARCHAR(4)),4),'') AS TIP_RIESGO_OPERACIONAL,
			ER.CALIFICACION_RIESGO, 
			ISNULL(LEFT(ER.DESCRIPCION_DEL_RIESGO_OPERACIONAL,200),'') AS DESCRIPCION_DEL_RIESGO_OPERACIONAL,
			ISNULL(ER.FECHA_INFO,'') AS FECHA_INFO
		FROM BRONZE.LMDA.EVENTOS_ROP ER
		LEFT JOIN BRONZE.RR.ROP_PRODUCTO CAT_P ON ER.PRODUCTO = CAT_P.DESCRIPCION
		LEFT JOIN BRONZE.RR.ROP_PROCESO CAT_R ON ER.PROCESO = CAT_R.DESCRIPCION
		LEFT JOIN BRONZE.RR.ROP_LINEA_NEGOCIO CAT_L ON ER.LINEA_NEGOCIO = CAT_L.DESCRIPCION
		LEFT JOIN BRONZE.RR.ROP_TIPO_RIESGO_OPERACIONAL CAT_T ON ER.TIP_RIESGO_OPERACIONAL = CAT_T.DESCRIPCION
		WHERE ER.FECHA_INFO >= @FechaIni
		AND ER.FECHA_INFO < @FechaFin

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[045_ENT_CARTERA_LINEA]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[045_ENT_CARTERA_LINEA] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[045_ENT_CARTERA_LINEA] ';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[045_ENT_CARTERA_LINEA]
				WHERE [FECHA] >= @FechaIni
					AND [FECHA] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[045_ENT_CARTERA_LINEA]
			WHERE [FECHA] >= @FechaIni
				AND [FECHA] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[045_ENT_CARTERA_LINEA] (
			[TIPO_PRODUCTO],
			[FECHA],
			[NUM_LINEA],
			[MONTO_NO_DISP],
			[DESTINO],
			[TIPO_GAR],
			[TIPO_LINEA],
			[PONDERADOR],
			[GRUPO_RIESGO],
			[GRADO_RIESGO],
			[POND_CODIGO_CONF]
			)
		SELECT [TIPO_PRODUCTO],
			[FECHA],
			[NUM_LINEA],
			[MONTO_NO_DISP],
			[DESTINO],
			[TIPO_GAR],
			[TIPO_LINEA],
			[PONDERADOR],
			[GRUPO_RIESGO],
			[GRADO_RIESGO],
			[POND_CODIGO_CONF]
		FROM [SILVER].[RR].[045_ENT_CARTERA_LINEA]
		WHERE [FECHA] >= @FechaIni
			AND [FECHA] < @FechaFin

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[047_ENT_BIENES_ADJUDICADOS]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[047_ENT_BIENES_ADJUDICADOS] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @FechaIni DATE,
		@FechaFin DATE;
	DECLARE @NombreJob NVARCHAR(128) = '[047_ENT_BIENES_ADJUDICADOS]';

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[047_ENT_BIENES_ADJUDICADOS]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[047_ENT_BIENES_ADJUDICADOS]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END

		INSERT INTO [RR].[047_ENT_BIENES_ADJUDICADOS] (
			[ID_ACT_ADJ],
			[TIPO_ACT_ADJ],
			[FECHA_ADJ],
			[IMPORTE_ACT_ADJ],
			[MOV_ADJ],
			[ESTIMACION_TRASP],
			[ESTIMACION_ADIC],
			[TIPO_ESTIMACION],
			[FECHA_INFO]
			)
		SELECT 
			BA.ID_ACT_ADJ AS ID_ACT_ADJ,
			CASE
				WHEN BA.TIPO_ACT_ADJ = 'Bienes muebles' THEN 1
				WHEN BA.TIPO_ACT_ADJ = 'Acciones' THEN 2
				WHEN BA.TIPO_ACT_ADJ = 'Títulos de deuda bancaria' THEN 3
				WHEN BA.TIPO_ACT_ADJ = 'Títulos de deuda gubernamental' THEN 4
				WHEN BA.TIPO_ACT_ADJ = 'Otros títulos de deuda' THEN 5
				WHEN BA.TIPO_ACT_ADJ = 'Derechos de cobro' THEN 6
				WHEN BA.TIPO_ACT_ADJ = 'Cartera de créditos' THEN 7 
				WHEN BA.TIPO_ACT_ADJ = 'Otros' THEN 8
				WHEN BA.TIPO_ACT_ADJ = 'Muebles restringidos' THEN 9
				WHEN BA.TIPO_ACT_ADJ = 'Inmuebles adjudicados, otorgados en arrendamiento puro' THEN 10
				WHEN BA.TIPO_ACT_ADJ = 'Inmuebles adjudicados, otorgados en arrendamiento con opción de compra' THEN 11
				WHEN BA.TIPO_ACT_ADJ = 'Otros inmuebles adjudicados' THEN 12
				WHEN BA.TIPO_ACT_ADJ = 'Inmuebles adjudicados restringidos' THEN 13
				WHEN BA.TIPO_ACT_ADJ = 'Incremento por actualización de inmuebles adjudicados' THEN 14 
				ELSE 0
			END AS TIPO_ACT_ADJ,
			BA.FECHA_ADJ AS FECHA_ADJ,
			ISNULL(BA.IMPORTE_ACT_ADJ,0) AS IMPORTE_ACT_ADJ,
			CASE
				WHEN BA.MOV_ADJ = 'Adjudicaciones ejecutadas del periodo' THEN '1'
				WHEN BA.MOV_ADJ = 'Adjudicaciones registradas en la contabilidad en el periodo' THEN '2'
				WHEN BA.MOV_ADJ = 'Ventas' THEN '3'
				WHEN BA.MOV_ADJ = 'Para uso de la institución' THEN '4'
				WHEN BA.MOV_ADJ = 'Cancelaciones' THEN '5'
				WHEN BA.MOV_ADJ = 'Traspasos' THEN 'X'
				ELSE ''
			END AS MOV_ADJ,
			ISNULL(BA.ESTIMACION_TRASP,0) AS ESTIMACION_TRASP,
			ISNULL(BA.ESTIMACION_ADIC,0) AS ESTIMACION_ADIC,
			CASE
				WHEN BA.TIPO_ESTIMACION = 'Por arrendamiento' THEN 1
				WHEN BA.TIPO_ESTIMACION = 'Por pérdida de valor del bien adjudicado' THEN 2
				WHEN BA.TIPO_ESTIMACION = 'Incremento por actualización de estimaciones por pérdida de valor de bienes adjudicados' THEN 3
				ELSE 0
			END AS TIPO_ESTIMACION,
			BA.FECHA_INFO AS FECHA_INFO
		FROM BRONZE.INF.BIENES_ADJUDICADOS BA
		WHERE BA.FECHA_INFO >= @FechaIni
		AND BA.FECHA_INFO < @FechaFin


		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[051_ENT_CONTRAPARTES]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[051_ENT_CONTRAPARTES] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[051_ENT_CONTRAPARTES] ';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[051_ENT_CONTRAPARTES]
				WHERE [FECHA_REPORTE] >= @FechaIni
					AND [FECHA_REPORTE] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[051_ENT_CONTRAPARTES]
			WHERE [FECHA_REPORTE] >= @FechaIni
				AND [FECHA_REPORTE] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[051_ENT_CONTRAPARTES] (
			[CONTRATO],
			[NOMBRE_CORTO],
			[NOMBRE_LARGO],
			[GRUPO],
			[GRADO_RIESGO],
			[CALIFICACION],
			[CLAVE_BANXICO],
			[ESCALA_CAL],
			[PLAZO_CAL],
			[AGENCIA],
			[TIPOCONTRAPARTESECTOR],
			[RFC_PER_REL],
			[PERSONA_RELACIONADA],
			[LEI],
			[JURISDICCION],
			[FECHA_REPORTE]
			)
		SELECT [CONTRATO],
			[NOMBRE_CORTO],
			[NOMBRE_LARGO],
			[GRUPO],
			[GRADO_RIESGO],
			[CALIFICACION],
			[CLAVE_BANXICO],
			[ESCALA_CAL],
			[PLAZO_CAL],
			[AGENCIA],
			[TIPOCONTRAPARTESECTOR],
			[RFC_PER_REL],
			[PERSONA_RELACIONADA],
			[LEI],
			[JURISDICCION],
			@FechaIni AS [FECHA_REPORTE]
		FROM [BRONZE].[LMDA].[CONTRAPARTES]
		WHERE [FECHA_REPORTE] >= @FechaIni
			AND [FECHA_REPORTE] < @FechaFin

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[054_ENT_RC12IN_IN_MIN]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[054_ENT_RC12IN_IN_MIN] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[054_ENT_RC12IN_IN_MIN]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[054_ENT_RC12IN_IN_MIN]
				WHERE [FECHA] >= @FechaIni
					AND [FECHA] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[054_ENT_RC12IN_IN_MIN]
			WHERE [FECHA] >= @FechaIni
				AND [FECHA] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[054_ENT_RC12IN_IN_MIN] (
			[FECHA],
			[I_INT],
			[G_INT],
			[I_DIV],
			[AP],
			[OIO],
			[OGO],
			[CTC],
			[CTP],
			[RC]
			)
		SELECT [FECHA],
			[I_INT],
			[G_INT],
			[I_DIV],
			[AP],
			[OIO],
			[OGO],
			[CTC],
			[CTP],
			[RC]
		FROM [BRONZE].[LMDA].[RC12IN_IN_MIN]
		WHERE [FECHA] >= @FechaIni
			AND [FECHA] < @FechaFin

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[058_ENT_SUCURSALES]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[058_ENT_SUCURSALES] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[058_ENT_SUCURSALES]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[058_ENT_SUCURSALES]
				WHERE [FECHA] >= @FechaIni
					AND [FECHA] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[058_ENT_SUCURSALES]
			WHERE [FECHA] >= @FechaIni
				AND [FECHA] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[058_ENT_SUCURSALES] (
			[MUNICIPIO],
			[CODIGO_POSTAL],
			[LOCALIDAD],
			[GENERO],
			[PERSONALIDAD_JURIDICA],
			[TIPO_CARTERA],
			[TIPO_INFORMACION_OPERATIVA],
			[NUMERO_TOTAL],
			[FECHA]
			)
		SELECT [MUNICIPIO],
			[CODIGO_POSTAL],
			[LOCALIDAD],
			[GENERO],
			[PERSONALIDAD_JURIDICA],
			[TIPO_CARTERA],
			[TIPO_INFORMACION_OPERATIVA],
			[NUMERO_TOTAL],
			[FECHA]
		FROM [SILVER].[RR].[058_ENT_SUCURSALES]
		WHERE [FECHA] >= @FechaIni
			AND [FECHA] < @FechaFin

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[067_CUP_ELEGIBILIDAD]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
--	Modificado: Alberto Vazquez
CREATE PROCEDURE [dbo].[067_CUP_ELEGIBILIDAD] 
	@CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[067_CUP_ELEGIBILIDAD] ';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[067_CUP_ELEGIBILIDAD]
				WHERE [fechaInicioOferta] >= @FechaIni
					AND [fechaInicioOferta] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[067_CUP_ELEGIBILIDAD]
			WHERE [fechaInicioOferta] >= @FechaIni
				AND [fechaInicioOferta] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[067_CUP_ELEGIBILIDAD] (
			[fechaInicioOferta],
			[identificadorProducto],
			[identificadorSubproducto],
			[tipo_de_instrumento],
			[Edad_minima],
			[Edad_maxima],
			[Antig_laborar_min],
			[Antig_resid_min],
			[Ingreso_mensual_min],
			[Interpretacion_ing_mensual],
			[Caract_especificas_consumidor],
			[Genero],
			[Publico],
			[Requiere_invitacion],
			[Primer_credito],
			[Score_credito_min]
			)
		SELECT  [fechaInicioOferta],
				[identificadorProducto],
				[identificadorSubproducto],
				[tipo_de_instrumento],
				[Edad_minima],
				[Edad_maxima],
				[Antig_laborar_min],
				[Antig_resid_min],
				[Ingreso_mensual_min],
				[Interpretacion_ing_mensual],
				[Caract_especificas_consumidor],
				[Genero],
				[Publico],
				[Requiere_invitacion],
				[Primer_credito],
				[Score_credito_min]
		FROM [SILVER].[RR].[067_CUP_ELEGIBILIDAD]
		WHERE [fechaInicioOferta] >= @FechaIni
			AND [fechaInicioOferta] < @FechaFin

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[070_CUP_CNR_PENALIZACION]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[070_CUP_CNR_PENALIZACION] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[070_CUP_CNR_PENALIZACION] ';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[070_CUP_CNR_PENALIZACION]
				WHERE [f_ini_oferta] >= @FechaIni
					AND [f_ini_oferta] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[070_CUP_CNR_PENALIZACION]
			WHERE [f_ini_oferta] >= @FechaIni
				AND [f_ini_oferta] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[070_CUP_CNR_PENALIZACION] (
			[f_ini_oferta],
			[id_producto],
			[id_subproducto],
			[id_operativo],
			[tipo_comision_pena_nom_per],
			[disponible],
			[operaciones_gratis],
			[moneda],
			[base_cobro],
			[tipo_valor],
			[tipo_periodo],
			[excepciones],
			[valor_comision],
			[comentario],
			[tipo_trabajador],
			[tipo_cliente]
			)
		SELECT [f_ini_oferta],
			[id_producto],
			[id_subproducto],
			[id_operativo],
			[tipo_comision_pena_nom_per],
			[disponible],
			[operaciones_gratis],
			[moneda],
			[base_cobro],
			[tipo_valor],
			[tipo_periodo],
			[excepciones],
			[valor_comision],
			[comentario],
			[tipo_trabajador],
			[tipo_cliente]
		FROM [SILVER].[RR].[070_CUP_CNR_PENALIZACION]
		WHERE [f_ini_oferta] >= @FechaIni
			AND [f_ini_oferta] < @FechaFin

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[073_CUP_DINPERSONAL]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
--	Modificado: Alberto Vazquez
CREATE PROCEDURE [dbo].[073_CUP_DINPERSONAL] 
	@CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[073_CUP_DINPERSONAL] ';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[073_CUP_DINPERSONAL]
				WHERE [f_ini_oferta] >= @FechaIni
					AND [f_ini_oferta] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[073_CUP_DINPERSONAL]
			WHERE [f_ini_oferta] >= @FechaIni
				AND [f_ini_oferta] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[073_CUP_DINPERSONAL] (
			[f_ini_oferta],
			[id_producto],
			[id_subproducto],
			[liga_solicitud],
			[tiem_prom_valor],
			[tiem_prom_tipo_peri],
			[domiciliacion_pagos]
			)
		SELECT	[f_ini_oferta],
				[id_producto],
				[id_subproducto],
				[liga_solicitud],
				[tiem_prom_valor],
				[tiem_prom_tipo_peri],
				[domiciliacion_pagos]
		FROM [SILVER].[RR].[073_CUP_DINPERSONAL]
		WHERE [f_ini_oferta] >= @FechaIni
			AND [f_ini_oferta] < @FechaFin

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[083_ENT_CVT_IN]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[083_ENT_CVT_IN] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[083_ENT_CVT_IN]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[083_ENT_CVT_IN]
				WHERE [FECHACONCERTACION] = @FechaSistema
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[083_ENT_CVT_IN]
			WHERE [FECHACONCERTACION] = @FechaSistema;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[083_ENT_CVT_IN] (
			[FECHACONCERTACION],
			[NUMEROOPERACIONINCUMPLIDA],
			[SECCION],
			[FECHAREGISTROINCUMPLIMIENTO],
			[TIPOINCUMPLIMIENTO],
			[NUMEROTITULOSINCUMPLIDOS],
			[NUMEROINCUMPLIMIENTO],
			[CAUSANTEINCUMPLIMIENTO],
			[RAZONINCUMPLIMIENTO],
			[TIPOMODIFICACION]
			)
		SELECT [FECHACONCERTACION],
			[NUMEROOPERACIONINCUMPLIDA],
			[SECCION],
			[FECHAREGISTROINCUMPLIMIENTO],
			[TIPOINCUMPLIMIENTO],
			[NUMEROTITULOSINCUMPLIDOS],
			[NUMEROINCUMPLIMIENTO],
			[CAUSANTEINCUMPLIMIENTO],
			[RAZONINCUMPLIMIENTO],
			[TIPOMODIFICACION]
		FROM [SILVER].[RR].[083_ENT_CVT_IN]
		WHERE [FECHACONCERTACION] = @FechaSistema

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[099_MDEF_BOVEDA_EXTERNA]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
--	Modificado: Alberto Vazquez
CREATE PROCEDURE [dbo].[099_MDEF_BOVEDA_EXTERNA] 
	@CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[099_MDEF_BOVEDA_EXTERNA] ';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[099_MDEF_BOVEDA_EXTERNA]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[099_MDEF_BOVEDA_EXTERNA]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[099_MDEF_BOVEDA_EXTERNA] (
			[CLAVE],
			[NOMBRE],
			[CALLE],
			[NUMERO],
			[COLONIA],
			[CODIGO_POSTAL],
			[REFERENCIA],
			[C_LOCALIDAD],
			[C_MUNICIPIO],
			[C_ESTADO],
			[LATITUD],
			[LONGITUD],
			[TIPO_BOVEDA],
			[CORREO],
			[TELEFONO],
			[FECHA_EVENTO],
			[TIPO_EVENTO],
			[FECHA_INFO]
			)
		SELECT	[CLAVE],
				[NOMBRE],
				[CALLE],
				[NUMERO],
				[COLONIA],
				[CODIGO_POSTAL],
				[REFERENCIA],
				[C_LOCALIDAD],
				[C_MUNICIPIO],
				[C_ESTADO],
				[LATITUD],
				[LONGITUD],
				[TIPO_BOVEDA],
				[CORREO],
				[TELEFONO],
				[FECHA_EVENTO],
				[TIPO_EVENTO],
				[FECHA_INFO]
		FROM [SILVER].[RR].[099_MDEF_BOVEDA_EXTERNA]
		WHERE [FECHA_INFO] >= @FechaIni
			AND [FECHA_INFO] < @FechaFin

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[102_ENT_TASAS_PASIVAS]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[102_ENT_TASAS_PASIVAS] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[102_ENT_TASAS_PASIVAS]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[102_ENT_TASAS_PASIVAS]
				WHERE [FECHA] >= @FechaIni
					AND [FECHA] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[102_ENT_TASAS_PASIVAS]
			WHERE [FECHA] >= @FechaIni
				AND [FECHA] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[102_ENT_TASAS_PASIVAS] (
			FECHA_EXTRACCION,
			CVECTA,
			PLAZO1,
			PLAZO2,
			IMPLIM1,
			IMPLIM2,
			TASA,
			FORTASA,
			TASAR,
			PUNTOS,
			FACTOR,
			FECHA
			)
		SELECT FECHA_EXTRACCION,
			CVECTA,
			PLAZO1,
			PLAZO2,
			IMPLIM1,
			IMPLIM2,
			TASA,
			FORTASA,
			TASAR,
			PUNTOS,
			FACTOR,
			FECHA
		FROM [SILVER].[RR].[102_ENT_TASAS_PASIVAS]
		WHERE [FECHA] >= @FechaIni
			AND [FECHA] < @FechaFin

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[105_SISPAGOS_IO]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[105_SISPAGOS_IO] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[105_SISPAGOS_IO]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE,
		@Fecha DATE,
		@TipoRep INT,
		@MesIni INT

	BEGIN TRY
		SET @TipoRep = 3 --meses del periodo
		SET @MesIni = FLOOR((month(@Fecha) + @TipoRep - 1) / @TipoRep) * @TipoRep - @TipoRep + 1
		SET @FechaIni = DATEfromparts(year(@Fecha), @MesIni, 1)
		SET @FechaFin = DATEadd(month, @TipoRep, @FechaIni)

		--------------------------------------------------------------------------------
		-- QUERY
		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[105_SISPAGOS_IO]
				WHERE [FECHA_REPORTE] >= @FechaIni
					AND [FECHA_REPORTE] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[105_SISPAGOS_IO]
			WHERE [FECHA_REPORTE] >= @FechaIni
				AND [FECHA_REPORTE] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[105_SISPAGOS_IO] (
			[AÑO],
			[MES],
			[SECCION],
			[MONEDA],
			[TIPO_DE_CUENTA],
			[TIPO_DE_PERSONA],
			[NUMERO_DE_CUENTAS],
			[SALDO_PROMEDIO_DE_LAS_CUENTAS],
			[FECHA_REPORTE]
			)
		SELECT [AÑO],
			[MES],
			[SECCION],
			[MONEDA],
			[TIPO_DE_CUENTA],
			[TIPO_DE_PERSONA],
			[NUMERO_DE_CUENTAS],
			[SALDO_PROMEDIO_DE_LAS_CUENTAS],
			@FechaIni AS [FECHA_REPORTE]
		FROM [SILVER].[RR].[105_SISPAGOS_IO]
		WHERE [FECHA_REPORTE] >= @FechaIni
			AND [FECHA_REPORTE] < @FechaFin

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[111_ENT_CIP]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[111_ENT_CIP] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[111_ENT_CIP]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[111_ENT_CIP]
				WHERE [fecha] >= @FechaIni
					AND [fecha] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[111_ENT_CIP]
			WHERE [fecha] >= @FechaIni
				AND [fecha] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[111_ENT_CIP] (
			[institucion],
			[fecha],
			[tipo_tarjeta],
			[giro_comercial],
			[nivel_ci],
			[variacion]
			)
		SELECT [institucion],
			[fecha],
			[tipo_tarjeta],
			[giro_comercial],
			[nivel_ci],
			[variacion]
		FROM [SILVER].[RR].[111_ENT_CIP]
		WHERE [fecha] >= @FechaIni
			AND [fecha] < @FechaFin

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[114_ENT_R15_1522]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
--	Modificado: Alberto Vazquez
CREATE PROCEDURE [dbo].[114_ENT_R15_1522] 
	@CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[114_ENT_R15_1522] ';
	DECLARE @FechaIni DATE,
		@FechaFin DATE,
		@Fecha DATE,
		@TipoRep INT,
		@MesIni INT

	BEGIN TRY
		SET @TipoRep = 3 --meses del periodo
		SET @MesIni = FLOOR((month(@Fecha) + @TipoRep - 1) / @TipoRep) * @TipoRep - @TipoRep + 1
		SET @FechaIni = DATEfromparts(year(@Fecha), @MesIni, 1)
		SET @FechaFin = DATEadd(month, @TipoRep, @FechaIni)

		--------------------------------------------------------------------------------
		-- QUERY
		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[114_ENT_R15_1522]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[114_ENT_R15_1522]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[114_ENT_R15_1522] (
			[TIPO_PRODUCTO],
			[SERVICIO_BANCA],
			[TIPO_OPERACION],
			[NUMERO_USUARIOS_OPERA],
			[NUMERO_OPERACIONES],
			[IMPORTE],
			[FECHA_INFO]
			)
		SELECT [TIPO_PRODUCTO],
			[SERVICIO_BANCA],
			[TIPO_OPERACION],
			[NUMERO_USUARIOS_OPERA],
			[NUMERO_OPERACIONES],
			[IMPORTE],
			[FECHA_INFO]
		FROM [SILVER].[RR].[114_ENT_R15_1522]
		WHERE [FECHA_INFO] >= @FechaIni AND [FECHA_INFO] < @FechaFin

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[117_R26_MODULO]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[117_R26_MODULO] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[117_R26_MODULO] ';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[117_R26_MODULO]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[117_R26_MODULO]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[117_R26_MODULO] (
			[RFC_COM],
			[CLAVE_MODULO],
			[LOCALIDAD_MODULO],
			[TIPO_MOV],
			[CAUSA_BAJA],
			[FECHA_INFO]
			)
		SELECT [RFC_COM],
			[CLAVE_MODULO],
			[LOCALIDAD_MODULO],
			[TIPO_MOV],
			[CAUSA_BAJA],
			[FECHA_INFO]
		FROM [SILVER].[RR].[117_R26_MODULO]
		WHERE [FECHA_INFO] >= @FechaIni
			AND [FECHA_INFO] < @FechaFin

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[120_ENT_TIPOCAMBIO_DIARIO]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[120_ENT_TIPOCAMBIO_DIARIO] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[120_ENT_TIPOCAMBIO_DIARIO]    ';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[120_ENT_TIPOCAMBIO_DIARIO]
				WHERE [FECHA] = @FechaSistema
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[120_ENT_TIPOCAMBIO_DIARIO]
			WHERE [FECHA] = @FechaSistema;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[120_ENT_TIPOCAMBIO_DIARIO] (
			[FECHA],
			[TIPODECAMBIO],
			[MONEDA]
			)
		SELECT [FECHA],
			[TIPODECAMBIO],
			[MONEDA]
		FROM [SILVER].[RR].[120_ENT_TIPOCAMBIO_DIARIO]
		WHERE [FECHA] = @FechaSistema

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[123_ENT_REPORTOS_VENC_ANT]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[123_ENT_REPORTOS_VENC_ANT] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[123_ENT_REPORTOS_VENC_ANT]  ';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[123_ENT_REPORTOS_VENC_ANT]
				WHERE [FECHACONCERTACION] = @FechaSistema
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[123_ENT_REPORTOS_VENC_ANT]
			WHERE [FECHACONCERTACION] = @FechaSistema;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[123_ENT_REPORTOS_VENC_ANT] (
			[FECHACONCERTACION],
			[FECHA_VENC_ANT],
			[MOTIVO_VEN_ANT],
			[NUMEROIDENTIFICACIONOPERACION],
			[NUMEROIDENTIFICACIONVENCANT],
			[SECCION]
			)
		SELECT [FECHACONCERTACION],
			[FECHA_VENC_ANT],
			[MOTIVO_VEN_ANT],
			[NUMEROIDENTIFICACIONOPERACION],
			[NUMEROIDENTIFICACIONVENCANT],
			[SECCION]
		FROM [SILVER].[RR].[123_ENT_REPORTOS_VENC_ANT]
		WHERE [FECHACONCERTACION] = @FechaSistema

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[131_ENT_OFF_CONVAL]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[131_ENT_OFF_CONVAL] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[131_ENT_OFF_CONVAL] ';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY --------------------------------------------------------------------------------
		-- QUERY
		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[131_ENT_OFF_CONVAL]
				WHERE [FECHA] = @FechaSistema
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[131_ENT_OFF_CONVAL]
			WHERE [FECHA] = @FechaSistema;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[131_ENT_OFF_CONVAL] (
			[FECHA],
			[FECVEN],
			[FECLIQ],
			[OFICINA],
			[OPERACION],
			[MERCADO],
			[PIZARRA],
			[OBJETIVO],
			[NUMCONTRAT],
			[NUMSUBY],
			[MONTOBASE],
			[MDAMONTO],
			[PRECIOENT],
			[MDAPRECIO],
			[LIQUIDA],
			[MDALIQUIDA],
			[SUBY],
			[MDASUBY],
			[CONTRAPAR],
			[NUM_ID],
			[MODIFICA],
			[ESTRATEGIA],
			[ID_ESTRA],
			[ENGRAPADO],
			[SPOT],
			[TIPOCONTRAPARTESECTOR],
			[FE_INI_OPE],
			[TIP_DER],
			[QUANTO],
			[TC_QUANT],
			[TIP_VEN],
			[TER_OPE],
			[CVE_TIT],
			[NUM_ID_OP_SUBY],
			[BROKER],
			[SOCIO_LIQ],
			[CAM_COM],
			[AG_CAL],
			[NUM_CONF],
			[FECVEN_A],
			[FECLIQ_A],
			[IMP_VEN],
			[MDA_VEN],
			[VEN_PAR],
			[PAG_VENA],
			[MOT_VENA],
			[VEN_ANT],
			[NUM_ID_CP],
			[PRO_TER],
			[OPER_DER],
			[DIASLIQ],
			[UTI_N],
			[UTI],
			[UPI],
			[ES_ESTRATEGIA]
			)
		SELECT [FECHA],
			[FECVEN],
			[FECLIQ],
			[OFICINA],
			[OPERACION],
			[MERCADO],
			[PIZARRA],
			[OBJETIVO],
			[NUMCONTRAT],
			[NUMSUBY],
			[MONTOBASE],
			[MDAMONTO],
			[PRECIOENT],
			[MDAPRECIO],
			[LIQUIDA],
			[MDALIQUIDA],
			[SUBY],
			[MDASUBY],
			[CONTRAPAR],
			[NUM_ID],
			[MODIFICA],
			[ESTRATEGIA],
			[ID_ESTRA],
			[ENGRAPADO],
			[SPOT],
			[TIPOCONTRAPARTESECTOR],
			[FE_INI_OPE],
			[TIP_DER],
			[QUANTO],
			[TC_QUANT],
			[TIP_VEN],
			[TER_OPE],
			[CVE_TIT],
			[NUM_ID_OP_SUBY],
			[BROKER],
			[SOCIO_LIQ],
			[CAM_COM],
			[AG_CAL],
			[NUM_CONF],
			[FECVEN_A],
			[FECLIQ_A],
			[IMP_VEN],
			[MDA_VEN],
			[VEN_PAR],
			[PAG_VENA],
			[MOT_VENA],
			[VEN_ANT],
			[NUM_ID_CP],
			[PRO_TER],
			[OPER_DER],
			[DIASLIQ],
			[UTI_N],
			[UTI],
			[UPI],
			[ES_ESTRATEGIA]
		FROM [BRONZE].[LMDA].[OFF_CONVAL]
		WHERE [FECHA] = @FechaSistema

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[141_ENT_GARANTIAS_III]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[141_ENT_GARANTIAS_III] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[141_ENT_GARANTIAS_III] ';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = DATEADD(DAY, - (DATEPART(WEEKDAY, @FechaSistema) + 5) % 7, @FechaSistema)
		SET @FechaFin = DATEADD(DAY, 7, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[141_ENT_GARANTIAS_III]
				WHERE [FECHA_REPORTE] >= @FechaIni
					AND [FECHA_REPORTE] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[141_ENT_GARANTIAS_III]
			WHERE [FECHA_REPORTE] >= @FechaIni
				AND [FECHA_REPORTE] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[141_ENT_GARANTIAS_III] (
			[OFICINA],
			[MERCADO],
			[SOCIO_LIQ],
			[EXP_POT],
			[RSG_MAX],
			[FECHA_REPORTE]
			)
		SELECT [OFICINA],
			[MERCADO],
			[SOCIO_LIQ],
			[EXP_POT],
			[RSG_MAX],
			@FechaIni AS [FECHA_REPORTE]
		FROM [SILVER].[RR].[141_ENT_GARANTIAS_III]
		WHERE [FECHA_REPORTE] >= @FechaIni
			AND [FECHA_REPORTE] < @FechaFin

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[152_ENT_OPTO_X]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[152_ENT_OPTO_X] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[152_ENT_OPTO_X]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[152_ENT_OPTO_X]
				WHERE [FE_CON_OPE] >= @FechaIni
					AND [FE_CON_OPE] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[152_ENT_OPTO_X]
			WHERE [FE_CON_OPE] >= @FechaIni
				AND [FE_CON_OPE] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[152_ENT_OPTO_X] (
			[FE_CON_OPE],
			[FE_CORTE],
			[LEI],
			[NUM_ID],
			[OBJ_OPE],
			[NATURALEZA],
			[VOL_IMPLIC],
			[DELTA_OPC],
			[MDADELTA],
			[GAMMA_OPC],
			[VEGA_OPC],
			[RHO],
			[TASA_RHO],
			[PHI],
			[TASA_PHI],
			[RHO_COL],
			[TASA_RHO_COL],
			[PRECIO_CORTE],
			[SEN_COMP],
			[CURVA_REF],
			[CURVA_REF2],
			[CURVA_REF3],
			[PAQ_OPC],
			[ID_PAQ],
			[CONSEC_PAQ],
			[PRI_ACT],
			[ID_GAR_VIG]
			)
		SELECT [FE_CON_OPE],
			[FE_CORTE],
			[LEI],
			[NUM_ID],
			[OBJ_OPE],
			[NATURALEZA],
			[VOL_IMPLIC],
			[DELTA_OPC],
			[MDADELTA],
			[GAMMA_OPC],
			[VEGA_OPC],
			[RHO],
			[TASA_RHO],
			[PHI],
			[TASA_PHI],
			[RHO_COL],
			[TASA_RHO_COL],
			[PRECIO_CORTE],
			[SEN_COMP],
			[CURVA_REF],
			[CURVA_REF2],
			[CURVA_REF3],
			[PAQ_OPC],
			[ID_PAQ],
			[CONSEC_PAQ],
			[PRI_ACT],
			[ID_GAR_VIG]
		FROM [SILVER].[RR].[152_ENT_OPTO_X]
		WHERE [FE_CON_OPE] >= @FechaIni
			AND [FE_CON_OPE] < @FechaFin

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[164_CUP_EST_HIPO_TERR_CRED]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[164_CUP_EST_HIPO_TERR_CRED] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[164_CUP_EST_HIPO_TERR_CRED]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[164_CUP_EST_HIPO_TERR_CRED]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[164_CUP_EST_HIPO_TERR_CRED]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[164_CUP_EST_HIPO_TERR_CRED] (
			[fechaInicioOferta],
			[fechaInicioSubproducto],
			[fechaFinSubproducto],
			[identificadorProducto],
			[Identificador_subproducto],
			[nombre_subproducto],
			[numeroReca],
			[Compracartera],
			[Aforo],
			[plazoMinimo],
			[plazoMaximo],
			[destino],
			[garantia],
			[limiteCreditoMin],
			[limiteCreditoMax],
			[Coacreditacion],
			[persona],
			[fiscal],
			[tipodebien],
			[AntiguedadBien],
			[tipoTasa],
			[referenciaTasa],
			[FECHA_INFO]
			)
		SELECT 
			[fechaInicioOferta],
			[fechaInicioSubproducto],
			[fechaFinSubproducto],
			[identificadorProducto],
			[Identificador_subproducto],
			[nombre_subproducto],
			[numeroReca],
			[Compracartera],
			[Aforo],
			[plazoMinimo],
			[plazoMaximo],
			[destino],
			[garantia],
			[limiteCreditoMin],
			[limiteCreditoMax],
			[Coacreditacion],
			[persona],
			[fiscal],
			[tipodebien],
			[AntiguedadBien],
			[tipoTasa],
			[referenciaTasa],
			[FECHA_INFO]
		FROM BRONZE.[ORI].[EST_HIPO_CREDITO]
		WHERE 
		FECHA_INFO >= @FechaIni
		AND FECHA_INFO < @FechaFin
		AND tipProd = 'terreno'

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log
	
	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[167_CUP_EST_EMP_PLIQ_CRED]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[167_CUP_EST_EMP_PLIQ_CRED] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[167_CUP_EST_EMP_PLIQ_CRED]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[167_CUP_EST_EMP_PLIQ_CRED]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[167_CUP_EST_EMP_PLIQ_CRED]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[167_CUP_EST_EMP_PLIQ_CRED] (
			[fechaInicioOferta],
			[fechaInicioSubproducto],
			[fechaFinSubproducto],
			[identificadorProducto],
			[Identificador_subproducto],
			[nombre_subproducto],
			[accesorio],
			[numeroReca],
			[destino],
			[arrendamiento],
			[redisponibiliidad],
			[tipoFactoraje],
			[garantia],
			[periodicidad],
			[limiteCreditoMin],
			[limiteCreditoMax],
			[aceptacion],
			[mediosDisposicion],
			[CAT],
			[tipoTasa],
			[referenciaTasa],
			[FECHA_INFO]
			)
		SELECT 
			[fechaInicioOferta],
			[fechaInicioSubproducto],
			[fechaFinSubproducto],
			[identificadorProducto],
			[Identificador_subproducto],
			[nombre_subproducto],
			[accesorio],
			[numeroReca],
			[destino],
			[arrendamiento],
			[redisponibiliidad],
			[tipoFactoraje],
			[garantia],
			[periodicidad],
			[limiteCreditoMin],
			[limiteCreditoMax],
			[aceptacion],
			[mediosDisposicion],
			[CAT],
			[tipoTasa],
			[referenciaTasa],
			[FECHA_INFO]
		FROM BRONZE.[ORI].[EST_EMPRESARIAL_CREDITO]
		WHERE 
		FECHA_INFO >= @FechaIni
		AND FECHA_INFO < @FechaFin
		AND tipProd = 'pyme liquidez'

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log
	
	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[170_CUP_BASE_OP_HIPO_TERR_CRED]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[170_CUP_BASE_OP_HIPO_TERR_CRED] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[170_CUP_BASE_OP_HIPO_TERR_CRED]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[170_CUP_BASE_OP_HIPO_TERR_CRED]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[170_CUP_BASE_OP_HIPO_TERR_CRED]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[170_CUP_BASE_OP_HIPO_TERR_CRED] (
			[fechaInicioOferta],
			[identificadorProducto],
			[Identificador_subproducto],
			[Identificadoroperativo],
			[Nombre_Operativo],
			[Programa],
			[fechaInicioOferta_Op],
			[fechaFinOferta_Op],
			[FECHA_INFO]
			)
		SELECT 
			[fechaInicioOferta],
			[identificadorProducto],
			[Identificador_subproducto],
			[Identificadoroperativo],
			[Nombre_Operativo],
			[Programa],
			[fechaInicioOferta_Op],
			[fechaFinOferta_Op],
			[FECHA_INFO]
		FROM BRONZE.[ORI].[BASE_OPERATIVA_HIPO_CREDITO]
		WHERE 
		FECHA_INFO >= @FechaIni
		AND FECHA_INFO < @FechaFin
		AND tipProd = 'terreno'

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log
	
	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[173_CUP_ELEG_HIPO_TERR_CRED]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[173_CUP_ELEG_HIPO_TERR_CRED] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[173_CUP_ELEG_HIPO_TERR_CRED]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[173_CUP_ELEG_HIPO_TERR_CRED]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[173_CUP_ELEG_HIPO_TERR_CRED]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[173_CUP_ELEG_HIPO_TERR_CRED] (
			[fechaInicioOferta],
			[identificadorProducto],
			[identificador_subproducto],
			[tipo_de_instrumento],
			[Edad_minima],
			[Edad_maxima],
			[Antig_laborar_min],
			[Antig_resid_min],
			[Ingreso_mensual_min],
			[Interpretacion_ing_mens_ual],
			[Caract_especificas_consu_midor],
			[Genero],
			[Publico],
			[Requiere_invitacion],
			[Primer_credito],
			[Score_credito_min],
			[FECHA_INFO]
			)
		SELECT 
			[fechaInicioOferta],
			[identificadorProducto],
			[identificador_subproducto],
			[tipo_de_instrumento],
			[Edad_minima],
			[Edad_maxima],
			[Antig_laborar_min],
			[Antig_resid_min],
			[Ingreso_mensual_min],
			[Interpretacion_ing_mens_ual],
			[Caract_especificas_consu_midor],
			[Genero],
			[Publico],
			[Requiere_invitacion],
			[Primer_credito],
			[Score_credito_min],
			[FECHA_INFO]
		FROM BRONZE.[ORI].[ELEGIBILIDAD_HIPO_CREDITO]
		WHERE 
		FECHA_INFO >= @FechaIni
		AND FECHA_INFO < @FechaFin
		AND tipProd = 'terreno'

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log
	
	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[176_CUP_MORA_HIPO_TERR_CRED]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[176_CUP_MORA_HIPO_TERR_CRED] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[176_CUP_MORA_HIPO_TERR_CRED]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[176_CUP_MORA_HIPO_TERR_CRED]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[176_CUP_MORA_HIPO_TERR_CRED]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[176_CUP_MORA_HIPO_TERR_CRED] (
			[f_ini_oferta],
			[id_producto],
			[id_subproducto],
			[tasa_mora_tipo_tasa],
			[tasa_mora_tasa_referencia],
			[tasa_mora_modo_calcular],
			[tasa_mora_promedio],
			[tasa_mora_come],
			[FECHA_INFO]
			)
		SELECT 
			[f_ini_oferta],
			[id_producto],
			[id_subproducto],
			[tasa_mora_tipo_tasa],
			[tasa_mora_tasa_referencia],
			[tasa_mora_modo_calcular],
			[tasa_mora_promedio],
			[tasa_mora_come],
			[FECHA_INFO]
		FROM BRONZE.[ORI].[MORATORIA_HIPO_CREDITO]
		WHERE 
		FECHA_INFO >= @FechaIni
		AND FECHA_INFO < @FechaFin
		AND tipProd = 'terreno'

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log
	
	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[179_CUP_COM_HIPO_TERR_CRED]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[179_CUP_COM_HIPO_TERR_CRED] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[179_CUP_COM_HIPO_TERR_CRED]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[179_CUP_COM_HIPO_TERR_CRED]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[179_CUP_COM_HIPO_TERR_CRED]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[179_CUP_COM_HIPO_TERR_CRED] (
			[f_ini_oferta],
			[id_producto],
			[id_subproducto],
			[id_operativo],
			[tipo_comision_uso_hip],
			[disponible],
			[operaciones_gratis],
			[moneda],
			[base_cobro],
			[tipo_valor],
			[tipo_periodo],
			[excepciones],
			[valor_comision],
			[comentario],
			[FECHA_INFO]
			)
		SELECT 
			[f_ini_oferta],
			[id_producto],
			[id_subproducto],
			[id_operativo],
			[tipo_comision_uso_hip],
			[disponible],
			[operaciones_gratis],
			[moneda],
			[base_cobro],
			[tipo_valor],
			[tipo_periodo],
			[excepciones],
			[valor_comision],
			[comentario],
			[FECHA_INFO]
		FROM BRONZE.[ORI].[COMISIONES_HIPO_CREDITO]
		WHERE 
		FECHA_INFO >= @FechaIni
		AND FECHA_INFO < @FechaFin
		AND tipProd= 'terreno'

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log
	
	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[182_CUP_PEN_HIPO_TERR_CRED]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[182_CUP_PEN_HIPO_TERR_CRED] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[182_CUP_PEN_HIPO_TERR_CRED]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[182_CUP_PEN_HIPO_TERR_CRED]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[182_CUP_PEN_HIPO_TERR_CRED]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[182_CUP_PEN_HIPO_TERR_CRED] (
			[f_ini_oferta],
			[id_producto],
			[id_subproducto],
			[id_operativo],
			[tipo_comision_pena_hip],
			[disponible],
			[operaciones_gratis],
			[moneda],
			[base_cobro],
			[tipo_valor],
			[tipo_periodo],
			[excepciones],
			[Valor_Comision],
			[comentario],
			[FECHA_INFO]
			)
		SELECT 
			[f_ini_oferta],
			[id_producto],
			[id_subproducto],
			[id_operativo],
			[tipo_comision_pena_hip],
			[disponible],
			[operaciones_gratis],
			[moneda],
			[base_cobro],
			[tipo_valor],
			[tipo_periodo],
			[excepciones],
			[Valor_Comision],
			[comentario],
			[FECHA_INFO]
		FROM BRONZE.[ORI].[PENALIZACIONES_HIPO_CREDITO]
		WHERE 
		FECHA_INFO >= @FechaIni
		AND FECHA_INFO < @FechaFin
		AND tipProd = 'terreno'

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log
	
	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[185_CUP_PEN_HIPO_2_TERR_CRED]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[185_CUP_PEN_HIPO_2_TERR_CRED] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[185_CUP_PEN_HIPO_2_TERR_CRED]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[185_CUP_PEN_HIPO_2_TERR_CRED]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[185_CUP_PEN_HIPO_2_TERR_CRED]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[185_CUP_PEN_HIPO_2_TERR_CRED] (
			[f_ini_oferta],
			[id_producto],
			[id_subproducto],
			[id_operativo],
			[tipo_comision_pena_hip],
			[disponible],
			[operaciones_gratis],
			[moneda],
			[base_cobro],
			[tipo_valor],
			[tipo_periodo],
			[excepciones],
			[Valor_Comision],
			[comentario],
			[FECHA_INFO]
			)
		SELECT 
			[f_ini_oferta],
			[id_producto],
			[id_subproducto],
			[id_operativo],
			[tipo_comision_pena_hip],
			[disponible],
			[operaciones_gratis],
			[moneda],
			[base_cobro],
			[tipo_valor],
			[tipo_periodo],
			[excepciones],
			[Valor_Comision],
			[comentario],
			[FECHA_INFO]
		FROM BRONZE.[ORI].[PENALIZACIONES_HIPO_2_CREDITO]
		WHERE 
		FECHA_INFO >= @FechaIni
		AND FECHA_INFO < @FechaFin
		AND tipProd = 'terreno'

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log
	
	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[191_CUP_CAT_HIPO_TERR_CRED]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[191_CUP_CAT_HIPO_TERR_CRED] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[191_CUP_CAT_HIPO_TERR_CRED]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[191_CUP_CAT_HIPO_TERR_CRED]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[191_CUP_CAT_HIPO_TERR_CRED]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[191_CUP_CAT_HIPO_TERR_CRED] (
			[f_ini_oferta],
			[id_producto],
			[id_subproducto],
			[aforo_min],
			[aforo_max],
			[plazo_min],
			[plazo_max],
			[plazo_tipo_peri],
			[tipo_vivienda],
			[tipo_programa],
			[tasa_ordinaria_tipo_tasa],
			[tasa_ordinaria_tasa_referencia],
			[tasa_ordinaria_modo_calcular],
			[tasa_ordinaria_promedio],
			[tasa_ordinaria_come],
			[CAT_de_publicidad],
			[cat_segmento],
			[cat_come],
			[pago_millar],
			[FECHA_INFO]
			)
		SELECT 
			[f_ini_oferta],
			[id_producto],
			[id_subproducto],
			[aforo_min],
			[aforo_max],
			[plazo_min],
			[plazo_max],
			[plazo_tipo_peri],
			[tipo_vivienda],
			[tipo_programa],
			[tasa_ordinaria_tipo_tasa],
			[tasa_ordinaria_tasa_referencia],
			[tasa_ordinaria_modo_calcular],
			[tasa_ordinaria_promedio],
			[tasa_ordinaria_come],
			[CAT_de_publicidad],
			[cat_segmento],
			[cat_come],
			[pago_millar],
			[FECHA_INFO]
		FROM BRONZE.[ORI].[CAT_HIPO_CREDITO]
		WHERE 
		FECHA_INFO >= @FechaIni
		AND FECHA_INFO < @FechaFin
		AND tipProd = 'terreno'

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log
	
	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[194_EST_EMP_REV_CRED]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[194_EST_EMP_REV_CRED] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[194_EST_EMP_REV_CRED]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[194_EST_EMP_REV_CRED]
				WHERE [FECHA_INFO] >= @FechaIni
					AND [FECHA_INFO] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[194_EST_EMP_REV_CRED]
			WHERE [FECHA_INFO] >= @FechaIni
				AND [FECHA_INFO] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[194_EST_EMP_REV_CRED] (
			[fechaInicioOferta],
			[fechaInicioSubproducto],
			[fechaFinSubproducto],
			[identificadorProducto],
			[Identificador_subproducto],
			[nombre_subproducto],
			[accesorio],
			[numeroReca],
			[destino],
			[arrendamiento],
			[redisponibiliidad],
			[tipoFactoraje],
			[garantia],
			[periodicidad],
			[limiteCreditoMin],
			[limiteCreditoMax],
			[aceptacion],
			[mediosDisposicion],
			[CAT],
			[tipoTasa],
			[referenciaTasa],
			[FECHA_INFO]
			)
		SELECT 
			[fechaInicioOferta],
			[fechaInicioSubproducto],
			[fechaFinSubproducto],
			[identificadorProducto],
			[Identificador_subproducto],
			[nombre_subproducto],
			[accesorio],
			[numeroReca],
			[destino],
			[arrendamiento],
			[redisponibiliidad],
			[tipoFactoraje],
			[garantia],
			[periodicidad],
			[limiteCreditoMin],
			[limiteCreditoMax],
			[aceptacion],
			[mediosDisposicion],
			[CAT],
			[tipoTasa],
			[referenciaTasa],
			[FECHA_INFO]
		FROM BRONZE.[ORI].[EST_EMPRESARIAL_CREDITO]
		WHERE 
		FECHA_INFO >= @FechaIni
		AND FECHA_INFO < @FechaFin
		AND tipProd = 'revolvente'

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log
	
	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[197_ENT_OPERACIONESOEMN]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[197_ENT_OPERACIONESOEMN] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME

AS
BEGIN	
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[197_ENT_OPERACIONESOEMN]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

   BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[197_ENT_OPERACIONESOEMN]
				WHERE [FECHA] >= @FechaIni
					AND [FECHA] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[197_ENT_OPERACIONESOEMN]
			WHERE [FECHA] >= @FechaIni
				AND [FECHA] < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [SILVER].[RR].[197_ENT_OPERACIONESOEMN] (
		[FECHA],
		[CONTRAPARTE],
		[FORMA_OPERACION],
		[TIPO_OPERACION],
		[MEDIO_DISTRIBUCION],
		[TIPO_MEDIO_DISTRIBUCION],
		[TRANSPORTADORA_VALORES],
		[DENOMINACION],
		[IMPORTE],
		[CLABE],
		[TIPO_CUENTA],
		[NIVEL_CUENTA],
		[TIPO_CLIENTE],
		[PAIS_IMPORT_EXPORT],
		[EMPRESA_GESTORA]
		)
		SELECT [FECHA],
		[CONTRAPARTE],
		[FORMA_OPERACION],
		[TIPO_OPERACION],
		[MEDIO_DISTRIBUCION],
		[TIPO_MEDIO_DISTRIBUCION],
		[TRANSPORTADORA_VALORES],
		[DENOMINACION],
		[IMPORTE],
		[CLABE],
		[TIPO_CUENTA],
		[NIVEL_CUENTA],
		[TIPO_CLIENTE],
		[PAIS_IMPORT_EXPORT],
		[EMPRESA_GESTORA]
		FROM [SILVER].[RR].[197_ENT_OPERACIONESOEMN]
		WHERE [FECHA] >= @FechaIni
			AND [FECHA] < @FechaFin


		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END

GO
```

```sql
/****** Object:  StoredProcedure [dbo].[202_ENT_SECCION_2B]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[202_ENT_SECCION_2B] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[202_ENT_SECCION_2B]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

   BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[202_ENT_SECCION_2B]
				WHERE FECHA_INFO >= @FechaIni
					AND FECHA_INFO < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[202_ENT_SECCION_2B]
			WHERE FECHA_INFO >= @FechaIni
				AND FECHA_INFO < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [SILVER].[RR].[202_ENT_SECCION_2B] (
		REPORTE,
		IDPRODUCTO,
		IDCREDITOINTERNO,
		IDCNBV,
		NUMDISPOSICION,
		FONDEOBD,
		FINICIOCREDITO,
		FVENCIMIENTOCREDITO,
		FINICIOLINEA,
		FVENCIMIENTOLINEA,
		MONEDACREDITO,
		MONEDALINEA,
		IDENTIFICACALIFICACION,
		DESTINO,
		INDICADORREESTRUCTURA,
		INCUMPLIMIENTO,
		TITULARDERECHO,
		DERECHOS,
		COMPROMETIDO,
		IDACREDITADO,
		IDACREDITADOCNBV,
		TIPOACREDITADO,
		NOMBREACREDITADO,
		RFC,
		RELACION,
		METODOLOGIA,
		RESERVAS,
		RIESGO,
		ETAPA,
		[PI],
		SI,
		EI,
		INDICADORGARANTIA,
		NUMGARANTIAS,
		TIPOGARANTIA,
		FECHAGARANTIACNBV,
		IMPORTEGARANTIA,
		PORTAFOLIOPASOYMEDIA,
		PASOYMEDIA,
		PORTAFOLIOPPERDIDAS,
		PRIMERASPERDIDAS,
		FULTIMOPAGO,
		MONTOORIGINAL,
		SALDOVIGENTE,
		TASAINTERES,
		FECHA_INFO
		)
		SELECT REPORTE,
		IDPRODUCTO,
		IDCREDITOINTERNO,
		IDCNBV,
		NUMDISPOSICION,
		FONDEOBD,
		FINICIOCREDITO,
		FVENCIMIENTOCREDITO,
		FINICIOLINEA,
		FVENCIMIENTOLINEA,
		MONEDACREDITO,
		MONEDALINEA,
		IDENTIFICACALIFICACION,
		DESTINO,
		INDICADORREESTRUCTURA,
		INCUMPLIMIENTO,
		TITULARDERECHO,
		DERECHOS,
		COMPROMETIDO,
		IDACREDITADO,
		IDACREDITADOCNBV,
		TIPOACREDITADO,
		NOMBREACREDITADO,
		RFC,
		RELACION,
		METODOLOGIA,
		RESERVAS,
		RIESGO,
		ETAPA,
		[PI],
		SI,
		EI,
		INDICADORGARANTIA,
		NUMGARANTIAS,
		TIPOGARANTIA,
		FECHAGARANTIACNBV,
		IMPORTEGARANTIA,
		PORTAFOLIOPASOYMEDIA,
		PASOYMEDIA,
		PORTAFOLIOPPERDIDAS,
		PRIMERASPERDIDAS,
		FULTIMOPAGO,
		MONTOORIGINAL,
		SALDOVIGENTE,
		TASAINTERES,
		FECHA_INFO
		FROM [SILVER].[RR].[202_ENT_SECCION_2B]
		WHERE FECHA_INFO >= @FechaIni
			AND FECHA_INFO < @FechaFin

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END

GO
```

```sql
/****** Object:  StoredProcedure [dbo].[205_ENT_SECCION_2E]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[205_ENT_SECCION_2E] @CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[205_ENT_SECCION_2E]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

   BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[205_ENT_SECCION_2E]
				WHERE FECHA_INFO >= @FechaIni
					AND FECHA_INFO < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[205_ENT_SECCION_2E]
			WHERE FECHA_INFO >= @FechaIni
				AND FECHA_INFO < @FechaFin;

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [SILVER].[RR].[205_ENT_SECCION_2E] (
		IDENTIFICADOR,
		FECHA_CALIFICACION,
		CALIFICACION,
		IDSECCION,
		FECHA_INFO		
		)
		SELECT IDENTIFICADOR,
		FECHA_CALIFICACION,
		CALIFICACION,
		IDSECCION,
		FECHA_INFO	
		FROM [SILVER].[RR].[205_ENT_SECCION_2E]
		WHERE FECHA_INFO >= @FechaIni
			AND FECHA_INFO < @FechaFin

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END

GO
```

```sql
/****** Object:  StoredProcedure [dbo].[208_ENT_CUOTAS_IPAB_DEF_FINAL]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[208_ENT_CUOTAS_IPAB_DEF_FINAL] 
	@CorreoNotificacion NVARCHAR(255) = NULL,
	@PerfilCorreo NVARCHAR(255) = NULL,
	@ProgramadorJob NVARCHAR(128) = NULL,
	@FechaSistema DATETIME
AS
BEGIN
	SET NOCOUNT ON;
	SET TRANSACTION ISOLATION LEVEL READ UNCOMMITTED;

	DECLARE @SQL NVARCHAR(MAX);
	DECLARE @MensajeError NVARCHAR(MAX) = '';
	DECLARE @ExitoEjecucion BIT = 1;
	DECLARE @FilasInsertadas INT = 0;
	DECLARE @LogMessage NVARCHAR(MAX) = '';
	DECLARE @DetallesLog NVARCHAR(MAX) = '';
	DECLARE @FechaInicio DATETIME = GETDATE();
	DECLARE @FilasEliminadas INT = 0;
	DECLARE @NombreJob NVARCHAR(128) = '[208_ENT_CUOTAS_IPAB_DEF_FINAL]';
	DECLARE @FechaIni DATE,
		@FechaFin DATE;

	BEGIN TRY
		SET @FechaIni = datefromparts(year(@FechaSistema), month(@FechaSistema), 1)
		SET @FechaFin = Dateadd(month, 1, @FechaIni) --------------------------------------------------------------------------------
			-- QUERY

		IF EXISTS (
				SELECT ID
				FROM [SILVER].[RR].[208_ENT_CUOTAS_IPAB_DEF_FINAL]
				WHERE [Fecha] >= @FechaIni
					AND [Fecha] < @FechaFin
				)
		BEGIN
			DELETE
			FROM [SILVER].[RR].[208_ENT_CUOTAS_IPAB_DEF_FINAL]
			WHERE [Fecha] >= @FechaIni
				AND [Fecha] < @FechaFin

			SET @FilasEliminadas = @@ROWCOUNT;
			SET @LogMessage = 'Registros eliminados: ' + CAST(@FilasEliminadas AS NVARCHAR(10));

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END;

		INSERT INTO [RR].[208_ENT_CUOTAS_IPAB_DEF_FINAL] (
			[Fecha],
			[CuentaInterna],
			[SaldoFinal],
			[SaldoPromedio],
			[Moneda]
			)
		SELECT 
			IPAB.Fecha AS Fecha,
			IPAB.CuentaInterna AS CuentaInterna,
			IPAB.SaldoFinal AS SaldoFinal,
			IPAB.SaldoPromedio AS SaldoPromedio,
			IPAB.Moneda AS Moneda
		FROM BRONZE.LMDA.ENT_CUOTAS_IPAB_DEF IPAB
		WHERE Fecha >= @FechaIni
				AND Fecha < @FechaFin
				AND BANDERA = 'F'

		---------------------------------------------------------------------------------------------
		SET @FilasInsertadas = @@ROWCOUNT;
		SET @LogMessage = 'Proceso completado. Filas totales: ' + CAST(@FilasInsertadas AS NVARCHAR(10));

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END TRY

	BEGIN CATCH
		SET @ExitoEjecucion = 0;
		SET @MensajeError = ERROR_MESSAGE();
		SET @LogMessage = 'Error durante la ejecución: ' + @MensajeError;

		PRINT @LogMessage;

		SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
	END CATCH -- Preparar el mensaje detallado para la alerta

	DECLARE @Asunto NVARCHAR(255);
	DECLARE @Cuerpo NVARCHAR(MAX);
	DECLARE @FechaFinalizacion DATETIME = GETDATE();
	DECLARE @DuracionEjecucion VARCHAR(20) = CAST(DATEDIFF(SECOND, @FechaInicio, @FechaFinalizacion) AS VARCHAR(10)) + ' segundos';

	-- Solo enviar alerta si hay un error
	IF @ExitoEjecucion = 0
		AND @CorreoNotificacion IS NOT NULL
		AND @PerfilCorreo IS NOT NULL
	BEGIN
		SET @Asunto = 'ALERTA: Error en ' + ISNULL(@NombreJob, 'Job Desconocido');
		SET @Cuerpo = 'Se ha producido un error durante la ejecución de.' + @NombreJob + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles del Job:' + CHAR(13) + CHAR(10) + '- Nombre del Job: ' + ISNULL(@NombreJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Programado por: ' + ISNULL(@ProgramadorJob, 'No especificado') + CHAR(13) + CHAR(10) + '- Fecha y hora de inicio: ' + CONVERT(VARCHAR, @FechaInicio, 120) + CHAR(13) + CHAR(10) + '- Fecha y hora de finalización: ' + CONVERT(VARCHAR, @FechaFinalizacion, 120) + CHAR(13) + CHAR(10) + '- Duración de la ejecución: ' + @DuracionEjecucion + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Detalles de la Ejecución:' + CHAR(13) + CHAR(10) + 'Mensaje de Error:' + CHAR(13) + CHAR(10) + @MensajeError + CHAR(13) + CHAR(10) + CHAR(13) + CHAR(10) + 'Log de Ejecución:' + CHAR(13) + CHAR(10) + @DetallesLog;

		BEGIN TRY
			EXEC msdb.dbo.sp_send_dbmail @profile_name = @PerfilCorreo,
				@recipients = @CorreoNotificacion,
				@subject = @Asunto,
				@body = @Cuerpo,
				@body_format = 'TEXT',
				@importance = 'High';

			SET @LogMessage = 'Alerta de error enviada exitosamente.';

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END TRY

		BEGIN CATCH
			SET @LogMessage = 'Error al enviar alerta: ' + ERROR_MESSAGE();

			PRINT @LogMessage;

			SET @DetallesLog = @DetallesLog + @LogMessage + CHAR(13) + CHAR(10);
		END CATCH
	END -- Registrar en la tabla de log

	INSERT INTO dbo.LogSilverDiario (
		FechaEjecucion,
		FilasInsertadas,
		EstadoEjecucion,
		MensajeError,
		DetallesLog,
		NombreJob,
		ProgramadorJob
		)
	VALUES (
		@FechaInicio,
		@FilasInsertadas,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN 'Exitoso'
			ELSE 'Error'
			END,
		CASE 
			WHEN @ExitoEjecucion = 1
				THEN NULL
			ELSE @MensajeError
			END,
		@DetallesLog,
		@NombreJob,
		@ProgramadorJob
		);

	SET @LogMessage = 'Proceso completado y registrado en la tabla de log.';

	PRINT @LogMessage;
END;
GO
```

# [dbo].[JobSilverDiario]

```sql
/****** Object:  StoredProcedure [dbo].[JobSilverDiario]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[JobSilverDiario]
    @FechaSistema        DATETIME,
    @CorreoNotificacion  NVARCHAR(255),
    @PerfilCorreo        NVARCHAR(255),
    @ProgramadorJob      NVARCHAR(128)
AS
BEGIN
    DECLARE @CurrentSchema    NVARCHAR(128)
    DECLARE @CurrentProc      NVARCHAR(128)
    DECLARE @CurrentFrecuency NVARCHAR(50)
    DECLARE @CurrentActivo    BIT
    DECLARE @SQL              NVARCHAR(MAX)
    DECLARE @ErrorMsg         NVARCHAR(MAX)

    -- Tabla temporal para acumular resultados
    CREATE TABLE #ResultadosEjecucion (
        Orden        INT IDENTITY(1,1),
        Procedimiento NVARCHAR(128),
        Frecuencia   NVARCHAR(50),
        Estado       NVARCHAR(10),   -- 'OK' | 'ERROR'
        Mensaje      NVARCHAR(MAX),
        FechaHora    DATETIME DEFAULT GETDATE()
    )

    DECLARE proc_cursor CURSOR FOR
    SELECT
        'dbo' AS SchemaName,
        CONCAT(FORMAT(numero, '000'), '_', nombre) AS ProcedureName,
        frecuencia AS Frecuency,
        activo
    FROM [ION].dbo.INDICE_REPORTES
    WHERE (
            frecuencia = 'Diaria'
        OR (frecuencia = 'Mensual'       AND DAY(DATEADD(DAY,1,@FechaSistema))=1)
        OR (frecuencia = 'Semanal'       AND DATEPART(WEEKDAY,@FechaSistema)=6)
        OR (frecuencia = 'Bimestral'     AND DAY(DATEADD(DAY,1,@FechaSistema))=1 AND MONTH(@FechaSistema)%2=0)
        OR (frecuencia = 'Trimestral'    AND DAY(DATEADD(DAY,1,@FechaSistema))=1 AND MONTH(@FechaSistema) IN (3,6,9,12))
        OR (frecuencia = 'Cuatrimestral' AND DAY(DATEADD(DAY,1,@FechaSistema))=1 AND MONTH(@FechaSistema) IN (4,8,12))
        OR (frecuencia = 'Semestral'     AND DAY(DATEADD(DAY,1,@FechaSistema))=1 AND MONTH(@FechaSistema) IN (6,12))
        OR (frecuencia = 'Anual'         AND DAY(DATEADD(DAY,1,@FechaSistema))=1 AND MONTH(@FechaSistema) IN (12))
    )
    AND activo = 1
    ORDER BY ProcedureName;

    OPEN proc_cursor
    FETCH NEXT FROM proc_cursor INTO @CurrentSchema, @CurrentProc, @CurrentFrecuency, @CurrentActivo

    WHILE @@FETCH_STATUS = 0
    BEGIN
        SET @SQL =
            'EXEC [' + @CurrentSchema + '].[' + @CurrentProc + '] '
            + '@CorreoNotificacion = ''' + @CorreoNotificacion + ''', '
            + '@PerfilCorreo = '''       + @PerfilCorreo       + ''', '
            + '@ProgramadorJob = '''     + @ProgramadorJob     + ''', '
            + '@FechaSistema = '''       + CONVERT(VARCHAR(23), @FechaSistema, 121) + '''';

        BEGIN TRY
            EXEC sp_executesql @SQL;
            INSERT INTO #ResultadosEjecucion (Procedimiento, Frecuencia, Estado, Mensaje)
            VALUES (@CurrentProc, @CurrentFrecuency, 'OK', 'Ejecutado correctamente')
        END TRY
        BEGIN CATCH
            SET @ErrorMsg = ERROR_MESSAGE()
            INSERT INTO #ResultadosEjecucion (Procedimiento, Frecuencia, Estado, Mensaje)
            VALUES (@CurrentProc, @CurrentFrecuency, 'ERROR', @ErrorMsg)
            PRINT 'Error en ' + @CurrentProc + ': ' + @ErrorMsg
        END CATCH

        FETCH NEXT FROM proc_cursor INTO @CurrentSchema, @CurrentProc, @CurrentFrecuency, @CurrentActivo
    END

    CLOSE proc_cursor
    DEALLOCATE proc_cursor

    -- =============================================
    -- Armar y enviar mail de resumen (solo si hubo
    -- al menos un SP ejecutado)
    -- =============================================
    DECLARE @TotalEjecutados INT
    DECLARE @TotalOK         INT
    DECLARE @TotalError      INT
    DECLARE @Asunto          NVARCHAR(500)
    DECLARE @Cuerpo          NVARCHAR(MAX)
    DECLARE @FilasHTML       NVARCHAR(MAX) = ''
    DECLARE @FechaStr        NVARCHAR(30)  = CONVERT(VARCHAR(30), @FechaSistema, 103)

    SELECT
        @TotalEjecutados = COUNT(*),
        @TotalOK         = SUM(CASE WHEN Estado = 'OK'    THEN 1 ELSE 0 END),
        @TotalError      = SUM(CASE WHEN Estado = 'ERROR' THEN 1 ELSE 0 END)
    FROM #ResultadosEjecucion

    IF @TotalEjecutados = 0
    BEGIN
        DROP TABLE #ResultadosEjecucion
        RETURN  -- Nada que reportar
    END

    -- Asunto dinámico según si hubo errores o no
    SET @Asunto =
        CASE
            WHEN @TotalError > 0
                THEN '⚠️ Silver Diario ' + @FechaStr + ' — ' + CAST(@TotalError AS NVARCHAR) + ' error(es) de ' + CAST(@TotalEjecutados AS NVARCHAR) + ' SPs'
            ELSE
                '✅ Silver Diario ' + @FechaStr + ' — ' + CAST(@TotalEjecutados AS NVARCHAR) + ' SPs ejecutados OK'
        END

    -- Construir filas HTML de la tabla de resultados
    SELECT @FilasHTML = @FilasHTML +
        '<tr style="background:' +
            CASE Estado
                WHEN 'OK'    THEN '#f0faf4'
                ELSE              '#fff5f5'
            END + ';">' +
        '<td style="padding:8px 12px;border-bottom:1px solid #e5e7eb;">' + CAST(Orden AS NVARCHAR) + '</td>' +
        '<td style="padding:8px 12px;border-bottom:1px solid #e5e7eb;font-family:monospace;font-size:13px;">' + Procedimiento + '</td>' +
        '<td style="padding:8px 12px;border-bottom:1px solid #e5e7eb;">' + Frecuencia + '</td>' +
        '<td style="padding:8px 12px;border-bottom:1px solid #e5e7eb;font-weight:600;color:' +
            CASE Estado
                WHEN 'OK'    THEN '#166534'
                ELSE              '#991b1b'
            END + ';">' + Estado + '</td>' +
        '<td style="padding:8px 12px;border-bottom:1px solid #e5e7eb;font-size:12px;color:#6b7280;">' + Mensaje + '</td>' +
        '</tr>'
    FROM #ResultadosEjecucion
    ORDER BY Orden

    -- Cuerpo HTML completo
    SET @Cuerpo = '
    <html><body style="font-family:Arial,sans-serif;font-size:14px;color:#111827;margin:0;padding:24px;">

      <h2 style="margin:0 0 4px;">Silver Diario — Resumen de ejecución</h2>
      <p style="margin:0 0 20px;color:#6b7280;">Fecha: ' + @FechaStr + ' &nbsp;|&nbsp; Programador: ' + @ProgramadorJob + '</p>

      <table style="border-collapse:collapse;margin-bottom:24px;">
        <tr>
          <td style="padding:12px 20px;background:#f9fafb;border:1px solid #e5e7eb;border-radius:6px;text-align:center;">
            <div style="font-size:28px;font-weight:700;">' + CAST(@TotalEjecutados AS NVARCHAR) + '</div>
            <div style="font-size:12px;color:#6b7280;">Total ejecutados</div>
          </td>
          <td style="width:12px;"></td>
          <td style="padding:12px 20px;background:#f0faf4;border:1px solid #bbf7d0;border-radius:6px;text-align:center;">
            <div style="font-size:28px;font-weight:700;color:#166534;">' + CAST(@TotalOK AS NVARCHAR) + '</div>
            <div style="font-size:12px;color:#166534;">Exitosos</div>
          </td>
          <td style="width:12px;"></td>
          <td style="padding:12px 20px;background:#fff5f5;border:1px solid #fecaca;border-radius:6px;text-align:center;">
            <div style="font-size:28px;font-weight:700;color:#991b1b;">' + CAST(@TotalError AS NVARCHAR) + '</div>
            <div style="font-size:12px;color:#991b1b;">Con error</div>
          </td>
        </tr>
      </table>

      <table style="border-collapse:collapse;width:100%;font-size:13px;">
        <thead>
          <tr style="background:#f3f4f6;">
            <th style="padding:8px 12px;text-align:left;border-bottom:2px solid #e5e7eb;">#</th>
            <th style="padding:8px 12px;text-align:left;border-bottom:2px solid #e5e7eb;">Procedimiento</th>
            <th style="padding:8px 12px;text-align:left;border-bottom:2px solid #e5e7eb;">Frecuencia</th>
            <th style="padding:8px 12px;text-align:left;border-bottom:2px solid #e5e7eb;">Estado</th>
            <th style="padding:8px 12px;text-align:left;border-bottom:2px solid #e5e7eb;">Detalle</th>
          </tr>
        </thead>
        <tbody>' + @FilasHTML + '</tbody>
      </table>

      <p style="margin:20px 0 0;font-size:12px;color:#9ca3af;">
        Generado automáticamente por ' + @ProgramadorJob + ' · ' + CONVERT(VARCHAR(30), GETDATE(), 120) + '
      </p>

    </body></html>'

    EXEC msdb.dbo.sp_send_dbmail
        @profile_name  = @PerfilCorreo,
        @recipients    = @CorreoNotificacion,
        @subject       = @Asunto,
        @body          = @Cuerpo,
        @body_format   = 'HTML'

    DROP TABLE #ResultadosEjecucion
END;
GO
```

# [dbo].[usp_IMPERA_Extraer]

```sql
/****** Object:  StoredProcedure [dbo].[usp_IMPERA_Extraer]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- 1. EXTRACTOR
CREATE   PROCEDURE [dbo].[usp_IMPERA_Extraer] (
    @IdLayout INT,
    @FechaInfo DATE,
    @IdCorrida BIGINT OUTPUT
)
AS
BEGIN
    SET NOCOUNT ON;
    PRINT '>> Ejecutando Extractor para LayoutId: ' + CAST(@IdLayout AS VARCHAR);

    -- 1. Insertar en Ejecucion para obtener IdCorrida
    INSERT INTO BRONZE.IMPERA_Aud.Ejecucion (IdLayout, TipoEjecucion, PeriodoIni, PeriodoFin)
    VALUES (@IdLayout, 'LAYOUT_INDIVIDUAL', @FechaInfo, @FechaInfo);
    SET @IdCorrida = SCOPE_IDENTITY();

    -- 2. Determinar tabla destino
    DECLARE @CodigoLayout VARCHAR(50);
    DECLARE @VersionEsquema VARCHAR(10);
    DECLARE @EsExtractDirecto BIT;
    SELECT @CodigoLayout = CodigoLayout, @VersionEsquema = VersionEsquema, @EsExtractDirecto = EsExtractDirecto
    FROM BRONZE.IMPERA_Meta.Layout WHERE IdLayout = @IdLayout;

    DECLARE @TargetTable VARCHAR(200) = 'BRONZE.IMPERA_Hml.[Hml_' + REPLACE(@CodigoLayout, '-', '_') + '_' + @VersionEsquema + ']';
    IF @EsExtractDirecto = 1
        SET @TargetTable = 'BRONZE.IMPERA_Lay.[Lay_' + REPLACE(@CodigoLayout, '-', '_') + '_' + @VersionEsquema + ']';

    -- 3. Construir SELECT (Agregando IdEjecucion)
    DECLARE @Select NVARCHAR(MAX) = 'SELECT ' + CAST(@IdCorrida AS NVARCHAR(20)) + ' AS [IdEjecucion], ';
    SELECT @Select = @Select + string_agg(
        ISNULL(mo.ColumnaOrigen, 'NULL') + ' AS [' + c.NombreCampo + ']', 
        ', '
    )
    FROM BRONZE.IMPERA_Meta.LayoutCampo c
    LEFT JOIN BRONZE.IMPERA_Meta.MapeoOrigen mo ON c.IdCampo = mo.IdCampo
    WHERE c.IdLayout = @IdLayout;

    -- 4. Construir FROM
    DECLARE @From NVARCHAR(MAX) = ' FROM ';
    SELECT @From = @From + 
        CASE WHEN o.TipoJoin = 'FROM' THEN '' ELSE ' ' + o.TipoJoin + ' JOIN ' END + 
        'BRONZE.' + ISNULL(o.SchemaOrigen, '') + '.' + o.TablaOrigen + ' ' + o.Alias
    FROM BRONZE.IMPERA_Meta.OrigenDato o
    WHERE o.IdLayout = @IdLayout
    ORDER BY o.Orden;

    -- 5. Construir WHERE
    DECLARE @Where NVARCHAR(MAX) = ' ';
    SELECT @Where = @Where + ISNULL(string_agg(' AND ' + f.ExpresionSql, ''), '')
    FROM BRONZE.IMPERA_Meta.OrigenFiltro f
    WHERE f.IdLayout = @IdLayout;

    IF @Where <> ' '
        SET @Where = ' WHERE 1=1 ' + @Where;

    -- 6. Armar Query Final
    DECLARE @SqlFinal NVARCHAR(MAX) = 'INSERT INTO ' + @TargetTable + ' ([IdEjecucion], ' + 
        (SELECT string_agg('[' + NombreCampo + ']', ', ') FROM BRONZE.IMPERA_Meta.LayoutCampo WHERE IdLayout = @IdLayout) +
        ') ' + @Select + @From + @Where;

    PRINT '--- QUERY GENERADO ---';
    PRINT @SqlFinal;

    EXEC sp_executesql @SqlFinal;
    
    PRINT '>> Extracción completada.';
END;
GO
```

# [dbo].[usp_IMPERA_GenerarTxt]

```sql
/****** Object:  StoredProcedure [dbo].[usp_IMPERA_GenerarTxt]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- 5. GENERADOR TXT
CREATE   PROCEDURE [dbo].[usp_IMPERA_GenerarTxt] (
    @IdCorrida BIGINT
)
AS
BEGIN
    SET NOCOUNT ON;
    PRINT '>> Ejecutando Generador TXT para IdCorrida: ' + CAST(@IdCorrida AS VARCHAR);

    DECLARE @IdLayout INT;
    SELECT @IdLayout = IdLayout FROM BRONZE.IMPERA_Aud.Ejecucion WHERE IdEjecucion = @IdCorrida;

    DECLARE @CodigoLayout VARCHAR(50);
    DECLARE @VersionEsquema VARCHAR(10);
    DECLARE @EsExtractDirecto BIT;
    DECLARE @SeparadorCampos CHAR(1);
    DECLARE @NombreArchivo NVARCHAR(300);
    DECLARE @IncluirCabecera BIT;

    SELECT @CodigoLayout = CodigoLayout, @VersionEsquema = VersionEsquema, @EsExtractDirecto = EsExtractDirecto, 
           @SeparadorCampos = SeparadorCampos, @NombreArchivo = ISNULL(PatronNombreArchivo, CodigoLayout + '_' + CAST(@IdCorrida AS VARCHAR) + '.txt'),
           @IncluirCabecera = IncluirCabecera
    FROM BRONZE.IMPERA_Meta.Layout WHERE IdLayout = @IdLayout;

    DECLARE @TargetTable VARCHAR(200) = 'BRONZE.IMPERA_Hml.[Hml_' + REPLACE(@CodigoLayout, '-', '_') + '_' + @VersionEsquema + ']';
    IF @EsExtractDirecto = 1
        SET @TargetTable = 'BRONZE.IMPERA_Lay.[Lay_' + REPLACE(@CodigoLayout, '-', '_') + '_' + @VersionEsquema + ']';

    -- Construir el string_agg de columnas para DATOS
    DECLARE @Cols NVARCHAR(MAX) = '';
    SELECT @Cols = @Cols + 
        'ISNULL(CAST([' + NombreCampo + '] AS VARCHAR(MAX)), '''') + ''' + @SeparadorCampos + ''' + '
    FROM BRONZE.IMPERA_Meta.LayoutCampo
    WHERE IdLayout = @IdLayout
    ORDER BY Posicion;

    IF LEN(@Cols) > 0
        SET @Cols = LEFT(@Cols, LEN(@Cols) - LEN(' + ''' + @SeparadorCampos + ''' + '));

    -- Cabeceras
    DECLARE @Header NVARCHAR(MAX) = '';
    IF @IncluirCabecera = 1
    BEGIN
        SELECT @Header = @Header + NombreCampo + @SeparadorCampos
        FROM BRONZE.IMPERA_Meta.LayoutCampo
        WHERE IdLayout = @IdLayout
        ORDER BY Posicion;
        
        IF LEN(@Header) > 0
            SET @Header = LEFT(@Header, LEN(@Header) - 1) + CHAR(13) + CHAR(10);
    END

    DECLARE @Sql NVARCHAR(MAX) = '
        DECLARE @Txt NVARCHAR(MAX);
        SELECT @Txt = STRING_AGG(CAST(' + @Cols + ' AS NVARCHAR(MAX)), CHAR(13)+CHAR(10))
        FROM ' + @TargetTable + ' WHERE IdEjecucion = ' + CAST(@IdCorrida AS VARCHAR) + ';
        
        IF @Txt IS NOT NULL
        BEGIN
            SET @Txt = ''' + @Header + ''' + @Txt;
        END

        INSERT INTO BRONZE.IMPERA_Txt.ArchivoGenerado (IdEjecucion, IdLayout, NombreArchivo, Encoding, Contenido, HashSHA256, FilasDatos)
        VALUES (' + CAST(@IdCorrida AS VARCHAR) + ', ' + CAST(@IdLayout AS VARCHAR) + ', ''' + @NombreArchivo + ''', ''UTF-8'', ISNULL(@Txt, ''''), HASHBYTES(''SHA2_256'', ISNULL(@Txt, '''')), (SELECT COUNT(1) FROM ' + @TargetTable + ' WHERE IdEjecucion = ' + CAST(@IdCorrida AS VARCHAR) + '));
    ';

    PRINT '--- QUERY GENERADO TXT ---';
    PRINT @Sql;

    EXEC sp_executesql @Sql;
    
    PRINT '>> Generación TXT completada.';
END;
GO
```

# [dbo].[usp_IMPERA_Homologar]

```sql
/****** Object:  StoredProcedure [dbo].[usp_IMPERA_Homologar]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- 2. HOMOLOGADOR (Aqui aplicamos nuestros Catalogos y Mapeos)
CREATE   PROCEDURE [dbo].[usp_IMPERA_Homologar] (
    @IdCorrida BIGINT
)
AS
BEGIN
    SET NOCOUNT ON;
    PRINT '>> Ejecutando Homologador para IdCorrida: ' + CAST(@IdCorrida AS VARCHAR);

    DECLARE @IdLayout INT;
    SELECT @IdLayout = IdLayout FROM BRONZE.IMPERA_Aud.Ejecucion WHERE IdEjecucion = @IdCorrida;

    DECLARE @CodigoLayout VARCHAR(50);
    DECLARE @VersionEsquema VARCHAR(10);
    SELECT @CodigoLayout = CodigoLayout, @VersionEsquema = VersionEsquema
    FROM BRONZE.IMPERA_Meta.Layout WHERE IdLayout = @IdLayout;

    DECLARE @TargetTable VARCHAR(200) = 'BRONZE.IMPERA_Hml.[Hml_' + REPLACE(@CodigoLayout, '-', '_') + '_' + @VersionEsquema + ']';

    DECLARE @SetClause NVARCHAR(MAX) = '';
    SELECT @SetClause = @SetClause + ISNULL(string_agg(
        '[' + c.NombreCampo + '] = ' + 
        CASE 
            WHEN r.TipoRegla = 'VALOR_FIJO' THEN '''' + ISNULL(r.ValorFijo, '') + ''''
            WHEN r.TipoRegla = 'CALCULADO' THEN ISNULL(r.ExpresionSql, 'NULL')
            ELSE 'NULL'
        END,
        ', '
    ), '')
    FROM BRONZE.IMPERA_Meta.HomologacionRegla r
    JOIN BRONZE.IMPERA_Meta.LayoutCampo c ON r.IdCampo = c.IdCampo
    WHERE c.IdLayout = @IdLayout AND r.TipoRegla IN ('VALOR_FIJO', 'CALCULADO');

    IF @SetClause <> ''
    BEGIN
        DECLARE @SqlUpdate NVARCHAR(MAX) = 'UPDATE ' + @TargetTable + ' SET ' + @SetClause + ' WHERE IdEjecucion = ' + CAST(@IdCorrida AS VARCHAR);
        PRINT '--- QUERY GENERADO FIJOS/CALCULADOS ---';
        PRINT @SqlUpdate;
        EXEC sp_executesql @SqlUpdate;
    END
    ELSE
    BEGIN
        PRINT 'No hay reglas de homologacion directas (VALOR_FIJO/CALCULADO) para este layout.';
    END

    -- HOMOLOGAR: LOOKUP_CATALOGO
    DECLARE @SqlUpdateCatalogo NVARCHAR(MAX) = '';
    SELECT @SqlUpdateCatalogo = @SqlUpdateCatalogo + 
        'UPDATE Hml SET [' + c.NombreCampo + '] = C.Codigo ' +
        'FROM ' + @TargetTable + ' Hml ' +
        'JOIN BRONZE.IMPERA_Meta.Catalogo C ON Hml.[' + c.NombreCampo + '] = C.' + ISNULL(r.ExpresionSql, 'Atributo1') + ' ' +
        'WHERE C.OrigenTabla = ''' + r.CatalogoCodigo + ''' AND Hml.IdEjecucion = ' + CAST(@IdCorrida AS VARCHAR) + '; ' + CHAR(13) + CHAR(10)
    FROM BRONZE.IMPERA_Meta.HomologacionRegla r
    JOIN BRONZE.IMPERA_Meta.LayoutCampo c ON r.IdCampo = c.IdCampo
    WHERE c.IdLayout = @IdLayout AND r.TipoRegla = 'LOOKUP_CATALOGO';

    IF @SqlUpdateCatalogo <> ''
    BEGIN
        PRINT '--- QUERY GENERADO LOOKUP CATALOGOS ---';
        PRINT @SqlUpdateCatalogo;
        EXEC sp_executesql @SqlUpdateCatalogo;
    END
    ELSE
    BEGIN
        PRINT 'No hay reglas de catálogo (LOOKUP_CATALOGO) para este layout.';
    END
    
    PRINT '>> Homologación completada.';
END;
GO
```

# [dbo].[usp_IMPERA_Motor_Ejecutar]

```sql
/****** Object:  StoredProcedure [dbo].[usp_IMPERA_Motor_Ejecutar]    Script Date: 01/06/2026 20:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- 6. ORQUESTADOR MAESTRO
CREATE   PROCEDURE [dbo].[usp_IMPERA_Motor_Ejecutar] (
    @CodigoLayout VARCHAR(50),
    @FechaInfo DATE
)
AS
BEGIN
    DECLARE @IdLayout INT;
    SELECT @IdLayout = IdLayout FROM BRONZE.IMPERA_Meta.Layout WHERE CodigoLayout = @CodigoLayout;
    
    IF @IdLayout IS NULL THROW 50000, 'Layout no existe', 1;
    
    DECLARE @IdCorrida BIGINT;
    
    EXEC dbo.usp_IMPERA_Extraer @IdLayout, @FechaInfo, @IdCorrida OUTPUT;
    EXEC dbo.usp_IMPERA_Homologar @IdCorrida;
    EXEC dbo.usp_IMPERA_Validar @IdCorrida;
    EXEC dbo.usp_IMPERA_GenerarDataset @IdCorrida;
    EXEC dbo.usp_IMPERA_GenerarTxt @IdCorrida;
    
    PRINT 'Ejecución completada con éxito. IdCorrida: ' + CAST(@IdCorrida AS VARCHAR);
END;
GO
```
